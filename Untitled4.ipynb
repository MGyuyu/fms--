```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import time

# 计算三点角度函数
def calculate_angle(a, b, c):
    """计算三点之间的角度"""
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle

    return angle

def identify_raised_leg(landmarks, mp_pose):
    """确定哪条腿是抬起的腿"""
    left_ankle_y = landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y
    right_ankle_y = landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y

    # Y坐标小的腿（在屏幕上更高）被视为抬起的腿
    is_right_leg_raised = right_ankle_y < left_ankle_y

    return is_right_leg_raised

def calculate_leg_angle(landmarks, mp_pose, is_right_leg_raised):
    """计算抬起腿的脚踝-膝盖-髋部角度"""
    if is_right_leg_raised:
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
        knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
    else:
        ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]

    # 计算膝盖角度
    angle = calculate_angle(ankle, knee, hip)

    return angle

def score_trunk_rotation(landmarks, mp_pose):
    """
    评估转动稳定动作
    - 3分: 腿部角度 > 160度
    - 2分: 腿部角度在130-160度之间
    - 1分: 腿部角度 < 130度
    """
    # 确定抬起的腿
    is_right_leg_raised = identify_raised_leg(landmarks, mp_pose)

    # 计算腿部角度
    leg_angle = calculate_leg_angle(landmarks, mp_pose, is_right_leg_raised)

    # 评分逻辑
    if leg_angle > 160:
        score = 3
        feedback = f"Excellent! Leg angle: {leg_angle:.1f}° (>160°)"
    elif leg_angle >= 130:
        score = 2
        feedback = f"Good rotation. Leg angle: {leg_angle:.1f}° (130-160°)"
    else:
        score = 1
        feedback = f"Limited rotation. Leg angle: {leg_angle:.1f}° (<130°)"

    return {
        'score': score,
        'feedback': feedback,
        'leg_angle': leg_angle,
        'is_right_leg_raised': is_right_leg_raised
    }

# 新增性能指标评估函数
def evaluate_performance(input_folder, output_folder, results):
    """计算并输出四个关键性能指标"""
    # 1. 检测成功率计算
    total_images = len([f for f in os.listdir(input_folder)
                      if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    detected_images = len(results)
    detection_success_rate = (detected_images / total_images) * 100 if total_images > 0 else 0

    # 2. 混淆矩阵分析 (使用文件名第一个字符作为专家打分)
    confusion_matrix = np.zeros((3, 3), dtype=int)  # 3×3矩阵对应1-3分

    for result in results:
        file_name = result['file']
        system_score = result['score'] - 1  # 转为0-2索引

        # 使用文件名第一个字符作为专家分数
        try:
            expert_score = int(file_name[0]) - 1  # 假设第一个字符是1-3的数字
            if 0 <= expert_score <= 2:  # 确保在有效范围内
                confusion_matrix[system_score, expert_score] += 1
        except (ValueError, IndexError):
            # 如果文件名格式不符合要求，跳过
            continue

    # 3. 端到端处理时间计算
    avg_processing_time = sum([r.get('processing_time', 0) for r in results]) / len(results) if results else 0

    # 4. 分数频率分析
    score_counts = {1: 0, 2: 0, 3: 0}
    for r in results:
        score = r['score']
        if score in score_counts:
            score_counts[score] += 1

    # 输出结果
    print("\n性能指标评估结果:")
    print("-" * 40)

    print(f"1. 检测成功率: {detection_success_rate:.2f}%")
    print(f"   总图像数: {total_images}")
    print(f"   成功检测数: {detected_images}")

    print("\n2. 混淆矩阵分析:")
    print("   系统评分 vs 专家评分 (专家评分取自文件名首字符)")
    print("   专家评分")
    print("              1分    2分    3分")
    print(f"   系统 1分:  {confusion_matrix[0, 0]}     {confusion_matrix[0, 1]}     {confusion_matrix[0, 2]}")
    print(f"   系统 2分:  {confusion_matrix[1, 0]}     {confusion_matrix[1, 1]}     {confusion_matrix[1, 2]}")
    print(f"   系统 3分:  {confusion_matrix[2, 0]}     {confusion_matrix[2, 1]}     {confusion_matrix[2, 2]}")

    # 计算总体准确率
    total_matches = sum([confusion_matrix[i, i] for i in range(3)])
    total_samples = np.sum(confusion_matrix)
    accuracy = (total_matches / total_samples) * 100 if total_samples > 0 else 0
    print(f"   评分准确率: {accuracy:.2f}%")

    print(f"\n3. 端到端处理时间: {avg_processing_time:.4f} 秒/图像")

    print("\n4. 分数频率分析:")
    total_scores = sum(score_counts.values())
    for score, count in score_counts.items():
        percentage = (count / total_scores) * 100 if total_scores > 0 else 0
        print(f"   {score}分: {count} 张图像 ({percentage:.2f}%)")

    # 保存指标结果到文件
    with open(os.path.join(output_folder, "performance_metrics.txt"), 'w') as f:
        f.write("性能指标评估结果\n")
        f.write("-" * 40 + "\n")
        f.write(f"1. 检测成功率: {detection_success_rate:.2f}%\n")
        f.write(f"   总图像数: {total_images}\n")
        f.write(f"   成功检测数: {detected_images}\n\n")

        f.write("2. 混淆矩阵分析:\n")
        f.write("   专家评分\n")
        f.write("              1分    2分    3分\n")
        f.write(f"   系统 1分:  {confusion_matrix[0, 0]}     {confusion_matrix[0, 1]}     {confusion_matrix[0, 2]}\n")
        f.write(f"   系统 2分:  {confusion_matrix[1, 0]}     {confusion_matrix[1, 1]}     {confusion_matrix[1, 2]}\n")
        f.write(f"   系统 3分:  {confusion_matrix[2, 0]}     {confusion_matrix[2, 1]}     {confusion_matrix[2, 2]}\n")
        f.write(f"   评分准确率: {accuracy:.2f}%\n\n")

        f.write(f"3. 端到端处理时间: {avg_processing_time:.4f} 秒/图像\n\n")

        f.write("4. 分数频率分析:\n")
        for score, count in score_counts.items():
            percentage = (count / total_scores) * 100 if total_scores > 0 else 0
            f.write(f"   {score}分: {count} 张图像 ({percentage:.2f}%)\n")

    # 绘制并保存混淆矩阵可视化
    plt.figure(figsize=(8, 6))
    plt.imshow(confusion_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('混淆矩阵')
    plt.colorbar()
    tick_marks = np.arange(3)
    plt.xticks(tick_marks, ['1分', '2分', '3分'], rotation=45)
    plt.yticks(tick_marks, ['1分', '2分', '3分'])
    plt.xlabel('专家评分')
    plt.ylabel('系统评分')

    # 在混淆矩阵中添加数值标注
    thresh = confusion_matrix.max() / 2.
    for i in range(confusion_matrix.shape[0]):
        for j in range(confusion_matrix.shape[1]):
            plt.text(j, i, format(confusion_matrix[i, j], 'd'),
                   horizontalalignment="center",
                   color="white" if confusion_matrix[i, j] > thresh else "black")

    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, "confusion_matrix.png"))

    # 绘制并保存分数分布图
    plt.figure(figsize=(8, 6))
    plt.bar(['1分', '2分', '3分'], [score_counts[1], score_counts[2], score_counts[3]])
    plt.title('分数频率分布')
    plt.xlabel('评分')
    plt.ylabel('图像数量')
    plt.savefig(os.path.join(output_folder, "score_distribution.png"))

# 视频处理函数
def process_video(input_path, output_path):
    """处理视频并进行姿态分析"""
    # 打开视频文件
    cap = cv2.VideoCapture(input_path)
    if not cap.isOpened():
        print(f"错误：无法打开视频文件 {input_path}")
        return

    # 获取视频属性
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"视频信息: 宽={width}, 高={height}, FPS={fps}, 总帧数={total_frames}")

    # 创建视频写入器
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # 初始化变量
    frame_count = 0
    pTime = 0
    angles_history = []  # 存储角度数据

    # 创建姿态检测模型
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    with mp_pose.Pose(
        static_image_mode=False,
        smooth_landmarks=True,
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5
    ) as pose:
        # 处理每一帧
        while cap.isOpened():
            success, img = cap.read()

            if not success:
                print("视频处理完成或读取错误")
                break

            # 将BGR格式转为RGB格式
            imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # 处理图像
            start_time = time.time()  # 开始计时
            results = pose.process(imgRGB)
            processing_time = time.time() - start_time  # 计算处理时间

            # 在原图上绘制姿态信息
            if results.pose_landmarks:
                # 绘制关键点和连接线
                mp_drawing.draw_landmarks(
                    img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(255,0,0), thickness=2, circle_radius=2),
                    mp_drawing.DrawingSpec(color=(0,255,0), thickness=2, circle_radius=2)
                )

                # 应用转动稳定性评分函数
                evaluation = score_trunk_rotation(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                feedback = evaluation['feedback']
                leg_angle = evaluation['leg_angle']
                is_right_leg_raised = evaluation['is_right_leg_raised']

                # 在图像上显示评分和反馈
                cv2.putText(img, f"Score: {score}/3", (10, 60),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2, cv2.LINE_AA)
                cv2.putText(img, feedback, (10, 90),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2, cv2.LINE_AA)

                # 存储角度数据
                current_angles = {
                    'frame': frame_count,
                    'leg_angle': leg_angle,
                    'score': score,
                    'processing_time': processing_time
                }
                angles_history.append(current_angles)

            # 显示进度
            frame_count += 1
            if frame_count % 30 == 0:  # 每30帧显示一次进度
                print(f"处理进度: {frame_count}/{total_frames} ({frame_count/total_frames*100:.1f}%)")

            # 计算和显示FPS
            cTime = time.time()
            if pTime > 0:
                fps_current = 1/(cTime-pTime)
                cv2.putText(img, f"FPS: {int(fps_current)}", (10, 30),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
            pTime = cTime

            # 写入输出视频
            out.write(img)

        # 释放资源
        cap.release()
        out.release()

    print(f"视频处理完成. 处理了 {frame_count} 帧.")

    # 返回分析结果
    return angles_history

def process_image_folder(input_folder, output_folder):
    """处理文件夹中的所有图像"""
    # 创建输出文件夹
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 设置MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # 获取图像列表
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []

    # 处理每张图像
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            start_time = time.time()  # 开始计时

            # 读取图像
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # 处理图像
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # 计算处理时间
            processing_time = time.time() - start_time

            # 检查是否检测到姿势
            if results.pose_landmarks:
                # 获取图像尺寸
                orig_height, orig_width, _ = img.shape

                # 创建更大的画布
                new_width = orig_width * 2
                new_height = orig_height * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)

                # 绘制姿势关键点
                mp_drawing.draw_landmarks(img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # 评分
                evaluation = score_trunk_rotation(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                feedback = evaluation['feedback']
                leg_angle = evaluation['leg_angle']
                is_right_leg_raised = evaluation['is_right_leg_raised']

                # 记录分数和处理时间
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'leg_angle': leg_angle,
                    'processing_time': processing_time
                })

                # 将原始图像放在左上角
                new_img[0:orig_height, 0:orig_width] = img

                # 在原始图像上绘制角度线
                landmarks = results.pose_landmarks.landmark

                if is_right_leg_raised:
                    ankle = (int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x * orig_width),
                            int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y * orig_height))
                    knee = (int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x * orig_width),
                           int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y * orig_height))
                    hip = (int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x * orig_width),
                          int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y * orig_height))
                    leg_label = "Right Leg (Raised)"
                else:
                    ankle = (int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x * orig_width),
                            int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y * orig_height))
                    knee = (int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x * orig_width),
                           int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y * orig_height))
                    hip = (int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x * orig_width),
                          int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y * orig_height))
                    leg_label = "Left Leg (Raised)"

                # 绘制线条
                cv2.line(new_img, ankle, knee, (0, 255, 255), 2)  # 脚踝到膝盖
                cv2.line(new_img, knee, hip, (0, 255, 255), 2)   # 膝盖到髋部

                # 添加角度文字
                cv2.putText(new_img, f"{leg_angle:.1f}°", (knee[0] + 10, knee[1] - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2, cv2.LINE_AA)

                # 添加评分信息
                score_box = np.zeros((orig_height, orig_width, 3), dtype=np.uint8)
                cv2.rectangle(score_box, (0, 0), (orig_width, orig_height), (30, 30, 30), -1)

                # 添加标题和分数
                cv2.putText(score_box, "Trunk Rotation Test", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, cv2.LINE_AA)

                cv2.putText(score_box, f"Testing: {leg_label}", (20, 80),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(score_box, f"SCORE: {score}/3", (20, 130),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.5, score_color, 2, cv2.LINE_AA)

                # 添加处理时间信息
                cv2.putText(score_box, f"Processing time: {processing_time:.4f}s", (20, 170),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)

                # 添加反馈
                cv2.putText(score_box, feedback, (20, 210),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 165, 255), 2, cv2.LINE_AA)

                # 添加评分标准
                cv2.putText(score_box, "Scoring Criteria:", (20, 260),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

                cv2.putText(score_box, "3 pts: Leg angle > 160°", (20, 300),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 1, cv2.LINE_AA)

                cv2.putText(score_box, "2 pts: Leg angle 130-160°", (20, 340),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 200, 200), 1, cv2.LINE_AA)

                cv2.putText(score_box, "1 pt: Leg angle < 130°", (20, 380),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 1, cv2.LINE_AA)

                # 将评分框放在右上角
                new_img[0:orig_height, orig_width:new_width] = score_box

                # 添加简化的视觉指南
                vis_box = np.zeros((orig_height, new_width, 3), dtype=np.uint8)
                cv2.rectangle(vis_box, (0, 0), (new_width, orig_height), (30, 30, 30), -1)

                # 添加标题
                cv2.putText(vis_box, "Trunk Rotation: Measured Angle", (new_width//2 - 180, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2, cv2.LINE_AA)

                # 绘制示例视觉指南
                center_x = new_width // 2

                # 绘制示例1 - 3分 (>160°)
                ex1_x = center_x - 300
                ex1_y = 150
                hip1_x, hip1_y = ex1_x, ex1_y
                knee1_x, knee1_y = hip1_x, hip1_y + 60
                ankle1_x, ankle1_y = knee1_x + 10, knee1_y + 100 # 165度角

                cv2.circle(vis_box, (hip1_x, hip1_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (knee1_x, knee1_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (ankle1_x, ankle1_y), 5, (255, 255, 255), -1)

                cv2.line(vis_box, (hip1_x, hip1_y), (knee1_x, knee1_y), (0, 255, 0), 2)
                cv2.line(vis_box, (knee1_x, knee1_y), (ankle1_x, ankle1_y), (0, 255, 0), 2)

                cv2.putText(vis_box, "Score 3 (>160°)", (hip1_x - 70, hip1_y - 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2, cv2.LINE_AA)

                # 绘制示例2 - 2分 (130-160°)
                ex2_x = center_x
                ex2_y = 150
                hip2_x, hip2_y = ex2_x, ex2_y
                knee2_x, knee2_y = hip2_x, hip2_y + 60
                ankle2_x, ankle2_y = knee2_x + 40, knee2_y + 90 # 约145度角

                cv2.circle(vis_box, (hip2_x, hip2_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (knee2_x, knee2_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (ankle2_x, ankle2_y), 5, (255, 255, 255), -1)

                cv2.line(vis_box, (hip2_x, hip2_y), (knee2_x, knee2_y), (0, 200, 200), 2)
                cv2.line(vis_box, (knee2_x, knee2_y), (ankle2_x, ankle2_y), (0, 200, 200), 2)

                cv2.putText(vis_box, "Score 2 (130-160°)", (hip2_x - 70, hip2_y - 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 200, 200), 2, cv2.LINE_AA)

                # 绘制示例3 - 1分 (<130°)
                ex3_x = center_x + 300
                ex3_y = 150
                hip3_x, hip3_y = ex3_x, ex3_y
                knee3_x, knee3_y = hip3_x, hip3_y + 60
                ankle3_x, ankle3_y = knee3_x + 60, knee3_y + 60 # 约120度角

                cv2.circle(vis_box, (hip3_x, hip3_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (knee3_x, knee3_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (ankle3_x, ankle3_y), 5, (255, 255, 255), -1)

                cv2.line(vis_box, (hip3_x, hip3_y), (knee3_x, knee3_y), (0, 0, 255), 2)
                cv2.line(vis_box, (knee3_x, knee3_y), (ankle3_x, ankle3_y), (0, 0, 255), 2)

                cv2.putText(vis_box, "Score 1 (<130°)", (hip3_x - 70, hip3_y - 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2, cv2.LINE_AA)

                # 将视觉指南放在底部
                new_img[orig_height:new_height, 0:new_width] = vis_box

                # 保存图像
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

            else:
                # 未检测到姿势
                # 未检测到姿势
                print(f"No pose detected in {img_file}")
                # 创建错误图像
                new_width = img.shape[1] * 2
                new_height = img.shape[0] * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)
                new_img[0:img.shape[0], 0:img.shape[1]] = img
                cv2.putText(new_img, "NO POSE DETECTED", (new_width//4, new_height//4),
                          cv2.FONT_HERSHEY_SIMPLEX, 2.0, (0, 0, 255), 3, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

    # 评估性能指标
    evaluate_performance(input_folder, output_folder, score_results)

    # 生成角度分布图
    if score_results:
        # 计算统计数据
        scores = [r['score'] for r in score_results]
        angles = [r['leg_angle'] for r in score_results]
        processing_times = [r['processing_time'] for r in score_results]
        avg_score = sum(scores) / len(scores)
        avg_angle = sum(angles) / len(angles)
        avg_processing_time = sum(processing_times) / len(processing_times)

        # 创建角度分布图
        plt.figure(figsize=(10, 6))
        plt.hist(angles, bins=10, color='blue', alpha=0.7)
        plt.title('Leg Angle Distribution')
        plt.xlabel('Angle (degrees)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.axvline(x=130, color='orange', linestyle='--', label='130° (Score 1-2 boundary)')
        plt.axvline(x=160, color='green', linestyle='--', label='160° (Score 2-3 boundary)')
        plt.legend()
        angle_chart_path = os.path.join(output_folder, "angle_distribution.png")
        plt.savefig(angle_chart_path)

        # 创建处理时间分布图
        plt.figure(figsize=(10, 6))
        plt.hist(processing_times, bins=10, color='green', alpha=0.7)
        plt.title('Processing Time Distribution')
        plt.xlabel('Processing Time (seconds)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.axvline(x=avg_processing_time, color='red', linestyle='--',
                   label=f'Average: {avg_processing_time:.4f}s')
        plt.legend()
        time_chart_path = os.path.join(output_folder, "processing_time_distribution.png")
        plt.savefig(time_chart_path)

        print(f"Charts saved to {output_folder}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# 示例用法
if __name__ == "__main__":
    # 图像处理
    input_folder = "r"
    output_folder = "r1"
    results = process_image_folder(input_folder, output_folder)

    # 视频处理（可选）
    # video_input = "test2.mp4"
    # video_output = "output_video2.mp4"
    # angles_history = process_video(video_input, video_output)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import time

# 计算三点角度函数
def calculate_angle(a, b, c):
    """计算三点之间的角度"""
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle

    return angle

def identify_raised_leg(landmarks, mp_pose):
    """确定哪条腿是抬起的腿"""
    left_ankle_y = landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y
    right_ankle_y = landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y

    # Y坐标小的腿（在屏幕上更高）被视为抬起的腿
    is_right_leg_raised = right_ankle_y < left_ankle_y

    return is_right_leg_raised

def calculate_leg_angle(landmarks, mp_pose, is_right_leg_raised):
    """计算抬起腿的脚踝-膝盖-髋部角度"""
    if is_right_leg_raised:
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
        knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
    else:
        ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]

    # 计算膝盖角度
    angle = calculate_angle(ankle, knee, hip)

    return angle

def score_trunk_rotation(landmarks, mp_pose):
    """
    评估转动稳定动作
    - 3分: 腿部角度 > 160度
    - 2分: 腿部角度在130-160度之间
    - 1分: 腿部角度 < 130度
    """
    # 确定抬起的腿
    is_right_leg_raised = identify_raised_leg(landmarks, mp_pose)

    # 计算腿部角度
    leg_angle = calculate_leg_angle(landmarks, mp_pose, is_right_leg_raised)

    # 评分逻辑
    if leg_angle > 160:
        score = 3
        feedback = f"Excellent! Leg angle: {leg_angle:.1f} deg (>160 deg)"
    elif leg_angle >= 130:
        score = 2
        feedback = f"Good rotation. Leg angle: {leg_angle:.1f} deg (130-160 deg)"
    else:
        score = 1
        feedback = f"Limited rotation. Leg angle: {leg_angle:.1f} deg (<130 deg)"

    return {
        'score': score,
        'feedback': feedback,
        'leg_angle': leg_angle,
        'is_right_leg_raised': is_right_leg_raised
    }

# 性能指标评估函数
def evaluate_performance(input_folder, output_folder, results):
    """计算并输出四个关键性能指标"""
    # 1. 检测成功率计算
    total_images = len([f for f in os.listdir(input_folder)
                      if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    detected_images = len(results)
    detection_success_rate = (detected_images / total_images) * 100 if total_images > 0 else 0

    # 2. 混淆矩阵分析 (使用文件名第一个字符作为专家打分)
    confusion_matrix = np.zeros((3, 3), dtype=int)  # 3×3矩阵对应1-3分

    for result in results:
        file_name = result['file']
        system_score = result['score'] - 1  # 转为0-2索引

        # 使用文件名第一个字符作为专家分数
        try:
            expert_score = int(file_name[0]) - 1  # 假设第一个字符是1-3的数字
            if 0 <= expert_score <= 2:  # 确保在有效范围内
                confusion_matrix[system_score, expert_score] += 1
        except (ValueError, IndexError):
            # 如果文件名格式不符合要求，跳过
            continue

    # 3. 端到端处理时间计算
    avg_processing_time = sum([r.get('processing_time', 0) for r in results]) / len(results) if results else 0

    # 4. 分数频率分析
    score_counts = {1: 0, 2: 0, 3: 0}
    for r in results:
        score = r['score']
        if score in score_counts:
            score_counts[score] += 1

    # 输出结果
    print("\nPerformance Metrics:")
    print("-" * 40)

    print(f"1. Detection Success Rate: {detection_success_rate:.2f}%")
    print(f"   Total Images: {total_images}")
    print(f"   Successfully Detected: {detected_images}")

    print("\n2. Confusion Matrix Analysis:")
    print("   System Score vs Expert Score (Expert scores from filename first digit)")
    print("   Expert Score")
    print("              Score 1    Score 2    Score 3")
    print(f"   System Score 1:  {confusion_matrix[0, 0]}     {confusion_matrix[0, 1]}     {confusion_matrix[0, 2]}")
    print(f"   System Score 2:  {confusion_matrix[1, 0]}     {confusion_matrix[1, 1]}     {confusion_matrix[1, 2]}")
    print(f"   System Score 3:  {confusion_matrix[2, 0]}     {confusion_matrix[2, 1]}     {confusion_matrix[2, 2]}")

    # 计算总体准确率
    total_matches = sum([confusion_matrix[i, i] for i in range(3)])
    total_samples = np.sum(confusion_matrix)
    accuracy = (total_matches / total_samples) * 100 if total_samples > 0 else 0
    print(f"   Scoring Accuracy: {accuracy:.2f}%")

    print(f"\n3. End-to-End Processing Time: {avg_processing_time:.4f} seconds/image")

    print("\n4. Score Frequency Analysis:")
    total_scores = sum(score_counts.values())
    for score, count in score_counts.items():
        percentage = (count / total_scores) * 100 if total_scores > 0 else 0
        print(f"   Score {score}: {count} images ({percentage:.2f}%)")

    # 保存指标结果到文件
    with open(os.path.join(output_folder, "performance_metrics.txt"), 'w') as f:
        f.write("Performance Metrics\n")
        f.write("-" * 40 + "\n")
        f.write(f"1. Detection Success Rate: {detection_success_rate:.2f}%\n")
        f.write(f"   Total Images: {total_images}\n")
        f.write(f"   Successfully Detected: {detected_images}\n\n")

        f.write("2. Confusion Matrix Analysis:\n")
        f.write("   Expert Score\n")
        f.write("              Score 1    Score 2    Score 3\n")
        f.write(f"   System Score 1:  {confusion_matrix[0, 0]}     {confusion_matrix[0, 1]}     {confusion_matrix[0, 2]}\n")
        f.write(f"   System Score 2:  {confusion_matrix[1, 0]}     {confusion_matrix[1, 1]}     {confusion_matrix[1, 2]}\n")
        f.write(f"   System Score 3:  {confusion_matrix[2, 0]}     {confusion_matrix[2, 1]}     {confusion_matrix[2, 2]}\n")
        f.write(f"   Scoring Accuracy: {accuracy:.2f}%\n\n")

        f.write(f"3. End-to-End Processing Time: {avg_processing_time:.4f} seconds/image\n\n")

        f.write("4. Score Frequency Analysis:\n")
        for score, count in score_counts.items():
            percentage = (count / total_scores) * 100 if total_scores > 0 else 0
            f.write(f"   Score {score}: {count} images ({percentage:.2f}%)\n")

def process_image_folder(input_folder, output_folder):
    """处理文件夹中的所有图像"""
    # 创建输出文件夹
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 设置MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # 获取图像列表
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []

    # 处理每张图像
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            start_time = time.time()  # 开始计时

            # 读取图像
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # 处理图像
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # 计算处理时间
            processing_time = time.time() - start_time

            # 检查是否检测到姿势
            if results.pose_landmarks:
                # 绘制姿势关键点
                mp_drawing.draw_landmarks(
                    img,
                    results.pose_landmarks,
                    mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=3),
                    mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2)
                )

                # 评分
                evaluation = score_trunk_rotation(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                leg_angle = evaluation['leg_angle']
                is_right_leg_raised = evaluation['is_right_leg_raised']

                # 记录分数和处理时间
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'leg_angle': leg_angle,
                    'processing_time': processing_time
                })

                # 获取关键点坐标
                landmarks = results.pose_landmarks.landmark
                height, width, _ = img.shape

                if is_right_leg_raised:
                    ankle = (int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x * width),
                            int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y * height))
                    knee = (int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x * width),
                           int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y * height))
                    hip = (int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x * width),
                          int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y * height))
                else:
                    ankle = (int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x * width),
                            int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y * height))
                    knee = (int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x * width),
                           int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y * height))
                    hip = (int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x * width),
                          int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y * height))

                # 绘制线条
                cv2.line(img, ankle, knee, (0, 255, 255), 3)  # 脚踝到膝盖
                cv2.line(img, knee, hip, (0, 255, 255), 3)    # 膝盖到髋部

                # 添加角度文字
                cv2.putText(img, f"{leg_angle:.1f} deg", (knee[0] + 10, knee[1] - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 255), 3, cv2.LINE_AA)

                # 添加分数
                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(img, f"SCORE: {score}/3", (20, 60),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.8, score_color, 4, cv2.LINE_AA)

                # 保存图像
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, img)

            else:
                # 未检测到姿势
                print(f"No pose detected in {img_file}")
                # 在原图上添加错误信息
                cv2.putText(img, "NO POSE DETECTED", (img.shape[1]//4, img.shape[0]//2),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 3, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, img)

    # 评估性能指标
    evaluate_performance(input_folder, output_folder, score_results)

    # 创建图表，使用英文并增大字体
    if score_results:
        # 计算统计数据
        scores = [r['score'] for r in score_results]
        angles = [r['leg_angle'] for r in score_results]
        processing_times = [r['processing_time'] for r in score_results]

        # 设置全局字体大小
        plt.rcParams.update({'font.size': 14})

        # 创建角度分布图
        plt.figure(figsize=(10, 6))
        plt.hist(angles, bins=10, color='blue', alpha=0.7)
        plt.title('Leg Angle Distribution', fontsize=18)
        plt.xlabel('Angle (degrees)', fontsize=16)
        plt.ylabel('Number of Images', fontsize=16)
        plt.grid(True, alpha=0.3)
        plt.axvline(x=130, color='orange', linestyle='--', linewidth=2,
                   label='130 deg (Score 1-2 boundary)')
        plt.axvline(x=160, color='green', linestyle='--', linewidth=2,
                   label='160 deg (Score 2-3 boundary)')
        plt.legend(fontsize=14)
        plt.tight_layout()
        angle_chart_path = os.path.join(output_folder, "angle_distribution.png")
        plt.savefig(angle_chart_path, dpi=300)

        # 创建分数分布图
        plt.figure(figsize=(10, 6))
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1
        plt.bar(['Score 1', 'Score 2', 'Score 3'],
               [score_counts[1], score_counts[2], score_counts[3]],
               color=['red', 'gold', 'green'])
        plt.title('Score Distribution', fontsize=18)
        plt.xlabel('Score', fontsize=16)
        plt.ylabel('Number of Images', fontsize=16)
        plt.grid(axis='y', alpha=0.3)
        for i, v in enumerate([score_counts[1], score_counts[2], score_counts[3]]):
            plt.text(i, v + 0.5, str(v), ha='center', fontsize=14, fontweight='bold')
        plt.tight_layout()
        score_chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(score_chart_path, dpi=300)

        # 创建处理时间分布图
        plt.figure(figsize=(10, 6))
        plt.hist(processing_times, bins=10, color='green', alpha=0.7)
        plt.title('Processing Time Distribution', fontsize=18)
        plt.xlabel('Processing Time (seconds)', fontsize=16)
        plt.ylabel('Number of Images', fontsize=16)
        plt.grid(True, alpha=0.3)
        avg_time = sum(processing_times) / len(processing_times)
        plt.axvline(x=avg_time, color='red', linestyle='--', linewidth=2,
                   label=f'Average: {avg_time:.4f} sec')
        plt.legend(fontsize=14)
        plt.tight_layout()
        time_chart_path = os.path.join(output_folder, "processing_time_distribution.png")
        plt.savefig(time_chart_path, dpi=300)

        # 创建混淆矩阵可视化
        plt.figure(figsize=(10, 8))
        # 使用之前计算的混淆矩阵数据
        confusion_matrix = np.zeros((3, 3), dtype=int)
        for result in score_results:
            file_name = result['file']
            system_score = result['score'] - 1  # 转为0-2索引

            # 使用文件名第一个字符作为专家分数
            try:
                expert_score = int(file_name[0]) - 1  # 假设第一个字符是1-3的数字
                if 0 <= expert_score <= 2:  # 确保在有效范围内
                    confusion_matrix[system_score, expert_score] += 1
            except (ValueError, IndexError):
                # 如果文件名格式不符合要求，跳过
                continue

        plt.imshow(confusion_matrix, interpolation='nearest', cmap='Blues')
        plt.title('Confusion Matrix', fontsize=18)
        plt.colorbar()
        tick_marks = np.arange(3)
        plt.xticks(tick_marks, ['Score 1', 'Score 2', 'Score 3'], fontsize=14)
        plt.yticks(tick_marks, ['Score 1', 'Score 2', 'Score 3'], fontsize=14)
        plt.xlabel('Expert Score', fontsize=16)
        plt.ylabel('System Score', fontsize=16)

        # 在混淆矩阵中添加数值标注
        thresh = confusion_matrix.max() / 2.
        for i in range(confusion_matrix.shape[0]):
            for j in range(confusion_matrix.shape[1]):
                plt.text(j, i, format(confusion_matrix[i, j], 'd'),
                       ha="center", va="center", fontsize=16, fontweight='bold',
                       color="white" if confusion_matrix[i, j] > thresh else "black")

        plt.tight_layout()
        conf_matrix_path = os.path.join(output_folder, "confusion_matrix.png")
        plt.savefig(conf_matrix_path, dpi=300)

        print(f"Charts saved to {output_folder}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# 视频处理函数 (可选功能)
def process_video(input_path, output_path):
    """处理视频并进行姿态分析"""
    # 打开视频文件
    cap = cv2.VideoCapture(input_path)
    if not cap.isOpened():
        print(f"Error: Cannot open video file {input_path}")
        return None

    # 获取视频属性
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"Video info: Width={width}, Height={height}, FPS={fps}, Total frames={total_frames}")

    # 创建视频写入器
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # 初始化变量
    frame_count = 0
    pTime = 0
    angles_history = []  # 存储角度数据

    # 创建姿态检测模型
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    with mp_pose.Pose(
        static_image_mode=False,
        smooth_landmarks=True,
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5
    ) as pose:
        # 处理每一帧
        while cap.isOpened():
            success, img = cap.read()

            if not success:
                print("Video processing complete or reading error")
                break

            # 将BGR格式转为RGB格式
            imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # 处理图像
            start_time = time.time()  # 开始计时
            results = pose.process(imgRGB)
            processing_time = time.time() - start_time  # 计算处理时间

            # 在原图上绘制姿态信息
            if results.pose_landmarks:
                # 绘制关键点和连接线
                mp_drawing.draw_landmarks(
                    img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(255,0,0), thickness=2, circle_radius=3),
                    mp_drawing.DrawingSpec(color=(0,255,0), thickness=2, circle_radius=2)
                )

                # 应用转动稳定性评分函数
                evaluation = score_trunk_rotation(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                leg_angle = evaluation['leg_angle']
                is_right_leg_raised = evaluation['is_right_leg_raised']

                # 在图像上显示评分和角度
                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(img, f"SCORE: {score}/3", (20, 60),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.8, score_color, 4, cv2.LINE_AA)

                # 获取关键点坐标
                landmarks = results.pose_landmarks.landmark

                if is_right_leg_raised:
                    ankle = (int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x * width),
                            int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y * height))
                    knee = (int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x * width),
                           int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y * height))
                    hip = (int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x * width),
                          int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y * height))
                else:
                    ankle = (int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x * width),
                            int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y * height))
                    knee = (int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x * width),
                           int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y * height))
                    hip = (int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x * width),
                          int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y * height))

                # 绘制线条
                cv2.line(img, ankle, knee, (0, 255, 255), 3)  # 脚踝到膝盖
                cv2.line(img, knee, hip, (0, 255, 255), 3)    # 膝盖到髋部

                # 添加角度文字
                cv2.putText(img, f"{leg_angle:.1f} deg", (knee[0] + 10, knee[1] - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 255), 3, cv2.LINE_AA)

                # 存储角度数据
                current_angles = {
                    'frame': frame_count,
                    'leg_angle': leg_angle,
                    'score': score,
                    'processing_time': processing_time
                }
                angles_history.append(current_angles)

            # 显示进度
            frame_count += 1
            if frame_count % 30 == 0:  # 每30帧显示一次进度
                print(f"Processing progress: {frame_count}/{total_frames} ({frame_count/total_frames*100:.1f}%)")

            # 计算和显示FPS
            cTime = time.time()
            if pTime > 0:
                fps_current = 1/(cTime-pTime)
                cv2.putText(img, f"FPS: {int(fps_current)}", (10, 30),
                           cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 2)
            pTime = cTime

            # 写入输出视频
            out.write(img)

        # 释放资源
        cap.release()
        out.release()

    print(f"Video processing complete. Processed {frame_count} frames.")

    # 返回分析结果
    return angles_history

# 示例用法
if __name__ == "__main__":
    # 图像处理
    input_folder = "rr"
    output_folder = "rr2"
    results = process_image_folder(input_folder, output_folder)

    # 视频处理（可选）
    # video_input = "test.mp4"
    # video_output = "output_video.mp4"
    # angles_history = process_video(video_input, video_output)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm  # For progress bar
import time

# Helper functions
def calculate_distance(a, b):
    """Calculate the Euclidean distance between two points"""
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)

def calculate_angle(a, b, c):
    """
    Calculate the angle between three points
    Args:
        a: First point [x, y]
        b: Middle point (vertex) [x, y]
        c: Last point [x, y]
    Returns:
        Angle in degrees
    """
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle

    return angle

def identify_raised_leg(landmarks, mp_pose):
    """
    Determine which leg is raised based on y-coordinate of ankles
    Returns:
        is_right_leg_raised: Boolean, True if right leg is raised, False if left leg is raised
    """
    left_ankle_y = landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y
    right_ankle_y = landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y

    # The leg with the smaller y-coordinate (higher on screen) is raised
    is_right_leg_raised = right_ankle_y < left_ankle_y

    return is_right_leg_raised

def calculate_leg_trunk_angle(landmarks, mp_pose, is_right_leg_raised):
    """
    Calculate the angle between the raised leg and the trunk
    Args:
        landmarks: MediaPipe pose landmarks
        mp_pose: MediaPipe pose model
        is_right_leg_raised: Boolean, True if right leg is raised, False if left leg is raised
    Returns:
        Angle in degrees
    """
    if is_right_leg_raised:
        # For right leg
        hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
        shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                   landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
    else:
        # For left leg
        hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                   landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]

    # Calculate the angle at the hip between the line from hip to ankle and the line from hip to shoulder
    angle = calculate_angle(ankle, hip, shoulder)

    return angle

def score_leg_raising(landmarks, mp_pose):
    """
    Score the active straight leg raise test:
    - 3 points: Angle between 90-110 degrees
    - 2 points: Angle between 110-150 degrees
    - 1 point: Angle > 150 degrees
    """
    # Identify which leg is raised
    is_right_leg_raised = identify_raised_leg(landmarks, mp_pose)

    # Calculate the leg-trunk angle
    leg_trunk_angle = calculate_leg_trunk_angle(landmarks, mp_pose, is_right_leg_raised)

    # Check if in lying position (simplified check)
    right_hip_y = landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y
    left_hip_y = landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y

    is_lying = abs(right_hip_y - left_hip_y) < 0.1  # Hips should be at similar height if lying

    # Also check if raised leg is straight (no significant knee bend)
    if is_right_leg_raised:
        knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
    else:
        knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]

    knee_angle = calculate_angle(hip, knee, ankle)
    leg_is_straight = knee_angle > 160  # Less than 20 degrees of knee bend

    # Determine score based on angle
    score = 1  # Default score
    feedback = []

    # Debug values
    debug_info = {
        'leg_trunk_angle': leg_trunk_angle,
        'knee_angle': knee_angle,
        'is_lying': is_lying,
        'leg_is_straight': leg_is_straight,
        'is_right_leg_raised': is_right_leg_raised
    }

    if not is_lying:
        score = 1
        feedback.append("Position incorrect. Body should be lying flat.")
    elif not leg_is_straight:
        score = 1
        feedback.append("Leg should be kept straight during the test.")
    elif leg_trunk_angle >= 90 and leg_trunk_angle <= 110:
        score = 3
        feedback.append(f"Excellent! High leg raise ({leg_trunk_angle:.1f} deg)")
    elif leg_trunk_angle > 110 and leg_trunk_angle <= 150:
        score = 2
        feedback.append(f"Good leg raise to moderate height ({leg_trunk_angle:.1f} deg)")
    else:  # > 150
        score = 1
        feedback.append(f"Limited leg raise. Try to lift higher ({leg_trunk_angle:.1f} deg)")

    return {
        'score': score,
        'feedback': feedback,
        'debug_info': debug_info
    }

# 性能指标评估函数
def evaluate_performance(input_folder, output_folder, results):
    """计算并输出四个关键性能指标"""
    # 1. 检测成功率计算
    total_images = len([f for f in os.listdir(input_folder)
                      if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    detected_images = len(results)
    detection_success_rate = (detected_images / total_images) * 100 if total_images > 0 else 0

    # 2. 混淆矩阵分析 (使用文件名第一个字符作为专家打分)
    confusion_matrix = np.zeros((3, 3), dtype=int)  # 3×3矩阵对应1-3分

    for result in results:
        file_name = result['file']
        system_score = result['score'] - 1  # 转为0-2索引

        # 使用文件名第一个字符作为专家分数
        try:
            expert_score = int(file_name[0]) - 1  # 假设第一个字符是1-3的数字
            if 0 <= expert_score <= 2:  # 确保在有效范围内
                confusion_matrix[system_score, expert_score] += 1
        except (ValueError, IndexError):
            # 如果文件名格式不符合要求，跳过
            continue

    # 3. 端到端处理时间计算
    avg_processing_time = sum([r.get('processing_time', 0) for r in results]) / len(results) if results else 0

    # 4. 分数频率分析
    score_counts = {1: 0, 2: 0, 3: 0}
    for r in results:
        score = r['score']
        if score in score_counts:
            score_counts[score] += 1

    # 输出结果
    print("\nPerformance Metrics:")
    print("-" * 40)

    print(f"1. Detection Success Rate: {detection_success_rate:.2f}%")
    print(f"   Total Images: {total_images}")
    print(f"   Successfully Detected: {detected_images}")

    print("\n2. Confusion Matrix Analysis:")
    print("   System Score vs Expert Score (Expert scores from filename first digit)")
    print("   Expert Score")
    print("              Score 1    Score 2    Score 3")
    print(f"   System Score 1:  {confusion_matrix[0, 0]}     {confusion_matrix[0, 1]}     {confusion_matrix[0, 2]}")
    print(f"   System Score 2:  {confusion_matrix[1, 0]}     {confusion_matrix[1, 1]}     {confusion_matrix[1, 2]}")
    print(f"   System Score 3:  {confusion_matrix[2, 0]}     {confusion_matrix[2, 1]}     {confusion_matrix[2, 2]}")

    # 计算总体准确率
    total_matches = sum([confusion_matrix[i, i] for i in range(3)])
    total_samples = np.sum(confusion_matrix)
    accuracy = (total_matches / total_samples) * 100 if total_samples > 0 else 0
    print(f"   Scoring Accuracy: {accuracy:.2f}%")

    print(f"\n3. End-to-End Processing Time: {avg_processing_time:.4f} seconds/image")

    print("\n4. Score Frequency Analysis:")
    total_scores = sum(score_counts.values())
    for score, count in score_counts.items():
        percentage = (count / total_scores) * 100 if total_scores > 0 else 0
        print(f"   Score {score}: {count} images ({percentage:.2f}%)")

    # 保存指标结果到文件
    with open(os.path.join(output_folder, "performance_metrics.txt"), 'w') as f:
        f.write("Performance Metrics\n")
        f.write("-" * 40 + "\n")
        f.write(f"1. Detection Success Rate: {detection_success_rate:.2f}%\n")
        f.write(f"   Total Images: {total_images}\n")
        f.write(f"   Successfully Detected: {detected_images}\n\n")

        f.write("2. Confusion Matrix Analysis:\n")
        f.write("   Expert Score\n")
        f.write("              Score 1    Score 2    Score 3\n")
        f.write(f"   System Score 1:  {confusion_matrix[0, 0]}     {confusion_matrix[0, 1]}     {confusion_matrix[0, 2]}\n")
        f.write(f"   System Score 2:  {confusion_matrix[1, 0]}     {confusion_matrix[1, 1]}     {confusion_matrix[1, 2]}\n")
        f.write(f"   System Score 3:  {confusion_matrix[2, 0]}     {confusion_matrix[2, 1]}     {confusion_matrix[2, 2]}\n")
        f.write(f"   Scoring Accuracy: {accuracy:.2f}%\n\n")

        f.write(f"3. End-to-End Processing Time: {avg_processing_time:.4f} seconds/image\n\n")

        f.write("4. Score Frequency Analysis:\n")
        for score, count in score_counts.items():
            percentage = (count / total_scores) * 100 if total_scores > 0 else 0
            f.write(f"   Score {score}: {count} images ({percentage:.2f}%)\n")

# Main function to process images
def process_image_folder(input_folder, output_folder):
    """
    Process all images in a folder with pose detection and leg raising test scoring
    """
    # Make sure output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []

    # Process each image with pose detection
    with mp_pose.Pose(
        static_image_mode=True,  # Use static mode for images
        model_complexity=2,  # Higher accuracy
        min_detection_confidence=0.5
    ) as pose:
        # Use tqdm for progress bar
        for img_file in tqdm(image_files):
            start_time = time.time()  # Start timing

            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)

            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Convert to RGB for MediaPipe
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Process image
            results = pose.process(img_rgb)

            # Calculate processing time
            processing_time = time.time() - start_time

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Get original image dimensions
                height, width, _ = img.shape

                # Draw pose landmarks on original image
                mp_drawing.draw_landmarks(
                    img,
                    results.pose_landmarks,
                    mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=3),
                    mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2)
                )

                # Score the pose
                evaluation = score_leg_raising(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                feedback = evaluation['feedback']
                debug_info = evaluation['debug_info']

                # Track scores
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'leg_trunk_angle': debug_info['leg_trunk_angle'],
                    'processing_time': processing_time
                })

                # Draw the leg-trunk angle on the image
                landmarks = results.pose_landmarks.landmark
                is_right_leg_raised = debug_info['is_right_leg_raised']

                if is_right_leg_raised:
                    hip_point = (int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x * width),
                                int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y * height))
                    ankle_point = (int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x * width),
                                  int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y * height))
                    shoulder_point = (int(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x * width),
                                     int(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y * height))
                else:
                    hip_point = (int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x * width),
                                int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y * height))
                    ankle_point = (int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x * width),
                                  int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y * height))
                    shoulder_point = (int(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x * width),
                                     int(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y * height))

                # Draw lines for angle visualization
                cv2.line(img, hip_point, ankle_point, (0, 255, 255), 3)  # Leg line
                cv2.line(img, hip_point, shoulder_point, (0, 255, 255), 3)  # Trunk line

                # Add angle text near hip
                angle = debug_info['leg_trunk_angle']
                cv2.putText(img, f"{angle:.1f} deg",
                          (hip_point[0] + 10, hip_point[1] - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 255), 3, cv2.LINE_AA)

                # Add score with larger text
                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(img, f"SCORE: {score}/3", (20, 60),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.8, score_color, 4, cv2.LINE_AA)

                # Save the image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, img)

            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                # Add error message to the original image
                cv2.putText(img, "NO POSE DETECTED", (img.shape[1]//4, img.shape[0]//2),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.5, (0, 0, 255), 3, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, img)

    # Evaluate performance metrics
    evaluate_performance(input_folder, output_folder, score_results)

    # Generate visualization charts
    if score_results:
        # Calculate statistics
        scores = [r['score'] for r in score_results]
        angles = [r['leg_trunk_angle'] for r in score_results]
        processing_times = [r['processing_time'] for r in score_results]

        # Set global font size
        plt.rcParams.update({'font.size': 14})

        # Create angle distribution chart
        plt.figure(figsize=(10, 6))
        plt.hist(angles, bins=10, color='blue', alpha=0.7)
        plt.title('Leg-Trunk Angle Distribution', fontsize=18)
        plt.xlabel('Angle (degrees)', fontsize=16)
        plt.ylabel('Number of Images', fontsize=16)
        plt.grid(True, alpha=0.3)
        plt.axvline(x=110, color='green', linestyle='--', linewidth=2,
                   label='110 deg (Score 3-2 boundary)')
        plt.axvline(x=150, color='orange', linestyle='--', linewidth=2,
                   label='150 deg (Score 2-1 boundary)')
        plt.legend(fontsize=14)
        plt.tight_layout()
        angle_chart_path = os.path.join(output_folder, "angle_distribution.png")
        plt.savefig(angle_chart_path, dpi=300)

        # Create score distribution chart
        plt.figure(figsize=(10, 6))
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1
        plt.bar(['Score 1', 'Score 2', 'Score 3'],
               [score_counts[1], score_counts[2], score_counts[3]],
               color=['red', 'gold', 'green'])
        plt.title('Score Distribution', fontsize=18)
        plt.xlabel('Score', fontsize=16)
        plt.ylabel('Number of Images', fontsize=16)
        plt.grid(axis='y', alpha=0.3)
        for i, v in enumerate([score_counts[1], score_counts[2], score_counts[3]]):
            plt.text(i, v + 0.5, str(v), ha='center', fontsize=14, fontweight='bold')
        plt.tight_layout()
        score_chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(score_chart_path, dpi=300)

        # Create processing time distribution chart
        plt.figure(figsize=(10, 6))
        plt.hist(processing_times, bins=10, color='green', alpha=0.7)
        plt.title('Processing Time Distribution', fontsize=18)
        plt.xlabel('Processing Time (seconds)', fontsize=16)
        plt.ylabel('Number of Images', fontsize=16)
        plt.grid(True, alpha=0.3)
        avg_time = sum(processing_times) / len(processing_times)
        plt.axvline(x=avg_time, color='red', linestyle='--', linewidth=2,
                   label=f'Average: {avg_time:.4f} sec')
        plt.legend(fontsize=14)
        plt.tight_layout()
        time_chart_path = os.path.join(output_folder, "processing_time_distribution.png")
        plt.savefig(time_chart_path, dpi=300)

        # Create confusion matrix visualization
        plt.figure(figsize=(10, 8))
        # Use previously calculated confusion matrix data
        confusion_matrix = np.zeros((3, 3), dtype=int)
        for result in score_results:
            file_name = result['file']
            system_score = result['score'] - 1  # Convert to 0-2 index

            # Use first character of filename as expert score
            try:
                expert_score = int(file_name[0]) - 1  # Assuming first character is 1-3
                if 0 <= expert_score <= 2:  # Ensure valid range
                    confusion_matrix[system_score, expert_score] += 1
            except (ValueError, IndexError):
                # Skip if filename format is invalid
                continue

        plt.imshow(confusion_matrix, interpolation='nearest', cmap='Blues')
        plt.title('Confusion Matrix', fontsize=18)
        plt.colorbar()
        tick_marks = np.arange(3)
        plt.xticks(tick_marks, ['Score 1', 'Score 2', 'Score 3'], fontsize=14)
        plt.yticks(tick_marks, ['Score 1', 'Score 2', 'Score 3'], fontsize=14)
        plt.xlabel('Expert Score', fontsize=16)
        plt.ylabel('System Score', fontsize=16)

        # Add value annotations to confusion matrix
        thresh = confusion_matrix.max() / 2.
        for i in range(confusion_matrix.shape[0]):
            for j in range(confusion_matrix.shape[1]):
                plt.text(j, i, format(confusion_matrix[i, j], 'd'),
                       ha="center", va="center", fontsize=16, fontweight='bold',
                       color="white" if confusion_matrix[i, j] > thresh else "black")

        plt.tight_layout()
        conf_matrix_path = os.path.join(output_folder, "confusion_matrix.png")
        plt.savefig(conf_matrix_path, dpi=300)

        # Create summary report
        report_path = os.path.join(output_folder, "score_summary.txt")
        with open(report_path, 'w') as f:
            f.write(f"Straight Leg Raise Test Score Summary\n")
            f.write(f"====================================\n\n")
            f.write(f"Images analyzed: {len(score_results)}\n")
            f.write(f"Average score: {sum(scores)/len(scores):.2f}/3\n")
            f.write(f"Average leg-trunk angle: {sum(angles)/len(angles):.2f} degrees\n\n")
            f.write(f"Score distribution:\n")
            for score, count in score_counts.items():
                percentage = (count / len(scores)) * 100 if scores else 0
                f.write(f"  Score {score}: {count} images ({percentage:.1f}%)\n")

            f.write("\nDetailed results:\n")
            for i, result in enumerate(score_results):
                f.write(f"\n{i+1}. {result['file']}\n")
                f.write(f"   Score: {result['score']}/3\n")
                f.write(f"   Angle: {result['leg_trunk_angle']:.1f} degrees\n")
                f.write(f"   Feedback: {', '.join(result['feedback'])}\n")

        print(f"\nReport saved to {report_path}")
        print(f"Charts saved to {output_folder}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# Example usage
if __name__ == "__main__":
    # Set these paths to your folders
    input_folder = "ll"
    output_folder = "l1"

    # Process the images
    results = process_image_folder(input_folder, output_folder)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import time

# Calculate angle between vectors function
def calculate_angle_between_vectors(v1, v2):
    v1, v2 = np.array(v1), np.array(v2)
    v1_norm = v1 / np.linalg.norm(v1)
    v2_norm = v2 / np.linalg.norm(v2)
    dot_product = np.clip(np.dot(v1_norm, v2_norm), -1.0, 1.0)
    return np.arccos(dot_product) * 180.0 / np.pi

# Calculate pushup angle
def calculate_pushup_angle(landmarks, mp_pose):
    # Get key points
    left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
    right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
    left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x, landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
    right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

    # Calculate midpoints
    mid_ankle = [(left_ankle[0] + right_ankle[0])/2, (left_ankle[1] + right_ankle[1])/2]
    mid_hip = [(left_hip[0] + right_hip[0])/2, (left_hip[1] + right_hip[1])/2]
    mid_wrist = [(left_wrist[0] + right_wrist[0])/2, (left_wrist[1] + right_wrist[1])/2]

    # Calculate vectors
    ankle_to_hip = [mid_hip[0] - mid_ankle[0], mid_hip[1] - mid_ankle[1]]
    ankle_to_hand = [mid_wrist[0] - mid_ankle[0], mid_wrist[1] - mid_ankle[1]]

    # Calculate angle
    return calculate_angle_between_vectors(ankle_to_hip, ankle_to_hand)

# Scoring function
def score_pushup(landmarks, mp_pose):
    pushup_angle = calculate_pushup_angle(landmarks, mp_pose)

    # Scoring logic
    if pushup_angle > 35:
        score = 3
        feedback = f"Excellent! Angle: {pushup_angle:.1f} deg (>35 deg)"
    elif pushup_angle >= 15:
        score = 2
        feedback = f"Good position. Angle: {pushup_angle:.1f} deg (15-35 deg)"
    else:
        score = 1
        feedback = f"Improve hand position. Angle: {pushup_angle:.1f} deg (<15 deg)"

    return {
        'score': score,
        'feedback': feedback,
        'pushup_angle': pushup_angle
    }

# Performance evaluation function
def evaluate_performance(results, expert_scores=None, total_images=None):
    """
    Calculate performance metrics for the pose detection and scoring system

    Parameters:
        results: List of dictionaries containing detection and scoring results
        expert_scores: Dictionary mapping filename to expert-assigned scores (optional)
        total_images: Total number of images processed (in case some failed)

    Returns:
        Dictionary containing performance metrics
    """
    # Initialize metrics
    metrics = {}

    # 1. Detection Success Rate
    if total_images:
        detection_rate = len(results) / total_images * 100
        metrics['detection_success_rate'] = detection_rate

    # If no results, return early
    if not results:
        return metrics

    # 2. End-to-end processing time (if available)
    if all('processing_time' in r for r in results):
        avg_time = sum(r['processing_time'] for r in results) / len(results)
        metrics['avg_processing_time'] = avg_time

    # 3. Score frequency analysis
    scores = [r['score'] for r in results]
    score_counts = {1: 0, 2: 0, 3: 0}
    for s in scores:
        score_counts[s] += 1
    metrics['score_counts'] = score_counts
    metrics['avg_score'] = sum(scores) / len(scores)

    # 4. Confusion matrix analysis (if expert scores are provided)
    if expert_scores:
        confusion_matrix = np.zeros((3, 3), dtype=int)  # 3x3 for scores 1, 2, 3
        agreement_count = 0

        for r in results:
            if r['file'] in expert_scores:
                pred_score = r['score']
                true_score = expert_scores[r['file']]

                # Adjust for 0-indexing
                confusion_matrix[true_score-1, pred_score-1] += 1

                if pred_score == true_score:
                    agreement_count += 1

        total_compared = sum(sum(row) for row in confusion_matrix)
        if total_compared > 0:
            accuracy = agreement_count / total_compared * 100
            metrics['accuracy'] = accuracy
            metrics['confusion_matrix'] = confusion_matrix.tolist()

    # 5. Angle statistics
    angles = [r['pushup_angle'] for r in results]
    metrics['avg_angle'] = sum(angles) / len(angles)
    metrics['min_angle'] = min(angles)
    metrics['max_angle'] = max(angles)

    return metrics

# Main processing function
def process_image_folder(input_folder, output_folder, expert_scores=None):
    # Create output folder
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Setup MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get image list
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []
    total_images = len(image_files)

    # Process each image
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            start_time = time.time()  # Start timing

            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Process image
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # Check if pose is detected
            if results.pose_landmarks:
                # Get image dimensions
                orig_height, orig_width, _ = img.shape

                # Create a copy of original image
                output_img = img.copy()

                # Draw pose landmarks
                mp_drawing.draw_landmarks(output_img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # Score
                evaluation = score_pushup(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                feedback = evaluation['feedback']
                angle = evaluation['pushup_angle']

                # Calculate processing time
                end_time = time.time()
                processing_time = end_time - start_time

                # Record scores
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'pushup_angle': angle,
                    'processing_time': processing_time
                })

                # Draw angle lines on the image
                landmarks = results.pose_landmarks.landmark
                mid_ankle_x = int(((landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x +
                                  landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x) / 2) * orig_width)
                mid_ankle_y = int(((landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y +
                                  landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y) / 2) * orig_height)
                mid_hip_x = int(((landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x +
                                landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x) / 2) * orig_width)
                mid_hip_y = int(((landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y +
                                landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y) / 2) * orig_height)
                mid_wrist_x = int(((landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x +
                                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x) / 2) * orig_width)
                mid_wrist_y = int(((landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y +
                                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y) / 2) * orig_height)

                # Draw lines
                cv2.line(output_img, (mid_ankle_x, mid_ankle_y), (mid_hip_x, mid_hip_y), (0, 255, 255), 2)
                cv2.line(output_img, (mid_ankle_x, mid_ankle_y), (mid_wrist_x, mid_wrist_y), (0, 255, 255), 2)

                # Add angle text
                cv2.putText(output_img, f"{angle:.1f} deg", (mid_ankle_x + 10, mid_ankle_y - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2, cv2.LINE_AA)

                # Add score
                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(output_img, f"Score: {score}/3", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, score_color, 2, cv2.LINE_AA)

                # Save image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, output_img)

            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                end_time = time.time()
                processing_time = end_time - start_time

                # Create a copy with "NO POSE DETECTED" text
                output_img = img.copy()
                cv2.putText(output_img, "NO POSE DETECTED", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, output_img)

    # Generate performance evaluation
    performance_metrics = evaluate_performance(score_results, expert_scores, total_images)

    # Generate summary report
    if score_results:
        # Calculate statistics
        scores = [r['score'] for r in score_results]
        angles = [r['pushup_angle'] for r in score_results]

        # Create score distribution chart
        plt.figure(figsize=(10, 6))
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1
        plt.bar(score_counts.keys(), score_counts.values(), color='green')
        plt.title('Pushup Position Test Score Distribution')
        plt.xlabel('Score')
        plt.ylabel('Number of Images')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)
        plt.close()

        # Create angle distribution chart
        plt.figure(figsize=(10, 6))
        plt.hist(angles, bins=10, color='blue', alpha=0.7)
        plt.title('Angle Distribution')
        plt.xlabel('Angle (degrees)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.axvline(x=15, color='orange', linestyle='--', label='15 deg (Score 1-2 boundary)')
        plt.axvline(x=35, color='green', linestyle='--', label='35 deg (Score 2-3 boundary)')
        plt.legend()
        angle_chart_path = os.path.join(output_folder, "angle_distribution.png")
        plt.savefig(angle_chart_path)
        plt.close()

        # If confusion matrix is available, visualize it
                # 如果有混淆矩阵数据，可视化它
        if 'confusion_matrix' in performance_metrics:
            plt.figure(figsize=(10, 8))
            cm = np.array(performance_metrics['confusion_matrix'])

            # 创建热力图
            ax = plt.subplot()
            im = ax.imshow(cm, interpolation='nearest', cmap='Blues')
            plt.colorbar(im)

            # 设置标签
            classes = ['Score 1', 'Score 2', 'Score 3']
            ax.set_xlabel('Expert Score', fontsize=12)
            ax.set_ylabel('System Score', fontsize=12)
            ax.set_title('Confusion Matrix', fontsize=14)

            # 设置刻度
            tick_marks = np.arange(len(classes))
            ax.set_xticks(tick_marks)
            ax.set_yticks(tick_marks)
            ax.set_xticklabels(classes)
            ax.set_yticklabels(classes)

            # 在每个单元格中添加数字
            thresh = cm.max() / 2.
            for i in range(len(classes)):
                for j in range(len(classes)):
                    ax.text(j, i, str(cm[i, j]),
                          ha="center", va="center",
                          color="white" if cm[i, j] > thresh else "black",
                          fontsize=14)

            # 调整布局
            plt.tight_layout()
            cm_path = os.path.join(output_folder, "confusion_matrix.png")
            plt.savefig(cm_path)
            plt.close()


        # Create processing time histogram if available
        if all('processing_time' in r for r in score_results):
            times = [r['processing_time'] for r in score_results]
            plt.figure(figsize=(10, 6))
            plt.hist(times, bins=10, color='purple', alpha=0.7)
            plt.title('Processing Time Distribution')
            plt.xlabel('Time (seconds)')
            plt.ylabel('Number of Images')
            plt.grid(True, alpha=0.3)
            time_chart_path = os.path.join(output_folder, "time_distribution.png")
            plt.savefig(time_chart_path)
            plt.close()

        print(f"Charts saved to {output_folder}")

        # Save performance metrics to file
        with open(os.path.join(output_folder, "performance_metrics.txt"), 'w') as f:
            f.write("Performance Metrics\n")
            f.write("===================\n\n")

            f.write(f"Total images: {total_images}\n")
            f.write(f"Images with detected pose: {len(score_results)}\n")
            f.write(f"Detection success rate: {performance_metrics.get('detection_success_rate', 0):.2f}%\n\n")

            f.write("Score Distribution:\n")
            for score, count in performance_metrics.get('score_counts', {}).items():
                f.write(f"  Score {score}: {count} images ({count/len(score_results)*100:.1f}%)\n")

            f.write(f"\nAverage score: {performance_metrics.get('avg_score', 0):.2f}\n")
            f.write(f"Average angle: {performance_metrics.get('avg_angle', 0):.2f} degrees\n")
            f.write(f"Minimum angle: {performance_metrics.get('min_angle', 0):.2f} degrees\n")
            f.write(f"Maximum angle: {performance_metrics.get('max_angle', 0):.2f} degrees\n\n")

            if 'avg_processing_time' in performance_metrics:
                f.write(f"Average processing time: {performance_metrics['avg_processing_time']:.3f} seconds\n\n")

            if 'accuracy' in performance_metrics:
                f.write(f"Agreement with expert scores: {performance_metrics['accuracy']:.2f}%\n\n")
                f.write("Confusion Matrix:\n")
                f.write("  (rows = expert scores, columns = predicted scores)\n")
                cm = np.array(performance_metrics['confusion_matrix'])
                for i in range(3):
                    f.write(f"  Score {i+1}: {cm[i,0]:3d} {cm[i,1]:3d} {cm[i,2]:3d}\n")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results, performance_metrics

# Example usage
if __name__ == "__main__":
    input_folder = "pp"
    output_folder = "p2"

    # Optional: Dictionary of expert scores for comparison, if available
    # Format: {filename: expert_score}
    expert_scores = None

    # Run the processing
    results, metrics = process_image_folder(input_folder, output_folder, expert_scores)

    # Print performance summary
    print("\nPerformance Metrics Summary:")
    print(f"Detection Success Rate: {metrics.get('detection_success_rate', 0):.2f}%")
    if 'avg_processing_time' in metrics:
        print(f"Average Processing Time: {metrics['avg_processing_time']:.3f} seconds")
    print(f"Average Score: {metrics.get('avg_score', 0):.2f}")
    if 'accuracy' in metrics:
        print(f"Agreement with Expert Scores: {metrics['accuracy']:.2f}%")
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import time
import re
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import pandas as pd
import seaborn as sns

# Helper functions
def calculate_distance(a, b):
    """Calculate the Euclidean distance between two points"""
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)

def calculate_shoulder_width(landmarks, mp_pose):
    """Calculate the shoulder width"""
    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]

    return calculate_distance(left_shoulder, right_shoulder)

def detect_hands_behind_back(landmarks, mp_pose):
    """Detect if hands are behind the back"""
    # Check if wrists are close to each other and behind torso
    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                 landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]

    # Calculate average shoulder position
    mid_shoulder_x = (left_shoulder[0] + right_shoulder[0]) / 2

    # Check if both wrists are behind the shoulder plane
    is_left_behind = left_wrist[0] < mid_shoulder_x
    is_right_behind = right_wrist[0] < mid_shoulder_x

    # Check if wrists are at similar depth
    wrist_distance = calculate_distance(left_wrist, right_wrist)
    shoulder_width = calculate_distance(left_shoulder, right_shoulder)

    # For test to be valid, wrists should be somewhat close and behind the body
    return (is_left_behind or is_right_behind) and wrist_distance < shoulder_width * 4

def score_shoulder_mobility(landmarks, mp_pose):
    """
    Score the shoulder mobility test based on ratio of hand distance to shoulder width:
    - 3 points: Hand distance/shoulder width < 2
    - 2 points: Hand distance/shoulder width between 2 and 2.5
    - 1 point: Hand distance/shoulder width > 2.5
    """
    # Get wrist positions for hand measurement
    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                 landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

    # Calculate distance between hands
    hand_distance = calculate_distance(left_wrist, right_wrist)

    # Calculate shoulder width as reference
    shoulder_width = calculate_shoulder_width(landmarks, mp_pose)

    # Check if hands appear to be behind back
    is_behind_back = detect_hands_behind_back(landmarks, mp_pose)

    # Calculate the ratio of hand distance to shoulder width
    distance_ratio = hand_distance / shoulder_width if shoulder_width > 0 else float('inf')

    # Determine score based on the ratio
    if not is_behind_back:
        score = 1
        feedback = "Hands not positioned correctly behind back"
    elif distance_ratio < 1.7:  # Ratio < 2
        score = 3
        feedback = f"Excellent mobility! Ratio: {distance_ratio:.2f}"
    elif distance_ratio <= 2:  # 2 <= Ratio <= 2.5
        score = 2
        feedback = f"Good mobility. Ratio: {distance_ratio:.2f}"
    else:  # Ratio > 2
        score = 1
        feedback = f"Limited mobility. Ratio: {distance_ratio:.2f}"

    return {
        'score': score,
        'feedback': feedback,
        'hand_distance': hand_distance,
        'shoulder_width': shoulder_width,
        'distance_value': distance_ratio,  # Now this value is the ratio
        'is_behind_back': is_behind_back
    }

def extract_ground_truth_from_filename(filename):
    """
    Extract ground truth score from the first digit in the filename
    Returns None if no digit is found
    """
    match = re.search(r'^\d', filename)
    if match:
        return int(match.group(0))
    return None

def evaluate_performance(score_results, processing_times=None):
    """
    Calculate and visualize performance metrics:
    1. Detection success rate
    2. Confusion matrix (compared to ground truth scores from filenames)
    3. End-to-end processing time
    4. Score frequency analysis

    Args:
        score_results: List of dictionaries with scoring results and ground truth
        processing_times: List of processing times for each image
    """
    # Create output directory for performance metrics
    performance_dir = "performance_metrics"
    if not os.path.exists(performance_dir):
        os.makedirs(performance_dir)

    # 1. Detection Success Rate
    total_images = len(score_results)
    detection_success_rate = 100.0  # Assume all were processed successfully

    # 2. Confusion Matrix using ground truth from filenames
    y_true = [result['ground_truth'] for result in score_results if result['ground_truth'] is not None]
    y_pred = [result['score'] for result in score_results if result['ground_truth'] is not None]

    if len(y_true) > 0:
        # Create confusion matrix
        cm = confusion_matrix(y_true, y_pred, labels=[1, 2, 3])

        # Plot confusion matrix
        plt.figure(figsize=(10, 8))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Score 1', 'Score 2', 'Score 3'])
        disp.plot(cmap=plt.cm.Blues)
        plt.title('Confusion Matrix: Algorithm vs Ground Truth Scores')
        plt.savefig(os.path.join(performance_dir, 'confusion_matrix.png'))
        plt.close()

        # Calculate accuracy
        accuracy = sum(1 for i, j in zip(y_true, y_pred) if i == j) / len(y_true)
        print(f"Accuracy compared to ground truth scores: {accuracy:.2%}")

        # Create detailed comparison table
        comparison_df = pd.DataFrame({
            'File': [result['file'] for result in score_results if result['ground_truth'] is not None],
            'Algorithm Score': [result['score'] for result in score_results if result['ground_truth'] is not None],
            'Ground Truth Score': [result['ground_truth'] for result in score_results if result['ground_truth'] is not None]
        })

        # Save comparison to CSV
        comparison_df.to_csv(os.path.join(performance_dir, 'score_comparison.csv'), index=False)

        # Create a heatmap of the confusion matrix with percentages
        plt.figure(figsize=(10, 8))
        cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100
        sns.heatmap(cm_percent, annot=True, fmt='.1f', cmap='Blues',
                   xticklabels=['Score 1', 'Score 2', 'Score 3'],
                   yticklabels=['Score 1', 'Score 2', 'Score 3'])
        plt.xlabel('Predicted Score')
        plt.ylabel('True Score')
        plt.title('Confusion Matrix (Percentage)')
        plt.savefig(os.path.join(performance_dir, 'confusion_matrix_percent.png'))
        plt.close()

    # 3. End-to-end Processing Time
    if processing_times:
        avg_time = sum(processing_times) / len(processing_times)
        min_time = min(processing_times)
        max_time = max(processing_times)

        # Plot processing time distribution
        plt.figure(figsize=(10, 6))
        plt.hist(processing_times, bins=20, color='green', alpha=0.7)
        plt.axvline(avg_time, color='red', linestyle='--', label=f'Mean: {avg_time:.2f}s')
        plt.title('Processing Time Distribution')
        plt.xlabel('Time (seconds)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.legend()
        plt.savefig(os.path.join(performance_dir, 'processing_time.png'))
        plt.close()

        print(f"Average processing time: {avg_time:.2f} seconds per image")
        print(f"Min processing time: {min_time:.2f} seconds")
        print(f"Max processing time: {max_time:.2f} seconds")

        # Save processing times to CSV
        times_df = pd.DataFrame({
            'File': [result['file'] for result in score_results],
            'Processing Time (s)': processing_times
        })
        times_df.to_csv(os.path.join(performance_dir, 'processing_times.csv'), index=False)

    # 4. Score Frequency Analysis
    scores = [r['score'] for r in score_results]

    # Create score distribution chart
    plt.figure(figsize=(10, 6))
    score_counts = {1: 0, 2: 0, 3: 0}
    for s in scores:
        score_counts[s] += 1

    # Bar chart
    plt.bar(score_counts.keys(), score_counts.values(), color=['#FF6B6B', '#FFD166', '#06D6A0'])
    plt.title('Shoulder Mobility Test Score Distribution')
    plt.xlabel('Score')
    plt.ylabel('Number of Images')
    plt.xticks([1, 2, 3])
    plt.grid(axis='y', alpha=0.3)

    # Add percentage labels
    total = sum(score_counts.values())
    for i, count in score_counts.items():
        percentage = count / total * 100 if total > 0 else 0
        plt.text(i, count + 0.5, f"{percentage:.1f}%", ha='center')

    plt.savefig(os.path.join(performance_dir, 'score_distribution.png'))
    plt.close()

    # Create a pie chart for score distribution
    plt.figure(figsize=(8, 8))
    plt.pie(
        score_counts.values(),
        labels=[f"Score {i}: {count}" for i, count in score_counts.items()],
        autopct='%1.1f%%',
        colors=['#FF6B6B', '#FFD166', '#06D6A0'],
        startangle=90
    )
    plt.title('Score Distribution by Percentage')
    plt.savefig(os.path.join(performance_dir, 'score_pie_chart.png'))
    plt.close()

    # Create distance value distribution
    distances = [r['distance_value'] for r in score_results]

    plt.figure(figsize=(10, 6))
    plt.hist(distances, bins=15, color='blue', alpha=0.7)
    plt.title('Hand-to-Shoulder Width Ratio Distribution')
    plt.xlabel('Ratio')
    plt.ylabel('Number of Images')
    plt.grid(True, alpha=0.3)
    plt.axvline(x=1.7, color='green', linestyle='--', label='1.7 (Score 3 threshold)')
    plt.axvline(x=2.0, color='orange', linestyle='--', label='2.0 (Score 2 threshold)')
    plt.legend()
    plt.savefig(os.path.join(performance_dir, 'distance_distribution.png'))
    plt.close()

    # Compare ground truth vs predicted scores
    if len(y_true) > 0:
        fig, ax = plt.subplots(figsize=(10, 6))
        ground_truth_counts = {1: 0, 2: 0, 3: 0}
        for gt in y_true:
            ground_truth_counts[gt] += 1

        width = 0.35
        x = np.arange(3)

        ax.bar(x - width/2, [ground_truth_counts[i+1] for i in range(3)], width, label='Ground Truth')
        ax.bar(x + width/2, [score_counts[i+1] for i in range(3)], width, label='Algorithm')

        ax.set_ylabel('Count')
        ax.set_title('Ground Truth vs Algorithm Score Distribution')
        ax.set_xticks(x)
        ax.set_xticklabels(['Score 1', 'Score 2', 'Score 3'])
        ax.legend()

        plt.savefig(os.path.join(performance_dir, 'ground_truth_vs_algorithm.png'))
        plt.close()

    # Return summary statistics
    return {
        'detection_success_rate': detection_success_rate,
        'total_images': total_images,
        'score_distribution': score_counts,
        'average_processing_time': avg_time if processing_times else None,
        'accuracy': accuracy if len(y_true) > 0 else None
    }

def process_image_folder(input_folder, output_folder):
    """Process all images in a folder and extract ground truth from filenames"""
    # Create output folder
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []
    processing_times = []

    # Process each image
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            start_time = time.time()

            # Extract ground truth from filename
            ground_truth = extract_ground_truth_from_filename(img_file)

            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Process image
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Score the pose
                evaluation = score_shoulder_mobility(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                feedback = evaluation['feedback']
                hand_distance = evaluation['hand_distance']
                shoulder_width = evaluation['shoulder_width']
                distance_value = evaluation['distance_value']
                is_behind_back = evaluation['is_behind_back']

                # Record scores with ground truth
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'distance_value': distance_value,
                    'ground_truth': ground_truth
                })

                # Get key positions for drawing
                landmarks = results.pose_landmarks.landmark
                left_wrist = (int(landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x * img.shape[1]),
                             int(landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y * img.shape[0]))
                right_wrist = (int(landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x * img.shape[1]),
                              int(landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y * img.shape[0]))
                left_shoulder = (int(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x * img.shape[1]),
                                int(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y * img.shape[0]))
                right_shoulder = (int(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x * img.shape[1]),
                                 int(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y * img.shape[0]))

                # Create a copy of the original image for drawing
                output_img = img.copy()

                # Draw pose landmarks
                mp_drawing.draw_landmarks(output_img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # Highlight wrists
                cv2.circle(output_img, left_wrist, 10, (0, 165, 255), -1)
                cv2.circle(output_img, right_wrist, 10, (0, 165, 255), -1)

                # Draw line between wrists
                cv2.line(output_img, left_wrist, right_wrist, (0, 165, 255), 2)

                # Draw shoulder width reference
                cv2.line(output_img, left_shoulder, right_shoulder, (0, 255, 0), 2)

                # Add just the score in the corner as requested
                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(output_img, f"Score: {score}", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, score_color, 2, cv2.LINE_AA)

                # Save image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, output_img)

            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                output_img = img.copy()
                cv2.putText(output_img, "NO POSE DETECTED", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 2, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, output_img)

                # Still record this file with no score
                score_results.append({
                    'file': img_file,
                    'score': None,
                    'feedback': "No pose detected",
                    'distance_value': None,
                    'ground_truth': ground_truth
                })

            # Record processing time
            end_time = time.time()
            processing_times.append(end_time - start_time)

    # Evaluate performance
    performance_stats = evaluate_performance(score_results, processing_times)

    # Print summary
    print(f"\nAll images processed and saved to {output_folder}")
    print(f"Detection success rate: {performance_stats['detection_success_rate']:.2f}%")
    if processing_times:
        print(f"Average processing time: {performance_stats['average_processing_time']:.2f} seconds per image")
    print("Score distribution:")
    for score, count in performance_stats['score_distribution'].items():
        print(f"  Score {score}: {count} images ({count/performance_stats['total_images']*100:.1f}%)")
    if performance_stats['accuracy'] is not None:
        print(f"Accuracy compared to ground truth: {performance_stats['accuracy']:.2f}")

    return score_results, performance_stats

# Example usage
if __name__ == "__main__":
    input_folder = "j"
    output_folder = "j1"

    results, performance = process_image_folder(input_folder, output_folder)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import time
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Helper functions
def calculate_distance(a, b):
    """Calculate the Euclidean distance between two points"""
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)

def calculate_shoulder_width(landmarks, mp_pose):
    """Calculate the shoulder width"""
    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]

    return calculate_distance(left_shoulder, right_shoulder)

def score_squat_mobility(landmarks, mp_pose):
    """
    Score the squat mobility test:
    - 3 points: Hip Y >= average knee Y (deep squat, hip at or below knee level)
    - 2 points: Hip Y < average knee Y and difference (avg_knee_y - mid_hip_y) > 0.5*shoulder_width
                but <= shoulder_width
    - 1 point: Hip Y < average knee Y and difference (avg_knee_y - mid_hip_y) > shoulder width
    """
    # Get hip positions
    left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
    right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]

    # Calculate middle hip point
    mid_hip_x = (left_hip[0] + right_hip[0]) / 2
    mid_hip_y = (left_hip[1] + right_hip[1]) / 2

    # Get knee positions
    left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
    right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                 landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]

    # Calculate average knee Y position
    avg_knee_y = (left_knee[1] + right_knee[1]) / 2

    # Calculate shoulder width as reference for scoring
    shoulder_width = calculate_shoulder_width(landmarks, mp_pose)

    # Calculate difference between average knee Y and hip Y
    # Note: In image coordinates, Y increases downward
    # If hip is higher than knees (poor squat), avg_knee_y > mid_hip_y
    hip_knee_diff = avg_knee_y - mid_hip_y

    # Determine score based on the criteria
    if mid_hip_y >= avg_knee_y:  # Hip Y >= average knee Y (hip at or below knee level)
        score = 3
        feedback = "Excellent squat mobility! Hip is at or below knee level."
    elif hip_knee_diff > 0.4 * shoulder_width:  # Hip is much higher than knees
        score = 1
        feedback = "Limited squat mobility. Hip is above knee level by more than shoulder width."
    elif hip_knee_diff > 0.1 * shoulder_width:  # Hip is moderately higher than knees
        score = 2
        feedback = "Moderate squat mobility. Hip is above knee level by 0.5-1x shoulder width."
    else:  # Hip is slightly higher than knees
        score = 3
        feedback = "Good squat mobility. Hip is only slightly above knee level."

    return {
        'score': score,
        'feedback': feedback,
        'hip_knee_diff': hip_knee_diff,
        'shoulder_width': shoulder_width,
        'hip_height': mid_hip_y,
        'avg_knee_height': avg_knee_y,
        'hip_knee_ratio': hip_knee_diff / shoulder_width if shoulder_width > 0 else float('inf'),
        'mid_hip_x': mid_hip_x,
        'left_knee': left_knee,
        'right_knee': right_knee
    }

def extract_ground_truth_from_filename(filename):
    """Extract the ground truth score from filename (first digit)"""
    for char in filename:
        if char.isdigit():
            return int(char)
    return None  # No digit found in filename

def evaluate_performance(score_results, processing_times, detection_failures):
    """
    Calculate performance metrics:
    1. Detection success rate
    2. Confusion matrix (compared to ground truth from filenames)
    3. End-to-end processing time
    4. Score frequency analysis
    """
    performance_metrics = {}

    # 1. Detection success rate
    total_images = len(score_results) + len(detection_failures)
    success_rate = len(score_results) / total_images if total_images > 0 else 0
    performance_metrics['detection_success_rate'] = success_rate

    # 2. Confusion matrix (only for successfully detected images)
    y_true = []
    y_pred = []

    for result in score_results:
        gt_score = extract_ground_truth_from_filename(result['file'])
        if gt_score is not None:
            y_true.append(gt_score)
            y_pred.append(result['score'])

    if len(y_true) > 0:
        # Create confusion matrix
        cm = confusion_matrix(y_true, y_pred, labels=[1, 2, 3])

        # Create confusion matrix visualization
        fig_cm = plt.figure(figsize=(8, 6))
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[1, 2, 3])
        disp.plot(cmap=plt.cm.Blues)
        plt.title('Confusion Matrix: Ground Truth vs. Predicted Scores')
        plt.xlabel('Predicted Score')
        plt.ylabel('Ground Truth Score')

        performance_metrics['confusion_matrix'] = cm
        performance_metrics['confusion_matrix_fig'] = fig_cm

    # 3. End-to-end processing time
    if processing_times:
        avg_processing_time = sum(processing_times) / len(processing_times)
        performance_metrics['avg_processing_time'] = avg_processing_time

        # Processing time distribution
        fig_time = plt.figure(figsize=(10, 6))
        plt.hist(processing_times, bins=min(20, len(processing_times)), alpha=0.7)
        plt.title('Processing Time Distribution')
        plt.xlabel('Processing Time (seconds)')
        plt.ylabel('Frequency')
        plt.grid(True, alpha=0.3)

        performance_metrics['processing_time_fig'] = fig_time

    # 4. Score frequency analysis
    if score_results:
        scores = [r['score'] for r in score_results]
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] = score_counts.get(s, 0) + 1

        fig_score = plt.figure(figsize=(8, 6))
        plt.bar(score_counts.keys(), score_counts.values(), color=['red', 'yellow', 'green'])
        plt.title('Score Frequency Analysis')
        plt.xlabel('Score')
        plt.ylabel('Count')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)

        performance_metrics['score_distribution'] = score_counts
        performance_metrics['score_distribution_fig'] = fig_score

    return performance_metrics

def process_squat_images(input_folder, output_folder):
    """Process all squat assessment images in a folder"""
    # Create output folder
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []
    processing_times = []
    detection_failures = []

    # Process each image
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            start_time = time.time()

            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Process image
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Get image dimensions
                height, width, _ = img.shape

                # Create output image (same size as input)
                output_img = img.copy()

                # Draw pose landmarks
                mp_drawing.draw_landmarks(output_img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # Score the pose
                evaluation = score_squat_mobility(results.pose_landmarks.landmark, mp_pose)

                score = evaluation['score']
                feedback = evaluation['feedback']
                shoulder_width = evaluation['shoulder_width']
                hip_knee_diff = evaluation['hip_knee_diff']
                mid_hip_x = evaluation['mid_hip_x']
                mid_hip_y = evaluation['hip_height']
                avg_knee_y = evaluation['avg_knee_height']
                left_knee = evaluation['left_knee']
                right_knee = evaluation['right_knee']
                hip_knee_ratio = evaluation['hip_knee_ratio']

                # Record scores
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'hip_knee_diff': hip_knee_diff,
                    'hip_knee_ratio': hip_knee_ratio
                })

                # Draw key points on original image
                landmarks = results.pose_landmarks.landmark

                # Draw hip point
                hip_point = (int(mid_hip_x * width), int(mid_hip_y * height))
                cv2.circle(output_img, hip_point, 10, (0, 0, 255), -1)

                # Draw knee points
                left_knee_point = (int(left_knee[0] * width), int(left_knee[1] * height))
                right_knee_point = (int(right_knee[0] * width), int(right_knee[1] * height))
                cv2.circle(output_img, left_knee_point, 10, (0, 255, 0), -1)
                cv2.circle(output_img, right_knee_point, 10, (0, 255, 0), -1)

                # Draw average knee height line
                avg_knee_line_start = (int(left_knee[0] * width), int(avg_knee_y * height))
                avg_knee_line_end = (int(right_knee[0] * width), int(avg_knee_y * height))
                cv2.line(output_img, avg_knee_line_start, avg_knee_line_end, (0, 255, 0), 2)

                # Draw height difference if hip is above knees
                if mid_hip_y > avg_knee_y:
                    hip_to_knee_line_start = (hip_point[0], hip_point[1])
                    hip_to_knee_line_end = (hip_point[0], int(avg_knee_y * height))
                    cv2.line(output_img, hip_to_knee_line_start, hip_to_knee_line_end, (0, 165, 255), 2)

                # Add only the score to the image
                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(output_img, f"SCORE: {score}/3", (20, 50),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.5, score_color, 2, cv2.LINE_AA)

                # Save processed image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, output_img)

            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                detection_failures.append(img_file)

                # Create error image
                error_img = img.copy()
                cv2.putText(error_img, "NO POSE DETECTED", (img.shape[1]//4, img.shape[0]//2),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 2, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, error_img)

            # Record processing time
            end_time = time.time()
            processing_times.append(end_time - start_time)

    # Performance evaluation
    performance_metrics = evaluate_performance(score_results, processing_times, detection_failures)

    # Save confusion matrix
    if 'confusion_matrix_fig' in performance_metrics:
        confusion_matrix_path = os.path.join(output_folder, "confusion_matrix.png")
        performance_metrics['confusion_matrix_fig'].savefig(confusion_matrix_path)
        plt.close(performance_metrics['confusion_matrix_fig'])

    # Save processing time distribution
    if 'processing_time_fig' in performance_metrics:
        time_dist_path = os.path.join(output_folder, "processing_time_distribution.png")
        performance_metrics['processing_time_fig'].savefig(time_dist_path)
        plt.close(performance_metrics['processing_time_fig'])

    # Save score distribution
    if 'score_distribution_fig' in performance_metrics:
        score_dist_path = os.path.join(output_folder, "score_distribution.png")
        performance_metrics['score_distribution_fig'].savefig(score_dist_path)
        plt.close(performance_metrics['score_distribution_fig'])

    # Print performance summary
    print("\n===== PERFORMANCE SUMMARY =====")
    print(f"Detection Success Rate: {performance_metrics.get('detection_success_rate', 0):.2%}")
    print(f"Average Processing Time: {performance_metrics.get('avg_processing_time', 0):.2f} seconds per image")
    if 'confusion_matrix' in performance_metrics:
        print("\nConfusion Matrix:")
        print(performance_metrics['confusion_matrix'])
    if 'score_distribution' in performance_metrics:
        print("\nScore Distribution:")
        for score, count in sorted(performance_metrics['score_distribution'].items()):
            print(f"Score {score}: {count} images")

    print(f"\nAll images processed and saved to {output_folder}")
    print(f"Performance metrics saved as images in {output_folder}")

    return score_results, performance_metrics

# Example usage
if __name__ == "__main__":
    input_folder = "s"
    output_folder = "s2"
    results, metrics = process_squat_images(input_folder, output_folder)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import time
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Helper functions
def calculate_distance(a, b):
    """Calculate the Euclidean distance between two points"""
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)

def calculate_shoulder_width(landmarks, mp_pose):
    """Calculate the shoulder width"""
    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]

    return calculate_distance(left_shoulder, right_shoulder)

def calculate_angle(a, b, c):
    """Calculate the angle between three points"""
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    ba = a - b
    bc = c - b

    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    cosine_angle = np.clip(cosine_angle, -1.0, 1.0)  # Ensure the value is in range [-1, 1]

    angle = np.arccos(cosine_angle)
    angle = np.degrees(angle)

    return angle

def score_lunge_mobility(landmarks, mp_pose):
    """
    Score the lunge mobility test:
    - 1 point: Upper body tilted left/right >5 degrees from vertical
    - 2 points: Upper body vertical (±5 degrees) but potential forward tilt
    - 3 points: Upper body vertical (±5 degrees) with optimal alignment
    """
    # Get key points
    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
    left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
    right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
    left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
    right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                 landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]

    # Calculate mid-points
    mid_shoulder_x = (left_shoulder[0] + right_shoulder[0]) / 2
    mid_shoulder_y = (left_shoulder[1] + right_shoulder[1]) / 2
    mid_hip_x = (left_hip[0] + right_hip[0]) / 2
    mid_hip_y = (left_hip[1] + right_hip[1]) / 2

    # Calculate lateral (left-right) tilt angle
    # Use the angle between vertical line and the line connecting mid-hips to mid-shoulders
    vertical_top = [mid_hip_x, 0]  # Point directly above hip
    vertical_bottom = [mid_hip_x, 1]  # Point directly below hip

    # Angle between vertical line and shoulder-hip line
    shoulder_hip_line = [mid_shoulder_x, mid_shoulder_y]
    vertical_line_top = vertical_top
    origin_point = [mid_hip_x, mid_hip_y]

    lateral_tilt_angle = calculate_angle(vertical_line_top, origin_point, shoulder_hip_line)

    # Adjust angle to be negative if tilted left, positive if tilted right
    if mid_shoulder_x < mid_hip_x:  # Tilted left
        lateral_tilt_angle = -lateral_tilt_angle

    # For forward tilt, we'll use the relative position of shoulders compared to hips
    # This is a very rough approximation since we only have frontal view
    # Look for a disproportionate shortening of the torso length compared to shoulder width

    # Calculate torso length (shoulder to hip)
    torso_length = abs(mid_shoulder_y - mid_hip_y)
    shoulder_width = calculate_distance(left_shoulder, right_shoulder)

    # Calculate a torso ratio (length/width)
    # A lower ratio might indicate forward tilt as torso appears shorter from front view when leaning forward
    torso_ratio = torso_length / shoulder_width if shoulder_width > 0 else 0

    # Threshold for potential forward tilt (needs calibration with actual data)
    forward_tilt_threshold = 1.5  # This is an arbitrary value, needs adjustment

    # Another indicator: front knee position relative to front ankle
    # In a good lunge, front knee should be over ankle, not too far forward
    # Determine which leg is forward (lower knee Y position)
    if left_knee[1] < right_knee[1]:  # Left leg is forward
        front_knee = left_knee
        front_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
    else:  # Right leg is forward
        front_knee = right_knee
        front_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                      landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]

    # Calculate horizontal distance between knee and ankle
    knee_ankle_x_diff = abs(front_knee[0] - front_ankle[0])
    knee_forward_indicator = knee_ankle_x_diff / shoulder_width if shoulder_width > 0 else 0

    # Combine indicators for potential forward lean assessment
    forward_tilt_indicator = (1 / torso_ratio if torso_ratio > 0 else 0) + knee_forward_indicator

    # Score based on criteria
    if abs(lateral_tilt_angle) > 5:
        score = 1
        feedback = f"Upper body is tilted to the {'left' if lateral_tilt_angle < 0 else 'right'} by {abs(lateral_tilt_angle):.1f} deg. Stand straighter with upper body aligned vertically."
    elif forward_tilt_indicator > forward_tilt_threshold:
        score = 2
        feedback = "Upper body appears to have some forward tilt. Try maintaining a more upright posture during the lunge."
    else:
        score = 3
        feedback = "Excellent lunge position with good upper body alignment."

    return {
        'score': score,
        'feedback': feedback,
        'lateral_tilt_angle': lateral_tilt_angle,
        'forward_tilt_indicator': forward_tilt_indicator,
        'torso_ratio': torso_ratio,
        'knee_forward_indicator': knee_forward_indicator,
        'mid_shoulder': [mid_shoulder_x, mid_shoulder_y],
        'mid_hip': [mid_hip_x, mid_hip_y]
    }

def evaluate_performance(score_results, total_images):
    """
    Calculate performance metrics:
    1. Detection success rate
    2. Confusion matrix analysis
    3. End-to-end processing time
    4. Score frequency analysis
    """
    performance = {}

    # 1. Detection success rate
    detection_rate = len(score_results) / total_images * 100 if total_images > 0 else 0
    performance['detection_rate'] = detection_rate

    # 2. End-to-end processing time
    if score_results and 'processing_time' in score_results[0]:
        avg_time = sum(result['processing_time'] for result in score_results) / len(score_results)
        performance['avg_processing_time'] = avg_time
        min_time = min(result['processing_time'] for result in score_results)
        max_time = max(result['processing_time'] for result in score_results)
        performance['min_processing_time'] = min_time
        performance['max_processing_time'] = max_time

    # 3. Score frequency analysis
    if score_results:
        scores = [result['score'] for result in score_results]
        score_counts = {1: 0, 2: 0, 3: 0}
        for score in scores:
            score_counts[score] += 1
        performance['score_counts'] = score_counts
        performance['avg_score'] = sum(scores) / len(scores)

    # 4. Confusion matrix
    # Extract ground truth scores from filenames (first digit of filename)
    if score_results:
        ground_truth = []
        predicted = []

        for result in score_results:
            filename = result['file']
            try:
                # Get first digit from the filename as ground truth score
                true_score = int(filename[0])
                if 1 <= true_score <= 3:  # Ensure it's a valid score
                    ground_truth.append(true_score)
                    predicted.append(result['score'])
            except (ValueError, IndexError):
                # Skip files where ground truth can't be determined
                continue

        if ground_truth and predicted:
            cm = confusion_matrix(ground_truth, predicted, labels=[1, 2, 3])
            performance['confusion_matrix'] = cm

            # Calculate accuracy
            accuracy = sum(1 for i in range(len(ground_truth)) if ground_truth[i] == predicted[i]) / len(ground_truth)
            performance['accuracy'] = accuracy * 100

    return performance

def process_lunge_images(input_folder, output_folder):
    """Process all lunge assessment images in a folder"""
    # Create output folder
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []
    total_images = len(image_files)

    # Process each image
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            start_time = time.time()  # Start timing

            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Process image
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Get image dimensions
                height, width, _ = img.shape

                # Create a copy for output
                output_img = img.copy()

                # Draw pose landmarks
                mp_drawing.draw_landmarks(output_img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # Score the pose
                evaluation = score_lunge_mobility(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                lateral_tilt_angle = evaluation['lateral_tilt_angle']
                forward_tilt_indicator = evaluation['forward_tilt_indicator']
                mid_shoulder = evaluation['mid_shoulder']
                mid_hip = evaluation['mid_hip']

                # Measure processing time
                end_time = time.time()
                processing_time = end_time - start_time

                # Record scores and timing
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'lateral_tilt_angle': lateral_tilt_angle,
                    'forward_tilt_indicator': forward_tilt_indicator,
                    'processing_time': processing_time
                })

                # Add score text (only display the score as requested)
                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(output_img, f"{score}/3", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, score_color, 2, cv2.LINE_AA)

                # Save image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, output_img)

            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                end_time = time.time()
                processing_time = end_time - start_time

                # Save image with just a note about no pose detected
                output_img = img.copy()
                cv2.putText(output_img, "NO POSE DETECTED", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 2, cv2.LINE_AA)

                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, output_img)

    # Generate performance evaluation
    performance_metrics = evaluate_performance(score_results, total_images)

    # Generate summary report and visualizations
    if score_results:
        # 1. Score frequency analysis chart
        plt.figure(figsize=(10, 6))
        score_counts = performance_metrics['score_counts']
        plt.bar(score_counts.keys(), score_counts.values(), color=['red', 'yellow', 'green'])
        plt.title('Lunge Assessment Score Distribution')
        plt.xlabel('Score')
        plt.ylabel('Number of Images')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)
        plt.close()

        # 2. Confusion matrix visualization
        if 'confusion_matrix' in performance_metrics:
            plt.figure(figsize=(10, 8))
            cm = performance_metrics['confusion_matrix']

            # Create heatmap using seaborn for better visualization
            ax = plt.subplot()
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=[1, 2, 3],
                       yticklabels=[1, 2, 3], cbar=True)

            # Adjust labels and title
            plt.xlabel('Predicted Score')
            plt.ylabel('Expert Score (Ground Truth)')
            plt.title('Confusion Matrix')

            # Save the figure
            cm_path = os.path.join(output_folder, "confusion_matrix.png")
            plt.tight_layout()
            plt.savefig(cm_path)
            plt.close()

        # 3. Processing time histogram
        plt.figure(figsize=(10, 6))
        processing_times = [result['processing_time'] for result in score_results]
        plt.hist(processing_times, bins=10, color='blue', alpha=0.7)
        plt.title('Processing Time Distribution')
        plt.xlabel('Time (seconds)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        time_chart_path = os.path.join(output_folder, "processing_time.png")
        plt.savefig(time_chart_path)
        plt.close()

        # 4. Angle distribution
        plt.figure(figsize=(10, 6))
        tilt_angles = [abs(result['lateral_tilt_angle']) for result in score_results]
        plt.hist(tilt_angles, bins=10, color='green', alpha=0.7)
        plt.title('Lateral Tilt Angle Distribution')
        plt.xlabel('Absolute Tilt Angle (deg)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.axvline(x=5, color='red', linestyle='--', label='5 deg Threshold')
        plt.legend()
        angle_chart_path = os.path.join(output_folder, "tilt_angle_distribution.png")
        plt.savefig(angle_chart_path)
        plt.close()

        # Write performance summary report
        with open(os.path.join(output_folder, "performance_summary.txt"), 'w') as f:
            f.write("LUNGE ASSESSMENT PERFORMANCE SUMMARY\n")
            f.write("===================================\n\n")

            f.write(f"Total images: {total_images}\n")
            f.write(f"Images with detected pose: {len(score_results)}\n")
            f.write(f"Detection success rate: {performance_metrics['detection_rate']:.2f}%\n\n")

            if 'avg_processing_time' in performance_metrics:
                f.write(f"Processing time:\n")
                f.write(f"  Average: {performance_metrics['avg_processing_time']:.4f} seconds\n")
                f.write(f"  Minimum: {performance_metrics['min_processing_time']:.4f} seconds\n")
                f.write(f"  Maximum: {performance_metrics['max_processing_time']:.4f} seconds\n\n")

            f.write("Score distribution:\n")
            for score, count in performance_metrics['score_counts'].items():
                percentage = count / len(score_results) * 100
                f.write(f"  Score {score}: {count} images ({percentage:.1f}%)\n")
            f.write(f"  Average score: {performance_metrics['avg_score']:.2f}\n\n")

            if 'accuracy' in performance_metrics:
                f.write(f"Agreement with ground truth (from filenames):\n")
                f.write(f"  Accuracy: {performance_metrics['accuracy']:.2f}%\n\n")

                if 'confusion_matrix' in performance_metrics:
                    f.write("Confusion Matrix:\n")
                    cm = performance_metrics['confusion_matrix']
                    f.write("  Expert Score (rows) vs Predicted Score (columns)\n")
                    f.write("      | 1  | 2  | 3  |\n")
                    f.write("  ----+----+----+----+\n")
                    for i in range(3):
                        f.write(f"   {i+1}  | {cm[i, 0]:2d} | {cm[i, 1]:2d} | {cm[i, 2]:2d} |\n")

        print(f"Charts and performance summary saved to {output_folder}")

    return score_results, performance_metrics

# Example usage
if __name__ == "__main__":
    input_folder = "lung"
    output_folder = "lunge_results"
    results, metrics = process_lunge_images(input_folder, output_folder)

    # Print key performance metrics
    print("\nPERFORMANCE SUMMARY:")
    print(f"Detection success rate: {metrics['detection_rate']:.2f}%")
    if 'avg_processing_time' in metrics:
        print(f"Average processing time: {metrics['avg_processing_time']:.4f} seconds")
    if 'accuracy' in metrics:
        print(f"Accuracy (vs ground truth): {metrics['accuracy']:.2f}%")
    print(f"Average score: {metrics['avg_score']:.2f}")
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm
import time
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Define calculation functions
def calculate_angle(a, b, c):
    """Calculate the angle between three points"""
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle

    return angle

def check_hands_horizontal(landmarks, mp_pose):
    """
    Check if hands are approximately horizontal using a height-to-width ratio
    Returns True if the vertical difference between hands is less than 1/10 of the horizontal distance
    """
    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                 landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

    # Calculate vertical difference between hands
    vertical_diff = abs(left_wrist[1] - right_wrist[1])

    # Calculate horizontal distance between hands
    horizontal_dist = abs(left_wrist[0] - right_wrist[0])

    # Return True if vertical difference is less than 1/10 of horizontal distance
    # (This is a more flexible criterion than using fixed angles)
    if horizontal_dist > 0:
        return vertical_diff < (horizontal_dist * 0.1)
    return False

def check_foot_orientation(landmarks, mp_pose, right_leg=True):
    """Check if raised foot is pointing forward/upward rather than sideways"""
    if right_leg:
        # Check right foot orientation
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
        foot_index = [landmarks[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX.value].y]
        heel = [landmarks[mp_pose.PoseLandmark.RIGHT_HEEL.value].x,
               landmarks[mp_pose.PoseLandmark.RIGHT_HEEL.value].y]
    else:
        # Check left foot orientation
        ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        foot_index = [landmarks[mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].x,
                     landmarks[mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].y]
        heel = [landmarks[mp_pose.PoseLandmark.LEFT_HEEL.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_HEEL.value].y]

    # Calculate the foot direction vector
    foot_vector = [foot_index[0] - heel[0], foot_index[1] - heel[1]]

    # Calculate angle with vertical (facing upward is 0 or 180 degrees)
    angle_with_vertical = np.abs(np.degrees(np.arctan2(foot_vector[0], foot_vector[1])))

    # If foot is pointing forward/upward, angle should be close to 0 or 180
    is_forward = angle_with_vertical < 45 or angle_with_vertical > 135

    return is_forward

def score_hurdle_step(landmarks, mp_pose, right_leg=True):
    """
    Score the hurdle step test with specified criteria:
    - 3 points: Hands horizontal AND foot pointing forward
    - 2 points: Hands horizontal BUT foot pointing sideways
    - 1 point: Hands not horizontal (regardless of foot orientation)
    """
    # Check if hands are horizontal (parallel to ground)
    hands_horizontal = check_hands_horizontal(landmarks, mp_pose)

    # Check foot orientation
    foot_forward = check_foot_orientation(landmarks, mp_pose, right_leg)

    # Scoring logic based on specified criteria
    score = 1  # Default score
    feedback = []

    if not hands_horizontal:
        score = 1
        feedback.append("Hands not horizontal (1/3)")
    elif hands_horizontal and foot_forward:
        score = 3
        feedback.append("Hands horizontal, foot forward (3/3)")
    elif hands_horizontal and not foot_forward:
        score = 2
        feedback.append("Hands horizontal, foot sideways (2/3)")

    return {
        'score': score,
        'feedback': feedback,
        'hands_horizontal': hands_horizontal,
        'foot_forward': foot_forward
    }

def evaluate_performance(results, total_images, expert_scores=None):
    """
    Calculate performance metrics:
    1. Detection success rate
    2. Confusion matrix analysis
    3. End-to-end processing time
    4. Score frequency analysis
    """
    performance = {}

    # 1. Detection success rate
    detection_rate = len(results) / total_images * 100 if total_images > 0 else 0
    performance['detection_rate'] = detection_rate

    # Only continue if we have results
    if not results:
        return performance

    # 2. Processing time analysis
    if 'processing_time' in results[0]:
        processing_times = [r['processing_time'] for r in results]
        performance['avg_processing_time'] = sum(processing_times) / len(processing_times)
        performance['min_processing_time'] = min(processing_times)
        performance['max_processing_time'] = max(processing_times)
        performance['total_processing_time'] = sum(processing_times)

    # 3. Score frequency analysis
    scores = [r['score'] for r in results]
    score_counts = {1: 0, 2: 0, 3: 0}
    for score in scores:
        score_counts[score] = score_counts.get(score, 0) + 1
    performance['score_counts'] = score_counts
    performance['avg_score'] = sum(scores) / len(scores)

    # 4. Confusion matrix analysis (compare with expert scores)
    # If expert_scores is None, try to extract from filenames (first digit)
    if expert_scores is None:
        expert_scores = {}
        for r in results:
            filename = r['file']
            try:
                # Assume first character of filename is the expert score (1, 2, or 3)
                score = int(filename[0])
                if 1 <= score <= 3:  # Valid score range
                    expert_scores[filename] = score
            except (ValueError, IndexError):
                # Skip if we can't extract a valid score
                continue

    if expert_scores:
        true_scores = []
        pred_scores = []

        for r in results:
            if r['file'] in expert_scores:
                true_scores.append(expert_scores[r['file']])
                pred_scores.append(r['score'])

        if true_scores and pred_scores:
            cm = confusion_matrix(true_scores, pred_scores, labels=[1, 2, 3])
            performance['confusion_matrix'] = cm

            # Calculate accuracy
            correct = sum(1 for i in range(len(true_scores)) if true_scores[i] == pred_scores[i])
            performance['accuracy'] = correct / len(true_scores) * 100

    return performance

# Main function to process images from a folder
def process_image_folder(input_folder, output_folder, right_leg=True):
    """
    Process all images in a folder with pose detection and scoring

    Args:
        input_folder: Path to folder containing input images
        output_folder: Path to folder where annotated images will be saved
        right_leg: Whether to evaluate right leg (True) or left leg (False)
    """
    # Make sure output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")

    # Score tracking
    score_results = []
    total_images = len(image_files)

    # Process each image with pose detection
    with mp_pose.Pose(
        static_image_mode=True,  # Use static mode for images
        model_complexity=2,  # Higher accuracy
        min_detection_confidence=0.5
    ) as pose:

        # Use tqdm for progress bar
        for img_file in tqdm(image_files):
            start_time = time.time()  # Start timing

            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)

            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Convert to RGB for MediaPipe
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Process image
            results = pose.process(img_rgb)

            end_time = time.time()  # End timing
            processing_time = end_time - start_time

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Draw pose landmarks on image
                output_img = img.copy()
                mp_drawing.draw_landmarks(
                    output_img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=2),
                    mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2)
                )

                # Score the pose
                evaluation = score_hurdle_step(results.pose_landmarks.landmark, mp_pose, right_leg)
                score = evaluation['score']
                hands_horizontal = evaluation['hands_horizontal']
                foot_forward = evaluation['foot_forward']

                # Track scores
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'hands_horizontal': hands_horizontal,
                    'foot_forward': foot_forward,
                    'processing_time': processing_time
                })

                # Get image dimensions for scaling text size
                height, width, _ = img.shape
                font_scale = min(width, height) / 500  # Increased font size
                thickness = max(2, int(font_scale * 2))  # Thicker text lines

                # Add only the score with larger text
                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(output_img, f"Score: {score}/3", (20, 60),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale*1.5, score_color, thickness, cv2.LINE_AA)

                # Save annotated image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, output_img)
            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                output_img = img.copy()

                # Get image dimensions for scaling text size
                height, width, _ = img.shape
                font_scale = min(width, height) / 500
                thickness = max(2, int(font_scale * 2))

                cv2.putText(output_img, "NO POSE DETECTED", (20, 60),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale*1.5, (0, 0, 255), thickness, cv2.LINE_AA)

                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, output_img)

    # Generate and save performance evaluation
    performance_metrics = evaluate_performance(score_results, total_images)

    # Generate visualizations and summary report
    if score_results:
        # Calculate statistics
        scores = [r['score'] for r in score_results]

        # Create score distribution chart
        plt.figure(figsize=(10, 6))
        score_counts = performance_metrics['score_counts']
        plt.bar(score_counts.keys(), score_counts.values(), color=['red', 'orange', 'green'])
        plt.title('Hurdle Step Test Score Distribution', fontsize=16)
        plt.xlabel('Score', fontsize=14)
        plt.ylabel('Number of Images', fontsize=14)
        plt.xticks([1, 2, 3], fontsize=12)
        plt.yticks(fontsize=12)
        plt.grid(axis='y', alpha=0.3)
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)
        plt.close()

        # Create processing time histogram
        plt.figure(figsize=(10, 6))
        times = [r['processing_time'] for r in score_results]
        plt.hist(times, bins=10, color='blue', alpha=0.7)
        plt.title('Processing Time Distribution', fontsize=16)
        plt.xlabel('Time (seconds)', fontsize=14)
        plt.ylabel('Number of Images', fontsize=14)
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        plt.grid(True, alpha=0.3)
        time_chart_path = os.path.join(output_folder, "processing_time.png")
        plt.savefig(time_chart_path)
        plt.close()

        # If confusion matrix is available, visualize it
        if 'confusion_matrix' in performance_metrics:
            plt.figure(figsize=(10, 8))
            cm = performance_metrics['confusion_matrix']

            # Use seaborn for better visualization
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                      xticklabels=['Score 1', 'Score 2', 'Score 3'],
                      yticklabels=['Score 1', 'Score 2', 'Score 3'])

            plt.title('Confusion Matrix', fontsize=16)
            plt.xlabel('Predicted Score', fontsize=14)
            plt.ylabel('Expert Score (Ground Truth)', fontsize=14)
            plt.xticks(fontsize=12)
            plt.yticks(fontsize=12)

            plt.tight_layout()
            cm_path = os.path.join(output_folder, "confusion_matrix.png")
            plt.savefig(cm_path)
            plt.close()

        # Write performance summary to file
        report_path = os.path.join(output_folder, "performance_summary.txt")
        with open(report_path, 'w') as f:
            f.write("HURDLE STEP TEST PERFORMANCE SUMMARY\n")
            f.write("===================================\n\n")

            f.write(f"Total images analyzed: {total_images}\n")
            f.write(f"Images with detected pose: {len(score_results)}\n")
            f.write(f"Detection success rate: {performance_metrics['detection_rate']:.2f}%\n\n")

            if 'avg_processing_time' in performance_metrics:
                f.write(f"Processing time statistics:\n")
                f.write(f"  Average: {performance_metrics['avg_processing_time']:.4f} seconds\n")
                f.write(f"  Minimum: {performance_metrics['min_processing_time']:.4f} seconds\n")
                f.write(f"  Maximum: {performance_metrics['max_processing_time']:.4f} seconds\n")
                f.write(f"  Total: {performance_metrics['total_processing_time']:.4f} seconds\n\n")

            f.write(f"Score distribution:\n")
            for score, count in performance_metrics['score_counts'].items():
                percentage = (count / len(scores)) * 100 if scores else 0
                f.write(f"  Score {score}: {count} images ({percentage:.1f}%)\n")

            f.write(f"  Average score: {performance_metrics['avg_score']:.2f}/3\n\n")

            if 'accuracy' in performance_metrics:
                f.write(f"Agreement with expert scores:\n")
                f.write(f"  Accuracy: {performance_metrics['accuracy']:.2f}%\n\n")

                if 'confusion_matrix' in performance_metrics:
                    f.write("Confusion Matrix:\n")
                    f.write("  Expert Score (rows) vs. Predicted Score (columns)\n")
                    f.write("      | 1  | 2  | 3  |\n")
                    f.write("  ----+----+----+----+\n")

                    cm = performance_metrics['confusion_matrix']
                    for i in range(3):
                        f.write(f"   {i+1}  | {cm[i, 0]:2d} | {cm[i, 1]:2d} | {cm[i, 2]:2d} |\n")

        print(f"\nReport saved to {report_path}")
        print(f"Charts saved to {output_folder}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results, performance_metrics

# Example usage
if __name__ == "__main__":
    # Set these paths to your folders
    input_folder = "stepp"
    output_folder = "stepp2"

    # Set which leg to evaluate (True for right leg, False for left leg)
    right_leg = True

    # Process the images
    results, metrics = process_image_folder(input_folder, output_folder, right_leg)

    # Print performance summary
    print("\nPERFORMANCE SUMMARY:")
    print(f"Detection success rate: {metrics['detection_rate']:.2f}%")
    if 'avg_processing_time' in metrics:
        print(f"Average processing time: {metrics['avg_processing_time']:.4f} seconds")
    if 'accuracy' in metrics:
        print(f"Agreement with expert scores: {metrics['accuracy']:.2f}%")
    print(f"Average score: {metrics['avg_score']:.2f}/3")
```
