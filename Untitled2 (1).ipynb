```python
# 导入必要的库
import cv2
import mediapipe as mp
import numpy as np
import time
import matplotlib.pyplot as plt

# 定义计算角度的函数
def calculate_angle(a, b, c):
    """
    计算三个点形成的角度
    """
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle

    return angle

# 设置MediaPipe模型
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

# 输入视频路径
input_path = 'test2.mp4'  # 修改为您的视频文件路径
output_path = 'output_video2.mp4'    # 可选，输出视频路径

# 打开视频文件
cap = cv2.VideoCapture(input_path)
if not cap.isOpened():
    print(f"错误：无法打开视频文件 {input_path}")
else:
    # 获取视频属性
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    print(f"视频信息: 宽={width}, 高={height}, FPS={fps}, 总帧数={total_frames}")

    # 创建视频写入器
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    # 初始化变量
    frame_count = 0
    pTime = 0
    angles_history = []  # 存储角度数据

    # 创建姿态检测模型
    with mp_pose.Pose(
        static_image_mode=False,
        smooth_landmarks=True,
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5
    ) as pose:

        # 处理每一帧
        while cap.isOpened():
            success, img = cap.read()

            if not success:
                print("视频处理完成或读取错误")
                break

            # 将BGR格式转为RGB格式
            imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # 处理图像
            results = pose.process(imgRGB)

            # 在原图上绘制姿态信息
            if results.pose_landmarks:
                # 绘制关键点和连接线
                mp_drawing.draw_landmarks(
                    img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(255,0,0), thickness=2, circle_radius=2),
                    mp_drawing.DrawingSpec(color=(0,255,0), thickness=2, circle_radius=2)
                )

                # 获取关键点
                landmarks = results.pose_landmarks.landmark

                # 计算各种角度
                try:
                    # 右肩-右肘-右腕角度
                    shoulder_r = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                                 landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
                    elbow_r = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
                              landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
                    wrist_r = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                              landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]
                    angle_r_arm = calculate_angle(shoulder_r, elbow_r, wrist_r)

                    # 右髋-右膝-右踝角度
                    hip_r = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                            landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
                    knee_r = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                              landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
                    ankle_r = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                              landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
                    angle_r_leg = calculate_angle(hip_r, knee_r, ankle_r)

                    # 在图像上显示角度
                    # 右肘角度
                    cv2.putText(img, f"{int(angle_r_arm)}",
                               tuple(np.multiply(elbow_r, [width, height]).astype(int)),
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

                    # 右膝角度
                    cv2.putText(img, f"{int(angle_r_leg)}",
                               tuple(np.multiply(knee_r, [width, height]).astype(int)),
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2, cv2.LINE_AA)

                    # 存储角度数据
                    current_angles = {
                        'frame': frame_count,
                        'right_arm': angle_r_arm,
                        'right_leg': angle_r_leg
                    }
                    angles_history.append(current_angles)

                except:
                    print(f"帧 {frame_count}: 无法计算某些角度")
            else:
                print(f"帧 {frame_count}: 未检测到姿态关键点")

            # 显示进度
            frame_count += 1
            if frame_count % 30 == 0:  # 每30帧显示一次进度
                print(f"处理进度: {frame_count}/{total_frames} ({frame_count/total_frames*100:.1f}%)")

            # 计算和显示FPS
            cTime = time.time()
            if pTime > 0:
                fps_current = 1/(cTime-pTime)
                cv2.putText(img, f"FPS: {int(fps_current)}", (10, 30),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
            pTime = cTime

            # 在Jupyter中显示图像
            # 注意：这里使用cv2.cvtColor将BGR转换为RGB以正确显示
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            plt.figure(figsize=(12, 8))
            plt.imshow(img_rgb)
            plt.title(f"Frame {frame_count}")
            plt.axis('off')
            plt.show()

            # 写入输出视频
            out.write(img)

            # 控制处理速度，模拟视频播放
            time.sleep(1/fps)  # 根据原始视频的FPS调整延迟

        # 释放资源
        cap.release()
        out.release()

    print(f"视频处理完成. 处理了 {frame_count} 帧.")

    # 绘制角度变化图表
    if angles_history:
        frames = [item['frame'] for item in angles_history]
        right_arm_angles = [item['right_arm'] for item in angles_history]
        right_leg_angles = [item['right_leg'] for item in angles_history]

        plt.figure(figsize=(12, 6))
        plt.plot(frames, right_arm_angles, 'r-', label='右臂角度')
        plt.plot(frames, right_leg_angles, 'b-', label='右腿角度')
        plt.xlabel('帧')
        plt.ylabel('角度 (度)')
        plt.title('角度变化分析')
        plt.legend()
        plt.grid(True)
        plt.show()

        # 简单的角度评估 (示例)
        # 这里只是一个简单示例，您可以根据特定动作定义自己的评分规则
        def evaluate_angles(arm_angles, leg_angles):
            # 假设我们在评估一个深蹲动作
            # 理想的腿部角度范围和手臂角度范围
            ideal_leg_min, ideal_leg_max = 80, 110  # 膝盖弯曲角度
            ideal_arm_min, ideal_arm_max = 150, 180  # 手臂伸直角度

            # 计算在理想范围内的百分比
            leg_in_range = sum(1 for a in leg_angles if ideal_leg_min <= a <= ideal_leg_max) / len(leg_angles) if leg_angles else 0
            arm_in_range = sum(1 for a in arm_angles if ideal_arm_min <= a <= ideal_arm_max) / len(arm_angles) if arm_angles else 0

            # 计算总评分 (简单加权)
            score = (leg_in_range * 0.7 + arm_in_range * 0.3) * 100  # 70%腿部，30%手臂

            return {
                'score': score,
                'leg_accuracy': leg_in_range * 100,
                'arm_accuracy': arm_in_range * 100,
                'feedback': generate_feedback(leg_angles, arm_angles, ideal_leg_min, ideal_leg_max, ideal_arm_min, ideal_arm_max)
            }

        def generate_feedback(leg_angles, arm_angles, leg_min, leg_max, arm_min, arm_max):
            feedback = []

            # 腿部角度分析
            avg_leg = sum(leg_angles) / len(leg_angles) if leg_angles else 0
            if avg_leg < leg_min:
                feedback.append("膝盖弯曲过度，尝试不要蹲得太低")
            elif avg_leg > leg_max:
                feedback.append("膝盖弯曲不足，尝试蹲得更低")

            # 手臂角度分析
            avg_arm = sum(arm_angles) / len(arm_angles) if arm_angles else 0
            if avg_arm < arm_min:
                feedback.append("手臂弯曲过多，尝试保持更直")
            elif avg_arm > arm_max:
                feedback.append("手臂过度伸直")

            return feedback if feedback else ["姿势良好!"]

        # 获取评估结果
        evaluation = evaluate_angles(right_arm_angles, right_leg_angles)

        print(f"动作评分: {evaluation['score']:.1f}/100")
        print(f"腿部准确度: {evaluation['leg_accuracy']:.1f}%")
        print(f"手臂准确度: {evaluation['arm_accuracy']:.1f}%")
        print("反馈建议:")
        for item in evaluation['feedback']:
            print(f"- {item}")
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm  # For progress bar

# Define calculation functions
def calculate_angle(a, b, c):
    """Calculate the angle between three points"""
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle

    return angle

def check_hands_horizontal(landmarks, mp_pose):
    """
    Check if hands are approximately horizontal using a height-to-width ratio
    Returns True if the vertical difference between hands is less than 1/10 of the horizontal distance
    """
    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                 landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

    # Calculate vertical difference between hands
    vertical_diff = abs(left_wrist[1] - right_wrist[1])

    # Calculate horizontal distance between hands
    horizontal_dist = abs(left_wrist[0] - right_wrist[0])

    # Return True if vertical difference is less than 1/10 of horizontal distance
    # (This is a more flexible criterion than using fixed angles)
    if horizontal_dist > 0:
        return vertical_diff < (horizontal_dist * 0.1)
    return False

def check_foot_orientation(landmarks, mp_pose, right_leg=True):
    """Check if raised foot is pointing forward/upward rather than sideways"""
    if right_leg:
        # Check right foot orientation
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
        foot_index = [landmarks[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_FOOT_INDEX.value].y]
        heel = [landmarks[mp_pose.PoseLandmark.RIGHT_HEEL.value].x,
               landmarks[mp_pose.PoseLandmark.RIGHT_HEEL.value].y]
    else:
        # Check left foot orientation
        ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        foot_index = [landmarks[mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].x,
                     landmarks[mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].y]
        heel = [landmarks[mp_pose.PoseLandmark.LEFT_HEEL.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_HEEL.value].y]

    # Calculate the foot direction vector
    foot_vector = [foot_index[0] - heel[0], foot_index[1] - heel[1]]

    # Calculate angle with vertical (facing upward is 0 or 180 degrees)
    angle_with_vertical = np.abs(np.degrees(np.arctan2(foot_vector[0], foot_vector[1])))

    # If foot is pointing forward/upward, angle should be close to 0 or 180
    is_forward = angle_with_vertical < 45 or angle_with_vertical > 135

    return is_forward

def score_hurdle_step(landmarks, mp_pose, right_leg=True):
    """
    Score the hurdle step test with specified criteria:
    - 3 points: Hands horizontal AND foot pointing forward
    - 2 points: Hands horizontal BUT foot pointing sideways
    - 1 point: Hands not horizontal (regardless of foot orientation)
    """
    # Check if hands are horizontal (parallel to ground)
    hands_horizontal = check_hands_horizontal(landmarks, mp_pose)

    # Check foot orientation
    foot_forward = check_foot_orientation(landmarks, mp_pose, right_leg)

    # Scoring logic based on specified criteria
    score = 1  # Default score
    feedback = []

    if not hands_horizontal:
        score = 1
        feedback.append("Hands not horizontal (1/3)")
    elif hands_horizontal and foot_forward:
        score = 3
        feedback.append("Hands horizontal, foot forward (3/3)")
    elif hands_horizontal and not foot_forward:
        score = 2
        feedback.append("Hands horizontal, foot sideways (2/3)")

    return {
        'score': score,
        'feedback': feedback,
        'hands_horizontal': hands_horizontal,
        'foot_forward': foot_forward
    }

# Main function to process images from a folder
def process_image_folder(input_folder, output_folder, right_leg=True):
    """
    Process all images in a folder with pose detection and scoring

    Args:
        input_folder: Path to folder containing input images
        output_folder: Path to folder where annotated images will be saved
        right_leg: Whether to evaluate right leg (True) or left leg (False)
    """
    # Make sure output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")

    # Score tracking
    score_results = []

    # Process each image with pose detection
    with mp_pose.Pose(
        static_image_mode=True,  # Use static mode for images
        model_complexity=2,  # Higher accuracy
        min_detection_confidence=0.5
    ) as pose:

        # Use tqdm for progress bar
        for img_file in tqdm(image_files):
            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)

            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Convert to RGB for MediaPipe
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Process image
            results = pose.process(img_rgb)

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Draw pose landmarks on image
                mp_drawing.draw_landmarks(
                    img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=2),
                    mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2)
                )

                # Score the pose
                evaluation = score_hurdle_step(results.pose_landmarks.landmark, mp_pose, right_leg)
                score = evaluation['score']
                feedback = evaluation['feedback']
                hands_horizontal = evaluation['hands_horizontal']
                foot_forward = evaluation['foot_forward']

                # Track scores
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback
                })

                # Get image dimensions for scaling text size
                height, width, _ = img.shape
                font_scale = min(width, height) / 1000  # Adjust font size based on image size

                # Create semi-transparent overlay (smaller)
                overlay = img.copy()
                overlay_width = min(width//2, 200)
                cv2.rectangle(overlay, (5, 5), (overlay_width, 75), (0, 0, 0), -1)
                alpha = 0.5  # More transparent
                cv2.addWeighted(overlay, alpha, img, 1-alpha, 0, img)

                # Add text (smaller font size)
                leg_text = "Right Leg" if right_leg else "Left Leg"
                cv2.putText(img, f"Leg: {leg_text}", (10, 20),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale*0.8, (255, 255, 255), 1, cv2.LINE_AA)

                cv2.putText(img, f"Score: {score}/3", (10, 35),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale*0.9, (0, 255, 0), 1, cv2.LINE_AA)

                # Add debug info
                cv2.putText(img, f"Hands horizontal: {'Yes' if hands_horizontal else 'No'}", (10, 50),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale*0.7, (255, 255, 255), 1, cv2.LINE_AA)
                cv2.putText(img, f"Foot forward: {'Yes' if foot_forward else 'No'}", (10, 65),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale*0.7, (255, 255, 255), 1, cv2.LINE_AA)

                # Add feedback with even smaller font
                y_pos = 85
                for fb in feedback:
                    # Split long feedback into multiple lines if needed
                    words = fb.split()
                    lines = []
                    current_line = ""

                    for word in words:
                        if len(current_line + " " + word) > 25:  # Shorter max length
                            lines.append(current_line)
                            current_line = word
                        else:
                            if current_line:
                                current_line += " " + word
                            else:
                                current_line = word

                    if current_line:
                        lines.append(current_line)

                    for line in lines:
                        cv2.putText(img, line, (10, y_pos),
                                  cv2.FONT_HERSHEY_SIMPLEX, font_scale*0.6, (0, 165, 255), 1, cv2.LINE_AA)
                        y_pos += int(15 * font_scale)
            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                cv2.putText(img, "No pose detected", (10, 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0, 0, 255), 1, cv2.LINE_AA)

            # Save annotated image
            output_path = os.path.join(output_folder, f"scored_{img_file}")
            cv2.imwrite(output_path, img)

    # Generate and save summary report
    if score_results:
        # Calculate statistics
        scores = [r['score'] for r in score_results]
        avg_score = sum(scores) / len(scores)

        # Count by score
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1

        # Create summary report
        report_path = os.path.join(output_folder, "score_summary.txt")
        with open(report_path, 'w') as f:
            f.write(f"Hurdle Step Test Score Summary\n")
            f.write(f"===========================\n\n")
            f.write(f"Images analyzed: {len(score_results)}\n")
            f.write(f"Average score: {avg_score:.2f}/3\n\n")
            f.write(f"Score distribution:\n")
            for score, count in score_counts.items():
                percentage = (count / len(scores)) * 100 if scores else 0
                f.write(f"  Score {score}: {count} images ({percentage:.1f}%)\n")

            f.write("\nDetailed results:\n")
            for i, result in enumerate(score_results):
                f.write(f"\n{i+1}. {result['file']}\n")
                f.write(f"   Score: {result['score']}/3\n")
                f.write(f"   Feedback: {', '.join(result['feedback'])}\n")

        print(f"\nReport saved to {report_path}")

        # Create score distribution chart
        plt.figure(figsize=(10, 6))
        plt.bar(score_counts.keys(), score_counts.values(), color='green')
        plt.title('Hurdle Step Test Score Distribution')
        plt.xlabel('Score')
        plt.ylabel('Number of Images')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)

        # Save chart
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)
        print(f"Score distribution chart saved to {chart_path}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# Example usage
if __name__ == "__main__":
    # Set these paths to your folders
    input_folder = "zz"
    output_folder = "zz1"

    # Set which leg to evaluate (True for right leg, False for left leg)
    right_leg = True

    # Process the images
    results = process_image_folder(input_folder, output_folder, right_leg)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm  # For progress bar

# Define helper functions
def calculate_distance(a, b):
    """Calculate the Euclidean distance between two points"""
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)

def calculate_angle(a, b, c):
    """
    Calculate the angle between three points
    Args:
        a: First point [x, y]
        b: Middle point (vertex) [x, y]
        c: Last point [x, y]
    Returns:
        Angle in degrees
    """
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle

    return angle

def identify_raised_leg(landmarks, mp_pose):
    """
    Determine which leg is raised based on y-coordinate of ankles
    Returns:
        is_right_leg_raised: Boolean, True if right leg is raised, False if left leg is raised
    """
    left_ankle_y = landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y
    right_ankle_y = landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y

    # The leg with the smaller y-coordinate (higher on screen) is raised
    is_right_leg_raised = right_ankle_y < left_ankle_y

    return is_right_leg_raised

def calculate_leg_trunk_angle(landmarks, mp_pose, is_right_leg_raised):
    """
    Calculate the angle between the raised leg and the trunk
    Args:
        landmarks: MediaPipe pose landmarks
        mp_pose: MediaPipe pose model
        is_right_leg_raised: Boolean, True if right leg is raised, False if left leg is raised
    Returns:
        Angle in degrees
    """
    if is_right_leg_raised:
        # For right leg
        hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
        shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                   landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
    else:
        # For left leg
        hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                   landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]

    # Calculate the angle at the hip between the line from hip to ankle and the line from hip to shoulder
    angle = calculate_angle(ankle, hip, shoulder)

    return angle

def score_leg_raising(landmarks, mp_pose):
    """
    Score the active straight leg raise test:
    - 3 points: Angle between 90-110 degrees
    - 2 points: Angle between 110-150 degrees
    - 1 point: Angle > 150 degrees
    """
    # Identify which leg is raised
    is_right_leg_raised = identify_raised_leg(landmarks, mp_pose)

    # Calculate the leg-trunk angle
    leg_trunk_angle = calculate_leg_trunk_angle(landmarks, mp_pose, is_right_leg_raised)

    # Check if in lying position (simplified check)
    right_hip_y = landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y
    left_hip_y = landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y

    is_lying = abs(right_hip_y - left_hip_y) < 0.1  # Hips should be at similar height if lying

    # Also check if raised leg is straight (no significant knee bend)
    if is_right_leg_raised:
        knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
    else:
        knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]

    knee_angle = calculate_angle(hip, knee, ankle)
    leg_is_straight = knee_angle > 160  # Less than 20 degrees of knee bend

    # Determine score based on angle
    score = 1  # Default score
    feedback = []

    # Debug values
    debug_info = {
        'leg_trunk_angle': leg_trunk_angle,
        'knee_angle': knee_angle,
        'is_lying': is_lying,
        'leg_is_straight': leg_is_straight,
        'is_right_leg_raised': is_right_leg_raised
    }

    if not is_lying:
        score = 1
        feedback.append("Position incorrect. Body should be lying flat.")
    elif not leg_is_straight:
        score = 1
        feedback.append("Leg should be kept straight during the test.")
    elif leg_trunk_angle >= 90 and leg_trunk_angle <= 110:
        score = 3
        feedback.append(f"Excellent! High leg raise ({leg_trunk_angle:.1f}°)")
    elif leg_trunk_angle > 110 and leg_trunk_angle <= 150:
        score = 2
        feedback.append(f"Good leg raise to moderate height ({leg_trunk_angle:.1f}°)")
    else:  # > 150
        score = 1
        feedback.append(f"Limited leg raise. Try to lift higher ({leg_trunk_angle:.1f}°)")

    return {
        'score': score,
        'feedback': feedback,
        'debug_info': debug_info
    }

# Main function to process images from a folder
def process_image_folder(input_folder, output_folder):
    """
    Process all images in a folder with pose detection and leg raising test scoring
    """
    # Make sure output folder exists
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")

    # Score tracking
    score_results = []

    # Process each image with pose detection
    with mp_pose.Pose(
        static_image_mode=True,  # Use static mode for images
        model_complexity=2,  # Higher accuracy
        min_detection_confidence=0.5
    ) as pose:

        # Use tqdm for progress bar
        for img_file in tqdm(image_files):
            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)

            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Convert to RGB for MediaPipe
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

            # Process image
            results = pose.process(img_rgb)

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Get original image dimensions
                orig_height, orig_width, _ = img.shape

                # Create a larger canvas (double the size + space for info)
                new_width = orig_width * 2
                new_height = orig_height * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)

                # Draw pose landmarks on original image
                mp_drawing.draw_landmarks(
                    img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                    mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2, circle_radius=2),
                    mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=2, circle_radius=2)
                )

                # Score the pose
                evaluation = score_leg_raising(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                feedback = evaluation['feedback']
                debug_info = evaluation['debug_info']

                # Track scores
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'leg_trunk_angle': debug_info['leg_trunk_angle']
                })

                # Place the original image in the top left of the new canvas
                new_img[0:orig_height, 0:orig_width] = img

                # Draw the leg-trunk angle on the original image
                # Get hip, ankle, and shoulder points of the raised leg
                landmarks = results.pose_landmarks.landmark
                is_right_leg_raised = debug_info['is_right_leg_raised']

                if is_right_leg_raised:
                    hip_point = (int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x * orig_width),
                                int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y * orig_height))
                    ankle_point = (int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x * orig_width),
                                  int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y * orig_height))
                    shoulder_point = (int(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x * orig_width),
                                     int(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y * orig_height))
                    knee_point = (int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x * orig_width),
                                 int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y * orig_height))
                    leg_label = "Right Leg (Raised)"
                else:
                    hip_point = (int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x * orig_width),
                                int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y * orig_height))
                    ankle_point = (int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x * orig_width),
                                  int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y * orig_height))
                    shoulder_point = (int(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x * orig_width),
                                     int(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y * orig_height))
                    knee_point = (int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x * orig_width),
                                 int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y * orig_height))
                    leg_label = "Left Leg (Raised)"

                # Draw lines for angle visualization
                cv2.line(new_img, hip_point, ankle_point, (0, 255, 255), 2)  # Leg line
                cv2.line(new_img, hip_point, shoulder_point, (0, 255, 255), 2)  # Trunk line

                # Draw angle arc
                angle = debug_info['leg_trunk_angle']
                # Add angle text near hip
                cv2.putText(new_img, f"{angle:.1f}°",
                          (hip_point[0] + 10, hip_point[1] - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2, cv2.LINE_AA)

                # Add score box in the top right
                score_box = np.zeros((orig_height, orig_width, 3), dtype=np.uint8)

                # Fill the score box with a dark background
                cv2.rectangle(score_box, (0, 0), (orig_width, orig_height), (30, 30, 30), -1)

                # Add score information with moderate sized, clear text
                font_scale = 1.0
                line_height = 40

                # Title
                cv2.putText(score_box, "Straight Leg Raise Test", (20, line_height),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 255, 255), 2, cv2.LINE_AA)

                # Which leg is being tested
                cv2.putText(score_box, f"Testing: {leg_label}", (20, line_height*2),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 255, 255), 2, cv2.LINE_AA)

                # Score with colorful highlight
                score_text = f"SCORE: {score}/3"
                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(score_box, score_text, (20, line_height*3),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.5, score_color, 2, cv2.LINE_AA)

                # Angle information
                angle_text = f"Leg-Trunk Angle: {angle:.1f}°"
                cv2.putText(score_box, angle_text, (20, line_height*4),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale, (0, 255, 255), 2, cv2.LINE_AA)

                # Validation info
                position_text = "Position OK" if debug_info['is_lying'] else "Not lying flat!"
                color = (0, 255, 0) if debug_info['is_lying'] else (0, 0, 255)
                cv2.putText(score_box, position_text, (20, line_height*5),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale, color, 2, cv2.LINE_AA)

                leg_text = "Leg straight OK" if debug_info['leg_is_straight'] else "Knee bent too much!"
                color = (0, 255, 0) if debug_info['leg_is_straight'] else (0, 0, 255)
                cv2.putText(score_box, leg_text, (20, line_height*6),
                          cv2.FONT_HERSHEY_SIMPLEX, font_scale, color, 2, cv2.LINE_AA)

                # Add feedback with clear text
                y_pos = line_height*7
                for fb in feedback:
                    wrapped_text = [fb[i:i+40] for i in range(0, len(fb), 40)]  # Wrap text at 40 chars
                    for line in wrapped_text:
                        cv2.putText(score_box, line, (20, y_pos),
                                  cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 165, 255), 2, cv2.LINE_AA)
                        y_pos += int(line_height * 0.8)

                # Add scoring criteria explanation
                y_pos += line_height
                cv2.putText(score_box, "Scoring Criteria:", (20, y_pos),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

                y_pos += line_height
                cv2.putText(score_box, "3 pts: Angle 90-110 degrees", (20, y_pos),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 1, cv2.LINE_AA)

                y_pos += line_height
                cv2.putText(score_box, "2 pts: Angle 110-150 degrees", (20, y_pos),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 200, 200), 1, cv2.LINE_AA)

                y_pos += line_height
                cv2.putText(score_box, "1 pt: Angle > 150 degrees", (20, y_pos),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 1, cv2.LINE_AA)

                # Place score box in top right of the new canvas
                new_img[0:orig_height, orig_width:new_width] = score_box

                # Add visual explanation in bottom half
                vis_box = np.zeros((orig_height, new_width, 3), dtype=np.uint8)

                # Fill the visualization box with a dark background
                cv2.rectangle(vis_box, (0, 0), (new_width, orig_height), (30, 30, 30), -1)

                # Draw simplified figures showing angle measurement
                margin = 30
                box_width = new_width - 2*margin
                box_height = orig_height - 2*margin

                # Function to draw a stick figure with angle
                def draw_angle_figure(img, center_x, y, leg_angle, score):
                    # Define points
                    hip_y = y + 100
                    hip_x = center_x

                    # Calculate shoulder position
                    shoulder_length = 80
                    shoulder_x = hip_x
                    shoulder_y = hip_y - shoulder_length

                    # Calculate ankle position based on angle
                    leg_length = 120
                    # Convert angle from "leg-trunk" format to actual drawing angle
                    drawing_angle = 180 - leg_angle
                    ankle_x = int(hip_x + leg_length * np.cos(np.radians(drawing_angle)))
                    ankle_y = int(hip_y - leg_length * np.sin(np.radians(drawing_angle)))

                    # Draw trunk (vertical line)
                    cv2.line(img, (hip_x, hip_y), (shoulder_x, shoulder_y), (255, 255, 255), 2)

                    # Draw leg line
                    cv2.line(img, (hip_x, hip_y), (ankle_x, ankle_y), (255, 255, 255), 2)

                    # Draw angle arc
                    radius = 30
                    start_angle = 270  # Vertical up
                    end_angle = 270 - leg_angle  # Based on leg-trunk angle

                    # Draw angle text
                    text_x = hip_x + int(radius * 1.5 * np.cos(np.radians((start_angle + end_angle)/2)))
                    text_y = hip_y - int(radius * 1.5 * np.sin(np.radians((start_angle + end_angle)/2)))
                    cv2.putText(img, f"{leg_angle}°", (text_x, text_y),
                              cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2, cv2.LINE_AA)

                    # Add score label
                    score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                    cv2.putText(img, f"Score: {score}", (hip_x - 40, hip_y + 40),
                              cv2.FONT_HERSHEY_SIMPLEX, 0.7, score_color, 2, cv2.LINE_AA)

                # Draw figures showing different angles and scores
                figure_width = (box_width) // 3
                draw_angle_figure(vis_box, margin + figure_width//2, margin, 105, 3)  # Score 3 (105°)
                draw_angle_figure(vis_box, margin + figure_width + figure_width//2, margin, 135, 2)  # Score 2 (135°)
                draw_angle_figure(vis_box, margin + 2*figure_width + figure_width//2, margin, 165, 1)  # Score 1 (165°)

                # Add title for visualization
                cv2.putText(vis_box, "Visual Angle Guide", (new_width//2 - 150, 30),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, cv2.LINE_AA)

                # Place visualization box in bottom half of the new canvas
                new_img[orig_height:new_height, 0:new_width] = vis_box

                # Save the enlarged composite image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

            else:
                # No pose detected - create a simple error image
                new_width = img.shape[1] * 2
                new_height = img.shape[0] * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)

                # Place original image in top left
                new_img[0:img.shape[0], 0:img.shape[1]] = img

                # Add error message
                cv2.putText(new_img, "NO POSE DETECTED", (new_width//4, new_height//4),
                          cv2.FONT_HERSHEY_SIMPLEX, 2.0, (0, 0, 255), 3, cv2.LINE_AA)

                # Save the enlarged image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

                print(f"No pose detected in {img_file}")

    # Generate and save summary report
    if score_results:
        # Calculate statistics
        scores = [r['score'] for r in score_results]
        angles = [r['leg_trunk_angle'] for r in score_results]
        avg_score = sum(scores) / len(scores)
        avg_angle = sum(angles) / len(angles)

        # Count by score
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1

        # Create summary report
        report_path = os.path.join(output_folder, "score_summary.txt")
        with open(report_path, 'w') as f:
            f.write(f"Straight Leg Raise Test Score Summary\n")
            f.write(f"====================================\n\n")
            f.write(f"Images analyzed: {len(score_results)}\n")
            f.write(f"Average score: {avg_score:.2f}/3\n")
            f.write(f"Average leg-trunk angle: {avg_angle:.2f} degrees\n\n")
            f.write(f"Score distribution:\n")
            for score, count in score_counts.items():
                percentage = (count / len(scores)) * 100 if scores else 0
                f.write(f"  Score {score}: {count} images ({percentage:.1f}%)\n")

            f.write("\nDetailed results:\n")
            for i, result in enumerate(score_results):
                f.write(f"\n{i+1}. {result['file']}\n")
                f.write(f"   Score: {result['score']}/3\n")
                f.write(f"   Angle: {result['leg_trunk_angle']:.1f} degrees\n")
                f.write(f"   Feedback: {', '.join(result['feedback'])}\n")

        print(f"\nReport saved to {report_path}")

        # Create score distribution chart
        plt.figure(figsize=(10, 6))
        plt.bar(score_counts.keys(), score_counts.values(), color='green')
        plt.title('Straight Leg Raise Test Score Distribution')
        plt.xlabel('Score')
        plt.ylabel('Number of Images')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)

        # Save chart
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)

        # Create angle histogram
        plt.figure(figsize=(10, 6))
        plt.hist(angles, bins=2, color='blue', alpha=0.7)
        plt.title('Leg-Trunk Angle Distribution')
        plt.xlabel('Angle (degrees)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)

        # Add vertical lines for scoring thresholds
        plt.axvline(x=110, color='green', linestyle='--', label='120° (Score 3-2 boundary)')
        plt.axvline(x=150, color='orange', linestyle='--', label='150° (Score 2-1 boundary)')
        plt.legend()

        # Save angle chart
        angle_chart_path = os.path.join(output_folder, "angle_distribution.png")
        plt.savefig(angle_chart_path)
        print(f"Charts saved to {output_folder}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# Example usage
if __name__ == "__main__":
    # Set these paths to your folders
    input_folder = "ll"
    output_folder = "ll5"

    # Process the images
    results = process_image_folder(input_folder, output_folder)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm

# 计算向量间夹角函数
def calculate_angle_between_vectors(v1, v2):
    v1, v2 = np.array(v1), np.array(v2)
    v1_norm = v1 / np.linalg.norm(v1)
    v2_norm = v2 / np.linalg.norm(v2)
    dot_product = np.clip(np.dot(v1_norm, v2_norm), -1.0, 1.0)
    return np.arccos(dot_product) * 180.0 / np.pi

# 计算俯卧撑角度
def calculate_pushup_angle(landmarks, mp_pose):
    # 获取关键点
    left_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
    right_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
    left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x, landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
    right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

    # 计算中点
    mid_ankle = [(left_ankle[0] + right_ankle[0])/2, (left_ankle[1] + right_ankle[1])/2]
    mid_hip = [(left_hip[0] + right_hip[0])/2, (left_hip[1] + right_hip[1])/2]
    mid_wrist = [(left_wrist[0] + right_wrist[0])/2, (left_wrist[1] + right_wrist[1])/2]

    # 计算向量
    ankle_to_hip = [mid_hip[0] - mid_ankle[0], mid_hip[1] - mid_ankle[1]]
    ankle_to_hand = [mid_wrist[0] - mid_ankle[0], mid_wrist[1] - mid_ankle[1]]

    # 计算夹角
    return calculate_angle_between_vectors(ankle_to_hip, ankle_to_hand)

# 评分函数
def score_pushup(landmarks, mp_pose):
    pushup_angle = calculate_pushup_angle(landmarks, mp_pose)

    # 评分逻辑
    if pushup_angle > 35:
        score = 3
        feedback = f"Excellent! Angle: {pushup_angle:.1f}° (>35°)"
    elif pushup_angle >= 15:
        score = 2
        feedback = f"Good position. Angle: {pushup_angle:.1f}° (15-35°)"
    else:
        score = 1
        feedback = f"Improve hand position. Angle: {pushup_angle:.1f}° (<15°)"

    return {
        'score': score,
        'feedback': feedback,
        'pushup_angle': pushup_angle
    }

# 主处理函数
def process_image_folder(input_folder, output_folder):
    # 创建输出文件夹
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 设置MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # 获取图像列表
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []

    # 处理每张图像
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            # 读取图像
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # 处理图像
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # 检查是否检测到姿势
            if results.pose_landmarks:
                # 获取图像尺寸
                orig_height, orig_width, _ = img.shape

                # 创建更大的画布
                new_width = orig_width * 2
                new_height = orig_height * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)

                # 绘制姿势关键点
                mp_drawing.draw_landmarks(img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # 评分
                evaluation = score_pushup(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                feedback = evaluation['feedback']
                angle = evaluation['pushup_angle']

                # 记录分数
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'pushup_angle': angle
                })

                # 将原始图像放在左上角
                new_img[0:orig_height, 0:orig_width] = img

                # 在原始图像上绘制角度线
                landmarks = results.pose_landmarks.landmark
                mid_ankle_x = int(((landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x +
                                  landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x) / 2) * orig_width)
                mid_ankle_y = int(((landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y +
                                  landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y) / 2) * orig_height)
                mid_hip_x = int(((landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x +
                                landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x) / 2) * orig_width)
                mid_hip_y = int(((landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y +
                                landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y) / 2) * orig_height)
                mid_wrist_x = int(((landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x +
                                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x) / 2) * orig_width)
                mid_wrist_y = int(((landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y +
                                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y) / 2) * orig_height)

                # 绘制线条
                cv2.line(new_img, (mid_ankle_x, mid_ankle_y), (mid_hip_x, mid_hip_y), (0, 255, 255), 2)
                cv2.line(new_img, (mid_ankle_x, mid_ankle_y), (mid_wrist_x, mid_wrist_y), (0, 255, 255), 2)

                # 添加角度文字
                cv2.putText(new_img, f"{angle:.1f}°", (mid_ankle_x + 10, mid_ankle_y - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2, cv2.LINE_AA)

                # 添加评分信息
                score_box = np.zeros((orig_height, orig_width, 3), dtype=np.uint8)
                cv2.rectangle(score_box, (0, 0), (orig_width, orig_height), (30, 30, 30), -1)

                # 添加标题和分数
                cv2.putText(score_box, "Pushup Position Test", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, cv2.LINE_AA)

                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(score_box, f"SCORE: {score}/3", (20, 100),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.5, score_color, 2, cv2.LINE_AA)

                # 添加反馈
                cv2.putText(score_box, feedback, (20, 160),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 165, 255), 2, cv2.LINE_AA)

                # 添加评分标准
                cv2.putText(score_box, "Scoring Criteria:", (20, 220),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

                cv2.putText(score_box, "3 pts: Angle > 35°", (20, 260),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 1, cv2.LINE_AA)

                cv2.putText(score_box, "2 pts: Angle 15-35°", (20, 300),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 200, 200), 1, cv2.LINE_AA)

                cv2.putText(score_box, "1 pt: Angle < 15°", (20, 340),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 1, cv2.LINE_AA)

                # 将评分框放在右上角
                new_img[0:orig_height, orig_width:new_width] = score_box

                # 添加简化的视觉指南
                vis_box = np.zeros((orig_height, new_width, 3), dtype=np.uint8)
                cv2.rectangle(vis_box, (0, 0), (new_width, orig_height), (30, 30, 30), -1)

                # 添加标题
                cv2.putText(vis_box, "Measured Angle:", (new_width//2 - 100, 50),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2, cv2.LINE_AA)

                # 绘制简单的示意图
                center_x = new_width // 2
                ankle_x = center_x - 100
                ankle_y = 150
                hip_x = center_x
                hip_y = 80
                hand_x = center_x + 100
                hand_y = 150

                # 绘制点和标签
                cv2.circle(vis_box, (ankle_x, ankle_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (hip_x, hip_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (hand_x, hand_y), 5, (255, 255, 255), -1)

                cv2.putText(vis_box, "Ankle", (ankle_x - 40, ankle_y + 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1, cv2.LINE_AA)
                cv2.putText(vis_box, "Hip", (hip_x - 15, hip_y - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1, cv2.LINE_AA)
                cv2.putText(vis_box, "Hand", (hand_x + 10, hand_y + 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1, cv2.LINE_AA)

                # 绘制线条
                cv2.line(vis_box, (ankle_x, ankle_y), (hip_x, hip_y), (0, 255, 255), 2)
                cv2.line(vis_box, (ankle_x, ankle_y), (hand_x, hand_y), (0, 255, 255), 2)

                # 将视觉指南放在底部
                new_img[orig_height:new_height, 0:new_width] = vis_box

                # 保存图像
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

            else:
                # 未检测到姿势
                print(f"No pose detected in {img_file}")
                # 创建错误图像
                new_width = img.shape[1] * 2
                new_height = img.shape[0] * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)
                new_img[0:img.shape[0], 0:img.shape[1]] = img
                cv2.putText(new_img, "NO POSE DETECTED", (new_width//4, new_height//4),
                          cv2.FONT_HERSHEY_SIMPLEX, 2.0, (0, 0, 255), 3, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

    # 生成总结报告
    if score_results:
        # 计算统计数据
        scores = [r['score'] for r in score_results]
        angles = [r['pushup_angle'] for r in score_results]
        avg_score = sum(scores) / len(scores)
        avg_angle = sum(angles) / len(angles)

        # 创建分数分布图
        plt.figure(figsize=(10, 6))
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1
        plt.bar(score_counts.keys(), score_counts.values(), color='green')
        plt.title('Pushup Position Test Score Distribution')
        plt.xlabel('Score')
        plt.ylabel('Number of Images')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)

        # 创建角度分布图
        plt.figure(figsize=(10, 6))
        plt.hist(angles, bins=10, color='blue', alpha=0.7)
        plt.title('Angle Distribution')
        plt.xlabel('Angle (degrees)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.axvline(x=15, color='orange', linestyle='--', label='15° (Score 1-2 boundary)')
        plt.axvline(x=35, color='green', linestyle='--', label='35° (Score 2-3 boundary)')
        plt.legend()
        angle_chart_path = os.path.join(output_folder, "angle_distribution.png")
        plt.savefig(angle_chart_path)

        print(f"Charts saved to {output_folder}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# 示例用法
if __name__ == "__main__":
    input_folder = "pppp"
    output_folder = "pppp1"
    results = process_image_folder(input_folder, output_folder)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm

# 计算三点角度函数
def calculate_angle(a, b, c):
    """计算三点之间的角度"""
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)

    if angle > 180.0:
        angle = 360-angle

    return angle

def identify_raised_leg(landmarks, mp_pose):
    """确定哪条腿是抬起的腿"""
    left_ankle_y = landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y
    right_ankle_y = landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y

    # Y坐标小的腿（在屏幕上更高）被视为抬起的腿
    is_right_leg_raised = right_ankle_y < left_ankle_y

    return is_right_leg_raised

def calculate_leg_angle(landmarks, mp_pose, is_right_leg_raised):
    """计算抬起腿的脚踝-膝盖-髋部角度"""
    if is_right_leg_raised:
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]
        knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
    else:
        ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
        knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
              landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]

    # 计算膝盖角度
    angle = calculate_angle(ankle, knee, hip)

    return angle

def score_trunk_rotation(landmarks, mp_pose):
    """
    评估转动稳定动作
    - 3分: 腿部角度 > 160度
    - 2分: 腿部角度在130-160度之间
    - 1分: 腿部角度 < 130度
    """
    # 确定抬起的腿
    is_right_leg_raised = identify_raised_leg(landmarks, mp_pose)

    # 计算腿部角度
    leg_angle = calculate_leg_angle(landmarks, mp_pose, is_right_leg_raised)

    # 评分逻辑
    if leg_angle > 160:
        score = 3
        feedback = f"Excellent! Leg angle: {leg_angle:.1f}° (>160°)"
    elif leg_angle >= 130:
        score = 2
        feedback = f"Good rotation. Leg angle: {leg_angle:.1f}° (130-160°)"
    else:
        score = 1
        feedback = f"Limited rotation. Leg angle: {leg_angle:.1f}° (<130°)"

    return {
        'score': score,
        'feedback': feedback,
        'leg_angle': leg_angle,
        'is_right_leg_raised': is_right_leg_raised
    }

def process_image_folder(input_folder, output_folder):
    """处理文件夹中的所有图像"""
    # 创建输出文件夹
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 设置MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # 获取图像列表
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []

    # 处理每张图像
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            # 读取图像
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # 处理图像
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # 检查是否检测到姿势
            if results.pose_landmarks:
                # 获取图像尺寸
                orig_height, orig_width, _ = img.shape

                # 创建更大的画布
                new_width = orig_width * 2
                new_height = orig_height * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)

                # 绘制姿势关键点
                mp_drawing.draw_landmarks(img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # 评分
                evaluation = score_trunk_rotation(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                feedback = evaluation['feedback']
                leg_angle = evaluation['leg_angle']
                is_right_leg_raised = evaluation['is_right_leg_raised']

                # 记录分数
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'leg_angle': leg_angle
                })

                # 将原始图像放在左上角
                new_img[0:orig_height, 0:orig_width] = img

                # 在原始图像上绘制角度线
                landmarks = results.pose_landmarks.landmark

                if is_right_leg_raised:
                    ankle = (int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x * orig_width),
                            int(landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y * orig_height))
                    knee = (int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x * orig_width),
                           int(landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y * orig_height))
                    hip = (int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x * orig_width),
                          int(landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y * orig_height))
                    leg_label = "Right Leg (Raised)"
                else:
                    ankle = (int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x * orig_width),
                            int(landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y * orig_height))
                    knee = (int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x * orig_width),
                           int(landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y * orig_height))
                    hip = (int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x * orig_width),
                          int(landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y * orig_height))
                    leg_label = "Left Leg (Raised)"

                # 绘制线条
                cv2.line(new_img, ankle, knee, (0, 255, 255), 2)  # 脚踝到膝盖
                cv2.line(new_img, knee, hip, (0, 255, 255), 2)   # 膝盖到髋部

                # 添加角度文字
                cv2.putText(new_img, f"{leg_angle:.1f}°", (knee[0] + 10, knee[1] - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2, cv2.LINE_AA)

                # 添加评分信息
                score_box = np.zeros((orig_height, orig_width, 3), dtype=np.uint8)
                cv2.rectangle(score_box, (0, 0), (orig_width, orig_height), (30, 30, 30), -1)

                # 添加标题和分数
                cv2.putText(score_box, "Trunk Rotation Test", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, cv2.LINE_AA)

                cv2.putText(score_box, f"Testing: {leg_label}", (20, 80),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(score_box, f"SCORE: {score}/3", (20, 130),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.5, score_color, 2, cv2.LINE_AA)

                # 添加反馈
                cv2.putText(score_box, feedback, (20, 180),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 165, 255), 2, cv2.LINE_AA)

                # 添加评分标准
                cv2.putText(score_box, "Scoring Criteria:", (20, 230),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2, cv2.LINE_AA)

                cv2.putText(score_box, "3 pts: Leg angle > 160°", (20, 270),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 1, cv2.LINE_AA)

                cv2.putText(score_box, "2 pts: Leg angle 130-160°", (20, 310),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 200, 200), 1, cv2.LINE_AA)

                cv2.putText(score_box, "1 pt: Leg angle < 130°", (20, 350),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 1, cv2.LINE_AA)

                # 将评分框放在右上角
                new_img[0:orig_height, orig_width:new_width] = score_box

                # 添加简化的视觉指南
                vis_box = np.zeros((orig_height, new_width, 3), dtype=np.uint8)
                cv2.rectangle(vis_box, (0, 0), (new_width, orig_height), (30, 30, 30), -1)

                # 添加标题
                cv2.putText(vis_box, "Trunk Rotation: Measured Angle", (new_width//2 - 180, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2, cv2.LINE_AA)

                # 绘制示例视觉指南
                center_x = new_width // 2

                # 绘制示例1 - 3分 (>160°)
                ex1_x = center_x - 300
                ex1_y = 150
                hip1_x, hip1_y = ex1_x, ex1_y
                knee1_x, knee1_y = hip1_x, hip1_y + 60
                ankle1_x, ankle1_y = knee1_x + 10, knee1_y + 100 # 165度角

                cv2.circle(vis_box, (hip1_x, hip1_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (knee1_x, knee1_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (ankle1_x, ankle1_y), 5, (255, 255, 255), -1)

                cv2.line(vis_box, (hip1_x, hip1_y), (knee1_x, knee1_y), (0, 255, 0), 2)
                cv2.line(vis_box, (knee1_x, knee1_y), (ankle1_x, ankle1_y), (0, 255, 0), 2)

                cv2.putText(vis_box, "Score 3 (>160°)", (hip1_x - 70, hip1_y - 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2, cv2.LINE_AA)

                # 绘制示例2 - 2分 (130-160°)
                ex2_x = center_x
                ex2_y = 150
                hip2_x, hip2_y = ex2_x, ex2_y
                knee2_x, knee2_y = hip2_x, hip2_y + 60
                ankle2_x, ankle2_y = knee2_x + 40, knee2_y + 90 # 约145度角

                cv2.circle(vis_box, (hip2_x, hip2_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (knee2_x, knee2_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (ankle2_x, ankle2_y), 5, (255, 255, 255), -1)

                cv2.line(vis_box, (hip2_x, hip2_y), (knee2_x, knee2_y), (0, 200, 200), 2)
                cv2.line(vis_box, (knee2_x, knee2_y), (ankle2_x, ankle2_y), (0, 200, 200), 2)

                cv2.putText(vis_box, "Score 2 (130-160°)", (hip2_x - 70, hip2_y - 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 200, 200), 2, cv2.LINE_AA)

                # 绘制示例3 - 1分 (<130°)
                ex3_x = center_x + 300
                ex3_y = 150
                hip3_x, hip3_y = ex3_x, ex3_y
                knee3_x, knee3_y = hip3_x, hip3_y + 60
                ankle3_x, ankle3_y = knee3_x + 60, knee3_y + 60 # 约120度角

                cv2.circle(vis_box, (hip3_x, hip3_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (knee3_x, knee3_y), 5, (255, 255, 255), -1)
                cv2.circle(vis_box, (ankle3_x, ankle3_y), 5, (255, 255, 255), -1)

                cv2.line(vis_box, (hip3_x, hip3_y), (knee3_x, knee3_y), (0, 0, 255), 2)
                cv2.line(vis_box, (knee3_x, knee3_y), (ankle3_x, ankle3_y), (0, 0, 255), 2)

                cv2.putText(vis_box, "Score 1 (<130°)", (hip3_x - 70, hip3_y - 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2, cv2.LINE_AA)

                # 将视觉指南放在底部
                new_img[orig_height:new_height, 0:new_width] = vis_box

                # 保存图像
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

            else:
                # 未检测到姿势
                print(f"No pose detected in {img_file}")
                # 创建错误图像
                new_width = img.shape[1] * 2
                new_height = img.shape[0] * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)
                new_img[0:img.shape[0], 0:img.shape[1]] = img
                cv2.putText(new_img, "NO POSE DETECTED", (new_width//4, new_height//4),
                          cv2.FONT_HERSHEY_SIMPLEX, 2.0, (0, 0, 255), 3, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

    # 生成总结报告
    if score_results:
        # 计算统计数据
        scores = [r['score'] for r in score_results]
        angles = [r['leg_angle'] for r in score_results]
        avg_score = sum(scores) / len(scores)
        avg_angle = sum(angles) / len(angles)

        # 创建分数分布图
        plt.figure(figsize=(10, 6))
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1
        plt.bar(score_counts.keys(), score_counts.values(), color='green')
        plt.title('Trunk Rotation Test Score Distribution')
        plt.xlabel('Score')
        plt.ylabel('Number of Images')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)

        # 创建角度分布图
        plt.figure(figsize=(10, 6))
        plt.hist(angles, bins=10, color='blue', alpha=0.7)
        plt.title('Leg Angle Distribution')
        plt.xlabel('Angle (degrees)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.axvline(x=130, color='orange', linestyle='--', label='130° (Score 1-2 boundary)')
        plt.axvline(x=160, color='green', linestyle='--', label='160° (Score 2-3 boundary)')
        plt.legend()
        angle_chart_path = os.path.join(output_folder, "angle_distribution.png")
        plt.savefig(angle_chart_path)

        print(f"Charts saved to {output_folder}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# 示例用法
if __name__ == "__main__":
    input_folder = "r"
    output_folder = "r1"
    results = process_image_folder(input_folder, output_folder)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm

# Helper functions
def calculate_distance(a, b):
    """Calculate the Euclidean distance between two points"""
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)

def calculate_shoulder_width(landmarks, mp_pose):
    """Calculate the shoulder width"""
    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]

    return calculate_distance(left_shoulder, right_shoulder)

def detect_hands_behind_back(landmarks, mp_pose):
    """Detect if hands are behind the back"""
    # Check if wrists are close to each other and behind torso
    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                 landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]

    # Calculate average shoulder position
    mid_shoulder_x = (left_shoulder[0] + right_shoulder[0]) / 2

    # Check if both wrists are behind the shoulder plane
    is_left_behind = left_wrist[0] < mid_shoulder_x
    is_right_behind = right_wrist[0] < mid_shoulder_x

    # Check if wrists are at similar depth
    wrist_distance = calculate_distance(left_wrist, right_wrist)
    shoulder_width = calculate_distance(left_shoulder, right_shoulder)

    # For test to be valid, wrists should be somewhat close and behind the body
    return (is_left_behind or is_right_behind) and wrist_distance < shoulder_width * 4

def score_shoulder_mobility(landmarks, mp_pose):
    """
    Score the shoulder mobility test based on ratio of hand distance to shoulder width:
    - 3 points: Hand distance/shoulder width < 2
    - 2 points: Hand distance/shoulder width between 2 and 2.5
    - 1 point: Hand distance/shoulder width > 2.5
    """
    # Get wrist positions for hand measurement
    left_wrist = [landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x,
                 landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y]
    right_wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
                  landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

    # Calculate distance between hands
    hand_distance = calculate_distance(left_wrist, right_wrist)

    # Calculate shoulder width as reference
    shoulder_width = calculate_shoulder_width(landmarks, mp_pose)

    # Check if hands appear to be behind back
    is_behind_back = detect_hands_behind_back(landmarks, mp_pose)

    # Calculate the ratio of hand distance to shoulder width
    distance_ratio = hand_distance / shoulder_width if shoulder_width > 0 else float('inf')

    # Determine score based on the ratio
    if not is_behind_back:
        score = 1
        feedback = "Hands not positioned correctly behind back"
    elif distance_ratio < 1.7:  # Ratio < 2
        score = 3
        feedback = f"Excellent mobility! Ratio: {distance_ratio:.2f}"
    elif distance_ratio <= 2:  # 2 <= Ratio <= 2.5
        score = 2
        feedback = f"Good mobility. Ratio: {distance_ratio:.2f}"
    else:  # Ratio > 2
        score = 1
        feedback = f"Limited mobility. Ratio: {distance_ratio:.2f}"

    return {
        'score': score,
        'feedback': feedback,
        'hand_distance': hand_distance,
        'shoulder_width': shoulder_width,
        'distance_value': distance_ratio,  # Now this value is the ratio
        'is_behind_back': is_behind_back
    }

def process_image_folder(input_folder, output_folder):
    """Process all images in a folder"""
    # Create output folder
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []

    # Process each image
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Process image
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Get image dimensions
                orig_height, orig_width, _ = img.shape

                # Create larger canvas
                new_width = orig_width * 2
                new_height = orig_height * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)

                # Draw pose landmarks
                mp_drawing.draw_landmarks(img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # Score the pose
                evaluation = score_shoulder_mobility(results.pose_landmarks.landmark, mp_pose)
                score = evaluation['score']
                feedback = evaluation['feedback']
                hand_distance = evaluation['hand_distance']
                shoulder_width = evaluation['shoulder_width']
                distance_value = evaluation['distance_value']
                is_behind_back = evaluation['is_behind_back']

                # Record scores
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'distance_value': distance_value
                })

                # Place original image in top left
                new_img[0:orig_height, 0:orig_width] = img

                # Draw key points on original image
                landmarks = results.pose_landmarks.landmark

                # Get key positions
                left_wrist = (int(landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].x * orig_width),
                             int(landmarks[mp_pose.PoseLandmark.LEFT_WRIST.value].y * orig_height))
                right_wrist = (int(landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x * orig_width),
                              int(landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y * orig_height))
                left_shoulder = (int(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x * orig_width),
                                int(landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y * orig_height))
                right_shoulder = (int(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x * orig_width),
                                 int(landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y * orig_height))

                # Highlight wrists
                cv2.circle(new_img, left_wrist, 10, (0, 165, 255), -1)
                cv2.circle(new_img, right_wrist, 10, (0, 165, 255), -1)

                # Draw line between wrists
                cv2.line(new_img, left_wrist, right_wrist, (0, 165, 255), 2)

                # Draw shoulder width reference
                cv2.line(new_img, left_shoulder, right_shoulder, (0, 255, 0), 2)

                # Add distance text
                mid_wrist_x = (left_wrist[0] + right_wrist[0]) // 2
                mid_wrist_y = (left_wrist[1] + right_wrist[1]) // 2

                cv2.putText(new_img, f"{distance_value:.2f} units",
                          (mid_wrist_x, mid_wrist_y - 10),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 2, cv2.LINE_AA)

                # Add score box
                score_box = np.zeros((orig_height, orig_width, 3), dtype=np.uint8)
                cv2.rectangle(score_box, (0, 0), (orig_width, orig_height), (30, 30, 30), -1)

                # Add title and score
                cv2.putText(score_box, "Shoulder Mobility Test", (20, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, cv2.LINE_AA)

                test_status = "Valid Test" if is_behind_back else "Invalid Position"
                status_color = (0, 255, 0) if is_behind_back else (0, 0, 255)
                cv2.putText(score_box, test_status, (20, 80),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, status_color, 2, cv2.LINE_AA)

                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(score_box, f"SCORE: {score}/3", (20, 130),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.5, score_color, 2, cv2.LINE_AA)

                # Add measurements
                cv2.putText(score_box, f"Hand distance: {distance_value:.2f} units", (20, 180),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)

                cv2.putText(score_box, f"Shoulder width: {shoulder_width:.2f} units", (20, 210),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)

                # Add feedback
                cv2.putText(score_box, feedback, (20, 250),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 1, cv2.LINE_AA)

                # Add scoring criteria - exact thresholds
                cv2.putText(score_box, "Scoring Criteria:", (20, 290),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 1, cv2.LINE_AA)

                cv2.putText(score_box, "3 pts: Distance < 2 units", (20, 320),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1, cv2.LINE_AA)

                cv2.putText(score_box, "2 pts: Distance 2-3 units", (20, 350),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 200, 200), 1, cv2.LINE_AA)

                cv2.putText(score_box, "1 pt: Distance > 3 units", (20, 380),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 1, cv2.LINE_AA)

                # Place score box in top right
                new_img[0:orig_height, orig_width:new_width] = score_box

                # Add visual guide in bottom half
                vis_box = np.zeros((orig_height, new_width, 3), dtype=np.uint8)
                cv2.rectangle(vis_box, (0, 0), (new_width, orig_height), (30, 30, 30), -1)

                # Add title
                cv2.putText(vis_box, "Shoulder Mobility Test Guide", (new_width//2 - 180, 40),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2, cv2.LINE_AA)

                # Draw a visual representing the test
                center_x = new_width // 2
                y_pos = 100

                # Draw an outline of a person from behind
                # Head
                head_x = center_x
                head_y = y_pos + 30
                cv2.circle(vis_box, (head_x, head_y), 25, (100, 100, 100), 2)

                # Shoulders
                left_sh_x = center_x - 60
                right_sh_x = center_x + 60
                sh_y = y_pos + 80

                # Draw shoulder line
                cv2.line(vis_box, (left_sh_x, sh_y), (right_sh_x, sh_y), (100, 100, 100), 2)

                # Draw torso
                torso_bottom_y = y_pos + 200
                cv2.line(vis_box, (left_sh_x, sh_y), (center_x - 40, torso_bottom_y), (100, 100, 100), 2)
                cv2.line(vis_box, (right_sh_x, sh_y), (center_x + 40, torso_bottom_y), (100, 100, 100), 2)

                # Draw spine
                cv2.line(vis_box, (center_x, head_y + 25), (center_x, torso_bottom_y), (100, 100, 100), 2)

                # Draw left arm reaching down
                left_elbow_x = left_sh_x - 15
                left_elbow_y = sh_y + 60
                left_wrist_x = center_x - 15
                left_wrist_y = torso_bottom_y - 30

                cv2.line(vis_box, (left_sh_x, sh_y), (left_elbow_x, left_elbow_y), (0, 165, 255), 2)
                cv2.line(vis_box, (left_elbow_x, left_elbow_y), (left_wrist_x, left_wrist_y), (0, 165, 255), 2)

                # Draw right arm reaching up
                right_elbow_x = right_sh_x + 15
                right_elbow_y = sh_y + 60
                right_wrist_x = center_x + 15
                right_wrist_y = y_pos + 160

                cv2.line(vis_box, (right_sh_x, sh_y), (right_elbow_x, right_elbow_y), (0, 165, 255), 2)
                cv2.line(vis_box, (right_elbow_x, right_elbow_y), (right_wrist_x, right_wrist_y), (0, 165, 255), 2)

                # Draw wrists
                cv2.circle(vis_box, (left_wrist_x, left_wrist_y), 5, (0, 165, 255), -1)
                cv2.circle(vis_box, (right_wrist_x, right_wrist_y), 5, (0, 165, 255), -1)

                # Draw a line showing the distance between hands
                cv2.line(vis_box, (left_wrist_x, left_wrist_y), (right_wrist_x, right_wrist_y), (255, 0, 0), 2)

                # Visual examples for the three scores
                ex_y = torso_bottom_y + 60

                # Score 3 example (< 2 units)
                s3_x = center_x - 250
                cv2.putText(vis_box, "Score 3: < 2 units", (s3_x - 70, ex_y - 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1, cv2.LINE_AA)
                cv2.circle(vis_box, (s3_x, ex_y), 5, (0, 255, 0), -1)
                cv2.circle(vis_box, (s3_x + 30, ex_y), 5, (0, 255, 0), -1)
                cv2.line(vis_box, (s3_x, ex_y), (s3_x + 30, ex_y), (0, 255, 0), 2)
                cv2.putText(vis_box, "1.5 units", (s3_x - 15, ex_y + 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1, cv2.LINE_AA)

                # Score 2 example (2-3 units)
                s2_x = center_x
                cv2.putText(vis_box, "Score 2: 2-3 units", (s2_x - 70, ex_y - 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 200, 200), 1, cv2.LINE_AA)
                cv2.circle(vis_box, (s2_x, ex_y), 5, (0, 200, 200), -1)
                cv2.circle(vis_box, (s2_x + 50, ex_y), 5, (0, 200, 200), -1)
                cv2.line(vis_box, (s2_x, ex_y), (s2_x + 50, ex_y), (0, 200, 200), 2)
                cv2.putText(vis_box, "2.5 units", (s2_x + 5, ex_y + 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 200, 200), 1, cv2.LINE_AA)

                # Score 1 example (> 3 units)
                s1_x = center_x + 250
                cv2.putText(vis_box, "Score 1: > 3 units", (s1_x - 70, ex_y - 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 1, cv2.LINE_AA)
                cv2.circle(vis_box, (s1_x, ex_y), 5, (0, 0, 255), -1)
                cv2.circle(vis_box, (s1_x + 80, ex_y), 5, (0, 0, 255), -1)
                cv2.line(vis_box, (s1_x, ex_y), (s1_x + 80, ex_y), (0, 0, 255), 2)
                cv2.putText(vis_box, "4 units", (s1_x + 25, ex_y + 20),
                          cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)

                # Place visual guide in bottom half
                new_img[orig_height:new_height, 0:new_width] = vis_box

                # Save image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                # Create error image
                new_width = img.shape[1] * 2
                new_height = img.shape[0] * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)
                new_img[0:img.shape[0], 0:img.shape[1]] = img
                cv2.putText(new_img, "NO POSE DETECTED", (new_width//4, new_height//4),
                          cv2.FONT_HERSHEY_SIMPLEX, 2.0, (0, 0, 255), 3, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

    # Generate summary report
    if score_results:
        # Calculate statistics
        scores = [r['score'] for r in score_results]
        distances = [r['distance_value'] for r in score_results]
        avg_score = sum(scores) / len(scores)
        avg_distance = sum(distances) / len(distances)

        # Create score distribution chart
        plt.figure(figsize=(10, 6))
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1
        plt.bar(score_counts.keys(), score_counts.values(), color='green')
        plt.title('Shoulder Mobility Test Score Distribution')
        plt.xlabel('Score')
        plt.ylabel('Number of Images')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)

        # Create distance distribution chart with updated thresholds
        plt.figure(figsize=(10, 6))
        plt.hist(distances, bins=10, color='blue', alpha=0.7)
        plt.title('Hand Distance Distribution')
        plt.xlabel('Distance (units)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.axvline(x=2.0, color='green', linestyle='--', label='2.0 (Score 2-3 boundary)')
        plt.axvline(x=3.0, color='orange', linestyle='--', label='3.0 (Score 1-2 boundary)')
        plt.legend()
        distance_chart_path = os.path.join(output_folder, "distance_distribution.png")
        plt.savefig(distance_chart_path)

        print(f"Charts saved to {output_folder}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# Example usage
if __name__ == "__main__":
    input_folder = "j"
    output_folder = "j5"
    results = process_image_folder(input_folder, output_folder)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm

# Helper functions
def calculate_distance(a, b):
    """Calculate the Euclidean distance between two points"""
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)

def calculate_shoulder_width(landmarks, mp_pose):
    """Calculate the shoulder width"""
    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]

    return calculate_distance(left_shoulder, right_shoulder)

def score_squat_mobility(landmarks, mp_pose):
    """
    Score the squat mobility test:
    - 3 points: Hip Y >= average knee Y (deep squat, hip at or below knee level)
    - 2 points: Hip Y < average knee Y and difference (avg_knee_y - mid_hip_y) > 0.5*shoulder_width
                but <= shoulder_width
    - 1 point: Hip Y < average knee Y and difference (avg_knee_y - mid_hip_y) > shoulder width
    """
    # Get hip positions
    left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
    right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]

    # Calculate middle hip point
    mid_hip_x = (left_hip[0] + right_hip[0]) / 2
    mid_hip_y = (left_hip[1] + right_hip[1]) / 2

    # Get knee positions
    left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
    right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                 landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]

    # Calculate average knee Y position
    avg_knee_y = (left_knee[1] + right_knee[1]) / 2

    # Calculate shoulder width as reference for scoring
    shoulder_width = calculate_shoulder_width(landmarks, mp_pose)

    # Calculate difference between average knee Y and hip Y
    # Note: In image coordinates, Y increases downward
    # If hip is higher than knees (poor squat), avg_knee_y > mid_hip_y
    hip_knee_diff = avg_knee_y - mid_hip_y

    # Determine score based on the criteria
    if mid_hip_y >= avg_knee_y:  # Hip Y >= average knee Y (hip at or below knee level)
        score = 3
        feedback = "Excellent squat mobility! Hip is at or below knee level."
    elif hip_knee_diff > 0.4 *shoulder_width:  # Hip is much higher than knees
        score = 1
        feedback = f"Limited squat mobility. Hip is above knee level by more than shoulder width."
    elif hip_knee_diff > 0.1 * shoulder_width:  # Hip is moderately higher than knees
        score = 2
        feedback = f"Moderate squat mobility. Hip is above knee level by 0.5-1x shoulder width."
    else:  # Hip is slightly higher than knees
        score = 3
        feedback = f"Good squat mobility. Hip is only slightly above knee level."

    return {
        'score': score,
        'feedback': feedback,
        'hip_knee_diff': hip_knee_diff,
        'shoulder_width': shoulder_width,
        'hip_height': mid_hip_y,
        'avg_knee_height': avg_knee_y,
        'hip_knee_ratio': hip_knee_diff / shoulder_width if shoulder_width > 0 else float('inf'),
        'mid_hip_x': mid_hip_x,
        'left_knee': left_knee,
        'right_knee': right_knee
    }

def process_squat_images(input_folder, output_folder):
    """Process all squat assessment images in a folder"""
    # Create output folder
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []

    # Process each image
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Process image
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Get image dimensions
                orig_height, orig_width, _ = img.shape

                # Create larger canvas
                new_width = orig_width * 2
                new_height = orig_height * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)

                # Draw pose landmarks
                mp_drawing.draw_landmarks(img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # Score the pose
                evaluation = score_squat_mobility(results.pose_landmarks.landmark, mp_pose)

                score = evaluation['score']
                feedback = evaluation['feedback']
                shoulder_width = evaluation['shoulder_width']
                hip_knee_diff = evaluation['hip_knee_diff']
                mid_hip_x = evaluation['mid_hip_x']
                mid_hip_y = evaluation['hip_height']
                avg_knee_y = evaluation['avg_knee_height']
                left_knee = evaluation['left_knee']
                right_knee = evaluation['right_knee']
                hip_knee_ratio = evaluation['hip_knee_ratio']

                # Record scores
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'hip_knee_diff': hip_knee_diff,
                    'hip_knee_ratio': hip_knee_ratio
                })

                # Place original image in top left
                new_img[0:orig_height, 0:orig_width] = img

                # Draw key points on original image
                landmarks = results.pose_landmarks.landmark

                # Draw hip point
                hip_point = (int(mid_hip_x * orig_width), int(mid_hip_y * orig_height))
                cv2.circle(new_img, hip_point, 10, (0, 0, 255), -1)

                # Draw knee points
                left_knee_point = (int(left_knee[0] * orig_width), int(left_knee[1] * orig_height))
                right_knee_point = (int(right_knee[0] * orig_width), int(right_knee[1] * orig_height))
                cv2.circle(new_img, left_knee_point, 10, (0, 255, 0), -1)
                cv2.circle(new_img, right_knee_point, 10, (0, 255, 0), -1)

                # Draw average knee height line
                avg_knee_line_start = (int(left_knee[0] * orig_width), int(avg_knee_y * orig_height))
                avg_knee_line_end = (int(right_knee[0] * orig_width), int(avg_knee_y * orig_height))
                cv2.line(new_img, avg_knee_line_start, avg_knee_line_end, (0, 255, 0), 2)

                # Draw height difference if hip is above knees
                if mid_hip_y > avg_knee_y:
                    hip_to_knee_line_start = (hip_point[0], hip_point[1])
                    hip_to_knee_line_end = (hip_point[0], int(avg_knee_y * orig_height))
                    cv2.line(new_img, hip_to_knee_line_start, hip_to_knee_line_end, (0, 165, 255), 2)

                    # Add difference text
                    diff_text_x = hip_point[0] + 10
                    diff_text_y = (hip_point[1] + int(avg_knee_y * orig_height)) // 2
                    cv2.putText(new_img, f"Diff: {hip_knee_diff:.2f} ({hip_knee_ratio:.2f}x SW)",
                            (diff_text_x, diff_text_y),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 2, cv2.LINE_AA)

                # Add score box
                score_box = np.zeros((orig_height, orig_width, 3), dtype=np.uint8)
                cv2.rectangle(score_box, (0, 0), (orig_width, orig_height), (30, 30, 30), -1)

                # Add title and score to score box
                cv2.putText(score_box, "Squat Mobility Test", (20, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, cv2.LINE_AA)

                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(score_box, f"SCORE: {score}/3", (20, 130),
                          cv2.FONT_HERSHEY_SIMPLEX, 1.5, score_color, 2, cv2.LINE_AA)

                # Add measurements
                cv2.putText(score_box, f"Hip-Knee Diff: {hip_knee_diff:.2f} units", (20, 150),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)

                cv2.putText(score_box, f"Hip-Knee/Shoulder Width: {hip_knee_ratio:.2f}", (20, 180),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)

                cv2.putText(score_box, f"Shoulder width: {shoulder_width:.2f} units", (20, 210),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)

                # Add feedback
                cv2.putText(score_box, feedback, (20, 250),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 1, cv2.LINE_AA)

                # Add scoring criteria
                cv2.putText(score_box, "Scoring Criteria:", (20, 290),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 1, cv2.LINE_AA)

                cv2.putText(score_box, "3 pts: Hip at/below knee level", (20, 320),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1, cv2.LINE_AA)

                cv2.putText(score_box, "2 pts: Hip > knee by 0.5-1x SW", (20, 350),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 200, 200), 1, cv2.LINE_AA)

                cv2.putText(score_box, "1 pt: Hip > knee by > 1x SW", (20, 380),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 1, cv2.LINE_AA)

                # Place score box in top right
                new_img[0:orig_height, orig_width:new_width] = score_box

                # Add visual guide in bottom half
                vis_box = np.zeros((orig_height, new_width, 3), dtype=np.uint8)
                cv2.rectangle(vis_box, (0, 0), (new_width, orig_height), (30, 30, 30), -1)

                # Add title
                cv2.putText(vis_box, "Squat Mobility Test Guide", (new_width//2 - 180, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2, cv2.LINE_AA)

                # Draw a stick figure in squat position
                center_x = new_width // 2
                y_pos = 100

                # Draw head
                head_y = y_pos + 30
                cv2.circle(vis_box, (center_x, head_y), 20, (100, 100, 100), 2)

                # Draw torso for good squat (score 3)
                good_torso_bottom_y = y_pos + 130
                cv2.line(vis_box, (center_x, head_y + 20), (center_x, good_torso_bottom_y), (0, 255, 0), 2)

                # Draw legs for good squat
                knee_x_offset = 40
                good_knee_y = good_torso_bottom_y + 40
                good_ankle_y = good_knee_y + 40

                # Left leg
                cv2.line(vis_box, (center_x, good_torso_bottom_y),
                       (center_x - knee_x_offset, good_knee_y), (0, 255, 0), 2)
                cv2.line(vis_box, (center_x - knee_x_offset, good_knee_y),
                       (center_x - knee_x_offset, good_ankle_y), (0, 255, 0), 2)

                # Right leg
                cv2.line(vis_box, (center_x, good_torso_bottom_y),
                       (center_x + knee_x_offset, good_knee_y), (0, 255, 0), 2)
                cv2.line(vis_box, (center_x + knee_x_offset, good_knee_y),
                       (center_x + knee_x_offset, good_ankle_y), (0, 255, 0), 2)

                # Mark knees
                cv2.circle(vis_box, (center_x - knee_x_offset, good_knee_y), 5, (0, 255, 0), -1)
                cv2.circle(vis_box, (center_x + knee_x_offset, good_knee_y), 5, (0, 255, 0), -1)

                # Mark hip
                cv2.circle(vis_box, (center_x, good_torso_bottom_y), 7, (0, 255, 0), -1)

                # Draw average knee line
                cv2.line(vis_box, (center_x - knee_x_offset - 20, good_knee_y),
                       (center_x + knee_x_offset + 20, good_knee_y), (0, 255, 0), 1)

                # Add label for good squat
                cv2.putText(vis_box, "Score 3: Hip at/below knee level", (center_x - 150, good_ankle_y + 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1, cv2.LINE_AA)

                # Draw poor squat example (score 1)
                x_offset = 200
                poor_torso_bottom_y = y_pos + 80  # Higher hip position
                poor_stick_x = center_x - x_offset

                # Draw head
                cv2.circle(vis_box, (poor_stick_x, head_y), 20, (100, 100, 100), 2)

                # Draw torso
                cv2.line(vis_box, (poor_stick_x, head_y + 20), (poor_stick_x, poor_torso_bottom_y), (0, 0, 255), 2)

                # Draw legs
                poor_knee_y = poor_torso_bottom_y + 40
                poor_ankle_y = poor_knee_y + 40

                # Left leg
                cv2.line(vis_box, (poor_stick_x, poor_torso_bottom_y),
                       (poor_stick_x - knee_x_offset, poor_knee_y), (0, 0, 255), 2)
                cv2.line(vis_box, (poor_stick_x - knee_x_offset, poor_knee_y),
                       (poor_stick_x - knee_x_offset, poor_ankle_y), (0, 0, 255), 2)

                # Right leg
                cv2.line(vis_box, (poor_stick_x, poor_torso_bottom_y),
                       (poor_stick_x + knee_x_offset, poor_knee_y), (0, 0, 255), 2)
                cv2.line(vis_box, (poor_stick_x + knee_x_offset, poor_knee_y),
                       (poor_stick_x + knee_x_offset, poor_ankle_y), (0, 0, 255), 2)

                # Mark knees
                cv2.circle(vis_box, (poor_stick_x - knee_x_offset, poor_knee_y), 5, (0, 0, 255), -1)
                cv2.circle(vis_box, (poor_stick_x + knee_x_offset, poor_knee_y), 5, (0, 0, 255), -1)

                # Mark hip
                cv2.circle(vis_box, (poor_stick_x, poor_torso_bottom_y), 7, (0, 0, 255), -1)

                # Draw average knee line
                cv2.line(vis_box, (poor_stick_x - knee_x_offset - 20, poor_knee_y),
                       (poor_stick_x + knee_x_offset + 20, poor_knee_y), (0, 0, 255), 1)

                # Draw hip-knee difference
                cv2.line(vis_box, (poor_stick_x, poor_torso_bottom_y),
                       (poor_stick_x, poor_knee_y), (0, 165, 255), 2)

                # Add label for poor squat
                cv2.putText(vis_box, "Score 1: Hip > knee by > 1x SW", (poor_stick_x - 150, poor_ankle_y + 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 1, cv2.LINE_AA)

                # Draw moderate squat example (score 2)
                moderate_torso_bottom_y = y_pos + 105  # Moderate hip position
                moderate_stick_x = center_x + x_offset

                # Draw head
                cv2.circle(vis_box, (moderate_stick_x, head_y), 20, (100, 100, 100), 2)

                # Draw torso
                cv2.line(vis_box, (moderate_stick_x, head_y + 20),
                       (moderate_stick_x, moderate_torso_bottom_y), (0, 200, 200), 2)

                # Draw legs
                moderate_knee_y = moderate_torso_bottom_y + 40
                moderate_ankle_y = moderate_knee_y + 40

                # Left leg
                cv2.line(vis_box, (moderate_stick_x, moderate_torso_bottom_y),
                       (moderate_stick_x - knee_x_offset, moderate_knee_y), (0, 200, 200), 2)
                cv2.line(vis_box, (moderate_stick_x - knee_x_offset, moderate_knee_y),
                       (moderate_stick_x - knee_x_offset, moderate_ankle_y), (0, 200, 200), 2)

                # Right leg
                cv2.line(vis_box, (moderate_stick_x, moderate_torso_bottom_y),
                       (moderate_stick_x + knee_x_offset, moderate_knee_y), (0, 200, 200), 2)
                cv2.line(vis_box, (moderate_stick_x + knee_x_offset, moderate_knee_y),
                       (moderate_stick_x + knee_x_offset, moderate_ankle_y), (0, 200, 200), 2)

                # Mark knees
                cv2.circle(vis_box, (moderate_stick_x - knee_x_offset, moderate_knee_y), 5, (0, 200, 200), -1)
                cv2.circle(vis_box, (moderate_stick_x + knee_x_offset, moderate_knee_y), 5, (0, 200, 200), -1)

                # Mark hip
                cv2.circle(vis_box, (moderate_stick_x, moderate_torso_bottom_y), 7, (0, 200, 200), -1)

                # Draw average knee line
                cv2.line(vis_box, (moderate_stick_x - knee_x_offset - 20, moderate_knee_y),
                       (moderate_stick_x + knee_x_offset + 20, moderate_knee_y), (0, 200, 200), 1)

                # Draw hip-knee difference
                cv2.line(vis_box, (moderate_stick_x, moderate_torso_bottom_y),
                       (moderate_stick_x, moderate_knee_y), (0, 165, 255), 2)

                # Add label for moderate squat
                cv2.putText(vis_box, "Score 2: Hip > knee by 0.5-1x SW", (moderate_stick_x - 170, moderate_ankle_y + 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 200, 200), 1, cv2.LINE_AA)

                # Place visual guide in bottom half
                new_img[orig_height:new_height, 0:new_width] = vis_box

                # Save image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                # Create error image
                new_width = img.shape[1] * 2
                new_height = img.shape[0] * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)
                new_img[0:img.shape[0], 0:img.shape[1]] = img
                cv2.putText(new_img, "NO POSE DETECTED", (new_width//4, new_height//4),
                          cv2.FONT_HERSHEY_SIMPLEX, 2.0, (0, 0, 255), 3, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

    # Generate summary report
    if score_results:
        # Calculate statistics
        scores = [r['score'] for r in score_results]
        diffs = [r['hip_knee_diff'] for r in score_results]
        ratios = [r['hip_knee_ratio'] for r in score_results]
        avg_score = sum(scores) / len(scores)
        avg_diff = sum(diffs) / len(diffs)
        avg_ratio = sum(ratios) / len(ratios)

        # Create score distribution chart
        plt.figure(figsize=(10, 6))
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1
        plt.bar(score_counts.keys(), score_counts.values(), color='green')
        plt.title('Squat Mobility Test Score Distribution')
        plt.xlabel('Score')
        plt.ylabel('Number of Images')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)

        # Create hip-knee difference distribution chart
        plt.figure(figsize=(10, 6))
        plt.hist(diffs, bins=10, color='blue', alpha=0.7)
        plt.title('Hip-Knee Difference Distribution')
        plt.xlabel('Difference (units)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.axvline(x=0, color='green', linestyle='--', label='Hip at knee level (Score 3)')
        plt.axvline(x=0.5, color='yellow', linestyle='--', label='0.5x SW (Score boundary)')
        plt.axvline(x=1.0, color='red', linestyle='--', label='1.0x SW (Score boundary)')
        plt.legend()
        distance_chart_path = os.path.join(output_folder, "hip_knee_diff_distribution.png")
        plt.savefig(distance_chart_path)

        print(f"Charts saved to {output_folder}")
        print(f"Average score: {avg_score:.2f}")
        print(f"Average hip-knee difference: {avg_diff:.2f}")
        print(f"Average hip-knee/shoulder width ratio: {avg_ratio:.2f}")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# Example usage
if __name__ == "__main__":
    input_folder = "s"
    output_folder = "s3"
    results = process_squat_images(input_folder, output_folder)
```

```python
import cv2
import mediapipe as mp
import numpy as np
import os
import matplotlib.pyplot as plt
from tqdm import tqdm

# Helper functions
def calculate_distance(a, b):
    """Calculate the Euclidean distance between two points"""
    a = np.array(a)
    b = np.array(b)
    return np.linalg.norm(a - b)

def calculate_shoulder_width(landmarks, mp_pose):
    """Calculate the shoulder width"""
    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]

    return calculate_distance(left_shoulder, right_shoulder)

def calculate_angle(a, b, c):
    """Calculate the angle between three points"""
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    ba = a - b
    bc = c - b

    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    cosine_angle = np.clip(cosine_angle, -1.0, 1.0)  # Ensure the value is in range [-1, 1]

    angle = np.arccos(cosine_angle)
    angle = np.degrees(angle)

    return angle

def score_lunge_mobility(landmarks, mp_pose):
    """
    Score the lunge mobility test:
    - 1 point: Upper body tilted left/right >5 degrees from vertical
    - 2 points: Upper body vertical (±5 degrees) but potential forward tilt
    - 3 points: Upper body vertical (±5 degrees) with optimal alignment

    Note: Since we only have frontal view, the forward tilt assessment is approximate
    """
    # Get key points
    left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
                    landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
    right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
                     landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
    left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,
               landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
    right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,
                landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
    left_knee = [landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].x,
                landmarks[mp_pose.PoseLandmark.LEFT_KNEE.value].y]
    right_knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
                 landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]

    # Calculate mid-points
    mid_shoulder_x = (left_shoulder[0] + right_shoulder[0]) / 2
    mid_shoulder_y = (left_shoulder[1] + right_shoulder[1]) / 2
    mid_hip_x = (left_hip[0] + right_hip[0]) / 2
    mid_hip_y = (left_hip[1] + right_hip[1]) / 2

    # Calculate lateral (left-right) tilt angle
    # Use the angle between vertical line and the line connecting mid-hips to mid-shoulders
    vertical_top = [mid_hip_x, 0]  # Point directly above hip
    vertical_bottom = [mid_hip_x, 1]  # Point directly below hip

    # Angle between vertical line and shoulder-hip line
    shoulder_hip_line = [mid_shoulder_x, mid_shoulder_y]
    vertical_line_top = vertical_top
    origin_point = [mid_hip_x, mid_hip_y]

    lateral_tilt_angle = calculate_angle(vertical_line_top, origin_point, shoulder_hip_line)

    # Adjust angle to be negative if tilted left, positive if tilted right
    if mid_shoulder_x < mid_hip_x:  # Tilted left
        lateral_tilt_angle = -lateral_tilt_angle

    # For forward tilt, we'll use the relative position of shoulders compared to hips
    # This is a very rough approximation since we only have frontal view
    # Look for a disproportionate shortening of the torso length compared to shoulder width

    # Calculate torso length (shoulder to hip)
    torso_length = abs(mid_shoulder_y - mid_hip_y)
    shoulder_width = calculate_distance(left_shoulder, right_shoulder)

    # Calculate a torso ratio (length/width)
    # A lower ratio might indicate forward tilt as torso appears shorter from front view when leaning forward
    torso_ratio = torso_length / shoulder_width if shoulder_width > 0 else 0

    # Threshold for potential forward tilt (needs calibration with actual data)
    forward_tilt_threshold = 1.5  # This is an arbitrary value, needs adjustment

    # Another indicator: front knee position relative to front ankle
    # In a good lunge, front knee should be over ankle, not too far forward
    # Determine which leg is forward (lower knee Y position)
    if left_knee[1] < right_knee[1]:  # Left leg is forward
        front_knee = left_knee
        front_ankle = [landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
                      landmarks[mp_pose.PoseLandmark.LEFT_ANKLE.value].y]
    else:  # Right leg is forward
        front_knee = right_knee
        front_ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
                      landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]

    # Calculate horizontal distance between knee and ankle
    knee_ankle_x_diff = abs(front_knee[0] - front_ankle[0])
    knee_forward_indicator = knee_ankle_x_diff / shoulder_width if shoulder_width > 0 else 0

    # Combine indicators for potential forward lean assessment
    forward_tilt_indicator = (1 / torso_ratio if torso_ratio > 0 else 0) + knee_forward_indicator

    # Score based on criteria
    if abs(lateral_tilt_angle) > 5:
        score = 1
        feedback = f"Upper body is tilted to the {'left' if lateral_tilt_angle < 0 else 'right'} by {abs(lateral_tilt_angle):.1f} degrees. Stand straighter with upper body aligned vertically."
    elif forward_tilt_indicator > forward_tilt_threshold:
        score = 2
        feedback = "Upper body appears to have some forward tilt. Try maintaining a more upright posture during the lunge."
    else:
        score = 3
        feedback = "Excellent lunge position with good upper body alignment."

    return {
        'score': score,
        'feedback': feedback,
        'lateral_tilt_angle': lateral_tilt_angle,
        'forward_tilt_indicator': forward_tilt_indicator,
        'torso_ratio': torso_ratio,
        'knee_forward_indicator': knee_forward_indicator,
        'mid_shoulder': [mid_shoulder_x, mid_shoulder_y],
        'mid_hip': [mid_hip_x, mid_hip_y]
    }

def process_lunge_images(input_folder, output_folder):
    """Process all lunge assessment images in a folder"""
    # Create output folder
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Set up MediaPipe
    mp_drawing = mp.solutions.drawing_utils
    mp_pose = mp.solutions.pose

    # Get list of image files
    image_files = [f for f in os.listdir(input_folder)
                  if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    print(f"Found {len(image_files)} images to process")
    score_results = []

    # Process each image
    with mp_pose.Pose(static_image_mode=True, model_complexity=2, min_detection_confidence=0.5) as pose:
        for img_file in tqdm(image_files):
            # Read image
            img_path = os.path.join(input_folder, img_file)
            img = cv2.imread(img_path)
            if img is None:
                print(f"Error reading image: {img_path}")
                continue

            # Process image
            img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            results = pose.process(img_rgb)

            # Check if pose landmarks were detected
            if results.pose_landmarks:
                # Get image dimensions
                orig_height, orig_width, _ = img.shape

                # Create larger canvas
                new_width = orig_width * 2
                new_height = orig_height * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)

                # Draw pose landmarks
                mp_drawing.draw_landmarks(img, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

                # Score the pose
                evaluation = score_lunge_mobility(results.pose_landmarks.landmark, mp_pose)

                score = evaluation['score']
                feedback = evaluation['feedback']
                lateral_tilt_angle = evaluation['lateral_tilt_angle']
                forward_tilt_indicator = evaluation['forward_tilt_indicator']
                mid_shoulder = evaluation['mid_shoulder']
                mid_hip = evaluation['mid_hip']

                # Record scores
                score_results.append({
                    'file': img_file,
                    'score': score,
                    'feedback': feedback,
                    'lateral_tilt_angle': lateral_tilt_angle,
                    'forward_tilt_indicator': forward_tilt_indicator
                })

                # Place original image in top left
                new_img[0:orig_height, 0:orig_width] = img

                # Draw key points on original image
                landmarks = results.pose_landmarks.landmark

                # Convert normalized coordinates to pixel coordinates
                mid_shoulder_px = (int(mid_shoulder[0] * orig_width), int(mid_shoulder[1] * orig_height))
                mid_hip_px = (int(mid_hip[0] * orig_width), int(mid_hip[1] * orig_height))

                # Draw vertical reference line from hip
                vertical_top_px = (mid_hip_px[0], 0)
                vertical_bottom_px = (mid_hip_px[0], orig_height)
                cv2.line(new_img, vertical_top_px, vertical_bottom_px, (0, 255, 0), 2)

                # Draw line connecting mid-shoulder to mid-hip
                cv2.line(new_img, mid_shoulder_px, mid_hip_px, (0, 165, 255), 2)

                # Mark mid points
                cv2.circle(new_img, mid_shoulder_px, 10, (0, 0, 255), -1)
                cv2.circle(new_img, mid_hip_px, 10, (255, 0, 0), -1)

                # Add tilt angle text
                tilt_text_pos = (mid_hip_px[0] + 20, (mid_shoulder_px[1] + mid_hip_px[1]) // 2)
                cv2.putText(new_img, f"Tilt: {lateral_tilt_angle:.1f}°",
                        tilt_text_pos, cv2.FONT_HERSHEY_SIMPLEX,
                        0.7, (0, 165, 255), 2, cv2.LINE_AA)

                # Add score box
                score_box = np.zeros((orig_height, orig_width, 3), dtype=np.uint8)
                cv2.rectangle(score_box, (0, 0), (orig_width, orig_height), (30, 30, 30), -1)

                # Add title and score to score box
                cv2.putText(score_box, "Lunge Assessment", (20, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 2, cv2.LINE_AA)

                score_color = (0, 255, 0) if score == 3 else ((0, 200, 200) if score == 2 else (0, 0, 255))
                cv2.putText(score_box, f"SCORE: {score}/3", (20, 130),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.5, score_color, 2, cv2.LINE_AA)

                # Add measurements
                cv2.putText(score_box, f"Lateral Tilt: {lateral_tilt_angle:.1f}°", (20, 180),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)

                cv2.putText(score_box, f"Forward Tilt Indicator: {forward_tilt_indicator:.2f}", (20, 210),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 1, cv2.LINE_AA)

                # Add feedback
                cv2.putText(score_box, feedback, (20, 250),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 1, cv2.LINE_AA)

                # Add scoring criteria
                cv2.putText(score_box, "Scoring Criteria:", (20, 290),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 1, cv2.LINE_AA)

                cv2.putText(score_box, "1 pt: Upper body tilt >5°", (20, 320),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 1, cv2.LINE_AA)

                cv2.putText(score_box, "2 pts: Vertical (±5°) but forward lean", (20, 350),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 200, 200), 1, cv2.LINE_AA)

                cv2.putText(score_box, "3 pts: Vertical (±5°) & upright", (20, 380),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1, cv2.LINE_AA)

                # Place score box in top right
                new_img[0:orig_height, orig_width:new_width] = score_box

                # Add visual guide in bottom half
                vis_box = np.zeros((orig_height, new_width, 3), dtype=np.uint8)
                cv2.rectangle(vis_box, (0, 0), (new_width, orig_height), (30, 30, 30), -1)

                # Add title
                cv2.putText(vis_box, "Lunge Assessment Guide", (new_width//2 - 180, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2, cv2.LINE_AA)

                # Draw example figures
                center_x = new_width // 2
                y_pos = 100

                # Example 1: Good lunge (score 3)
                good_x = center_x - 300
                # Draw head
                head_y = y_pos + 30
                cv2.circle(vis_box, (good_x, head_y), 20, (100, 100, 100), 2)

                # Draw torso - straight vertical
                torso_bottom_y = y_pos + 130
                cv2.line(vis_box, (good_x, head_y + 20), (good_x, torso_bottom_y), (0, 255, 0), 2)

                # Draw legs
                front_knee_x = good_x + 40
                back_knee_x = good_x - 40
                knee_y = torso_bottom_y + 40
                ankle_y = knee_y + 40

                # Front leg (bent)
                cv2.line(vis_box, (good_x, torso_bottom_y), (front_knee_x, knee_y), (0, 255, 0), 2)
                cv2.line(vis_box, (front_knee_x, knee_y), (front_knee_x, ankle_y), (0, 255, 0), 2)

                # Back leg (straight)
                cv2.line(vis_box, (good_x, torso_bottom_y), (back_knee_x, knee_y), (0, 255, 0), 2)
                cv2.line(vis_box, (back_knee_x, knee_y), (back_knee_x - 20, ankle_y), (0, 255, 0), 2)

                # Draw vertical reference line
                cv2.line(vis_box, (good_x, head_y - 30), (good_x, ankle_y + 20), (255, 255, 255), 1, cv2.LINE_AA)

                # Add label
                cv2.putText(vis_box, "Score 3: Perfect vertical alignment", (good_x - 120, ankle_y + 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 1, cv2.LINE_AA)

                # Example 2: Moderate lunge (score 2)
                moderate_x = center_x
                # Draw head
                cv2.circle(vis_box, (moderate_x, head_y), 20, (100, 100, 100), 2)

                # Draw torso - straight but with some forward tilt (from side view)
                # We'll simulate this with slightly shorter appearance
                mod_torso_bottom_y = y_pos + 120
                cv2.line(vis_box, (moderate_x, head_y + 20), (moderate_x, mod_torso_bottom_y), (0, 200, 200), 2)

                # Draw legs
                front_knee_x = moderate_x + 40
                back_knee_x = moderate_x - 40

                # Front leg (bent)
                cv2.line(vis_box, (moderate_x, mod_torso_bottom_y), (front_knee_x, knee_y), (0, 200, 200), 2)
                cv2.line(vis_box, (front_knee_x, knee_y), (front_knee_x - 10, ankle_y), (0, 200, 200), 2)

                # Back leg (straight)
                cv2.line(vis_box, (moderate_x, mod_torso_bottom_y), (back_knee_x, knee_y), (0, 200, 200), 2)
                cv2.line(vis_box, (back_knee_x, knee_y), (back_knee_x - 20, ankle_y), (0, 200, 200), 2)

                # Draw vertical reference line
                cv2.line(vis_box, (moderate_x, head_y - 30), (moderate_x, ankle_y + 20), (255, 255, 255), 1, cv2.LINE_AA)

                # Add label
                cv2.putText(vis_box, "Score 2: Vertical but with forward lean", (moderate_x - 150, ankle_y + 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 200, 200), 1, cv2.LINE_AA)

                # Example 3: Poor lunge (score 1)
                poor_x = center_x + 300
                # Draw head
                cv2.circle(vis_box, (poor_x, head_y), 20, (100, 100, 100), 2)

                # Draw torso - tilted
                poor_torso_bottom_x = poor_x - 30  # Tilted to the left
                cv2.line(vis_box, (poor_x, head_y + 20), (poor_torso_bottom_x, torso_bottom_y), (0, 0, 255), 2)

                # Draw legs
                front_knee_x = poor_torso_bottom_x + 40
                back_knee_x = poor_torso_bottom_x - 40

                # Front leg (bent)
                cv2.line(vis_box, (poor_torso_bottom_x, torso_bottom_y), (front_knee_x, knee_y), (0, 0, 255), 2)
                cv2.line(vis_box, (front_knee_x, knee_y), (front_knee_x, ankle_y), (0, 0, 255), 2)

                # Back leg (straight)
                cv2.line(vis_box, (poor_torso_bottom_x, torso_bottom_y), (back_knee_x, knee_y), (0, 0, 255), 2)
                cv2.line(vis_box, (back_knee_x, knee_y), (back_knee_x - 20, ankle_y), (0, 0, 255), 2)

                # Draw vertical reference line
                cv2.line(vis_box, (poor_x, head_y - 30), (poor_x, ankle_y + 20), (255, 255, 255), 1, cv2.LINE_AA)

                # Draw tilt angle
                cv2.putText(vis_box, "~8°", (poor_x - 40, torso_bottom_y - 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 165, 255), 1, cv2.LINE_AA)

                # Add label
                cv2.putText(vis_box, "Score 1: Tilted >5° to side", (poor_x - 120, ankle_y + 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 1, cv2.LINE_AA)

                # Place visual guide in bottom half
                new_img[orig_height:new_height, 0:new_width] = vis_box

                # Save image
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

            else:
                # No pose detected
                print(f"No pose detected in {img_file}")
                # Create error image
                new_width = img.shape[1] * 2
                new_height = img.shape[0] * 2
                new_img = np.zeros((new_height, new_width, 3), dtype=np.uint8)
                new_img[0:img.shape[0], 0:img.shape[1]] = img
                cv2.putText(new_img, "NO POSE DETECTED", (new_width//4, new_height//4),
                          cv2.FONT_HERSHEY_SIMPLEX, 2.0, (0, 0, 255), 3, cv2.LINE_AA)
                output_path = os.path.join(output_folder, f"scored_{img_file}")
                cv2.imwrite(output_path, new_img)

    # Generate summary report
    if score_results:
        # Calculate statistics
        scores = [r['score'] for r in score_results]
        tilt_angles = [abs(r['lateral_tilt_angle']) for r in score_results]
        forward_indicators = [r['forward_tilt_indicator'] for r in score_results]

        avg_score = sum(scores) / len(scores)
        avg_tilt = sum(tilt_angles) / len(tilt_angles)

        # Create score distribution chart
        plt.figure(figsize=(10, 6))
        score_counts = {1: 0, 2: 0, 3: 0}
        for s in scores:
            score_counts[s] += 1
        plt.bar(score_counts.keys(), score_counts.values(), color='green')
        plt.title('Lunge Assessment Score Distribution')
        plt.xlabel('Score')
        plt.ylabel('Number of Images')
        plt.xticks([1, 2, 3])
        plt.grid(axis='y', alpha=0.3)
        chart_path = os.path.join(output_folder, "score_distribution.png")
        plt.savefig(chart_path)

        # Create tilt angle distribution chart
        plt.figure(figsize=(10, 6))
        plt.hist(tilt_angles, bins=10, color='blue', alpha=0.7)
        plt.title('Lateral Tilt Angle Distribution')
        plt.xlabel('Absolute Tilt Angle (degrees)')
        plt.ylabel('Number of Images')
        plt.grid(True, alpha=0.3)
        plt.axvline(x=5, color='red', linestyle='--', label='5° Threshold')
        plt.legend()
        angle_chart_path = os.path.join(output_folder, "tilt_angle_distribution.png")
        plt.savefig(angle_chart_path)

        print(f"Charts saved to {output_folder}")
        print(f"Average score: {avg_score:.2f}")
        print(f"Average absolute lateral tilt: {avg_tilt:.2f} degrees")

    print(f"\nAll images processed and saved to {output_folder}")
    return score_results

# Example usage
if __name__ == "__main__":
    input_folder = "llll"
    output_folder = "lunge_results"
    results = process_lunge_images(input_folder, output_folder)
```
